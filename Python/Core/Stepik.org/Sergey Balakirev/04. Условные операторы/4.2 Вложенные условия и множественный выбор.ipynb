# step_04
Вводится вес боксера-любителя (в кг, в виде вещественного числа). Известно, что вес таков, что боксер может быть отнесен к одной из весовых категорий: <br>
1) легкий вес – до 60 кг (включительно); <br>
2) первый полусредний вес – до 64 кг (включительно); <br>
3) полусредний вес – до 69 кг (включительно); <br>
4) остальные - более 69 кг. <br>

Вывести на экран номер категории, в которой будет выступать боксер.

## sample_input:
62.4

## sample_output:
2


````python
# date: 28.08.2022

weight = float(input())

if weight <= 60:
    print("1")
elif 61 <= weight <= 64:
    print("2")
elif 65 <= weight <= 69:
    print("3")
elif weight > 69:
    print("4")
````

# step_07
Дата некоторого дня характеризуется двумя натуральными числами: m (порядковый номер месяца) и n (число). По введенным m и n (в одну строку через пробел) определить:

а) дату предыдущего дня (принять, что m и n не характеризуют 1 января);
б) дату следующего дня (принять, что m и n не характеризуют 31 декабря).

В задаче принять, что год не является високосным. Вывести предыдущую дату и следующую дату (в формате: mm.dd, где m - число месяца; d - номер дня) в одну строчку через пробел.

P.S. Число дней в месяцах не високосного года, начиная с января: 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31

## sample_input:
8 31

## sample_output:
08.30 09.01


````python
# date: 28.08.2022

month, day = map(int, input().split())

if month == 2:
    if day == 1:
        print(f"{str(month - 1).zfill(2)}.31 {str(month).zfill(2)}.{str(day + 1).zfill(2)}")

    elif day < 28:
        print(f"{str(month).zfill(2)}.{str(day - 1).zfill(2)} {str(month).zfill(2)}.{str(day + 1).zfill(2)}")

    else:
        print(f"{str(month).zfill(2)}.{str(day - 1).zfill(2)} {str(month + 1).zfill(2)}.01")

elif month in [1, 3, 5, 7, 8, 10, 12]:
    if day == 1:
        print(f"{str(month - 1).zfill(2)}.31 {str(month).zfill(2)}.{str(day + 1).zfill(2)}")

    elif day < 31:
        print(f"{str(month).zfill(2)}.{str(day - 1).zfill(2)} {str(month).zfill(2)}.{str(day + 1).zfill(2)}")

    else:
        print(f"{str(month).zfill(2)}.{str(day - 1).zfill(2)} {str(month + 1).zfill(2)}.01")

else:
    if day == 1:
        print(f"{str(month - 1).zfill(2)}.31 {str(month).zfill(2)}.{str(day + 1).zfill(2)}")
    elif day < 30:
        print(f"{str(month).zfill(2)}.{str(day - 1).zfill(2)} {str(month).zfill(2)}.{str(day + 1).zfill(2)}")
        
    else:
        print(f"{str(month).zfill(2)}.{str(day - 1).zfill(2)} {str(month + 1).zfill(2)}.01")
````

# step_08
Вводится целое число k (1 <= k <= 365). Определить, каким днем недели (понедельник, вторник, среда, четверг, пятница, суббота или воскресенье) является k-й день не високосного года, в котором 1 января является понедельником.

## sample_input:
121

## sample_output:
вторник


````python
# date: 28.08.2022

day = int(input()) % 7

match day:
    case 1:
        print("понедельник")
    case 2:
        print("вторник")
    case 3:
        print("среда")
    case 4:
        print("четверг")
    case 5:
        print("пятница")
    case 6:
        print("суббота")
    case 7:
        print("воскресенье")
````

````python
# date: 15.10.2022

weekdays: dict = {1 : 'понедельник',
                  2 : 'вторник',
                  3 : 'среда',
                  4 : 'четверг',
                  5 : 'пятница',
                  6 : 'суббота',
                  7 : 'воскресенье'}

print(weekdays[(int(input()) % 7)])
````
