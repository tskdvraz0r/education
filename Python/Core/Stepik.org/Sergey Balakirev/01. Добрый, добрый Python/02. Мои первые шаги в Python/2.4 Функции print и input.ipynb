# step_10
В магазине продается X синих ручек, Y зеленых, черных в два раза больше, чем синих, а красных в четыре раза больше зеленых. Напишите программу, в которой вводятся целые значения X, Y в одну строку через пробел с помощью команды: <br>

```python
X, Y = map(int, input().split())
```

и выводится на экран общее число ручек в виде целого числа.

## sample_input:
10 20

## sample_output:
130


````python
# date: 31.07.2022

blue_pen, green_pen = map(int, input().split())
black_pen, red_pen = blue_pen * 2, green_pen * 4

print(blue_pen + green_pen + black_pen + red_pen)
````
