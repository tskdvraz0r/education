# step_15 - Предикат делимости
Напишите функцию func(num1, num2), принимающую в качестве аргументов два натуральных числа num1 и num2 и возвращающую значение True если число num1 делится без остатка на число num2 и False в противном случае. <br>

Результатом вывода программы должно быть "делится" (если функция func() вернула True) и "не делится" (если функция func() вернула False).

**Примечание:** <br> 
Следующий программный код:
```python
print(func(10, 2))
print(func(5, 7))
print(func(15, 15))
```

должен выводить:
```python
True
False
True
```

а вся программа должна выводить:
```python
делится
не делится
делится
```

## sample_input:
10 <br>
2 <br>

## sample_output:
делится

````python
# date: 08.01.2023

def func(num1, num2):
    if isinstance(num1, int) and isinstance(num2, int):
        if num1 % num2 == 0:
            return True
        else:
            return False

if func(int(input()), int(input())):
    print("делится")
else:
    print("не делится")
````
