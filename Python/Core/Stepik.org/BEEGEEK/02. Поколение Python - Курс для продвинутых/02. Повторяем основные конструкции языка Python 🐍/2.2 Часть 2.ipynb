# step_01 - Координатные четверти
Дан набор точек на координатной плоскости. Необходимо подсчитать и вывести количество точек, лежащих в каждой координатной четверти.

**Формат входных данных:** <br>
В первой строке записано количество точек. Каждая следующая строка состоит из двух целых чисел — координат точки (сначала абсцисса – $x$, затем ордината – $y$), разделенных символом пробела.

**Формат выходных данных:** <br>
Программа должна вывести количество точек, лежащих в каждой координатной четверти, как в примерах.

**Примечание №1:** <br>
Учтите, что точки, лежащие на осях координат, не принято относить к какой-либо координатной четверти.

<img src="https://ucarecdn.com/48a2a71b-636d-499a-ad55-764e6b6ac7ba/" width=500>

## sample_input:
4 <br>
0 -1 <br>
1 2 <br>
0 9 <br>
-9 -5 <br>

## sample_output:
Первая четверть: 1 <br>
Вторая четверть: 0 <br>
Третья четверть: 1 <br>
Четвертая четверть: 0 <br>

```python
# date: 08.01.2023

# 2.9. Координатные четверти
n = int(input())
first_quarter = 0
second_quarter = 0
third_quarter = 0
fourth_quarter = 0

for i in range(n):
    coordinates = input().split(' ')
    if int(coordinates[0]) > 0 and int(coordinates[1]) > 0:
        first_quarter += 1
    if int(coordinates[0]) < 0 and int(coordinates[1]) > 0:
        second_quarter += 1
    if int(coordinates[0]) < 0 and int(coordinates[1]) < 0:
        third_quarter += 1
    if int(coordinates[0]) > 0 and int(coordinates[1]) < 0:
        fourth_quarter += 1

print(
    f'Первая четверть: {first_quarter}',
    f'Вторая четверть: {second_quarter}',
    f'Третья четверть: {third_quarter}',
    f'Четвертая четверть: {fourth_quarter}',
    sep = '\n'
)
```

# step_02 - Больше предыдущего
На вход программе подается строка текста из натуральных чисел. Из неё формируется список чисел. Напишите программу подсчета количества чисел, которые больше предшествующего им в этом списке числа, то есть, стоят вслед за меньшим числом.

**Формат входных данных:** <br>
На вход программе подается строка текста из разделенных пробелами натуральных чисел.

**Формат выходных данных:** <br>
Программа должна вывести одно число – количество элементов списка, больших предыдущего.

## sample_input:
1 2 3 4 5 <br>

## sample_output:
4 <br>

```python
# date: 08.01.2023

n = input().split(' ')
count = 0

for i in range(len(n) - 1):
    if int(n[i + 1]) > int(n[i]):
        count += 1

print(count)
```

# step_03 - Назад, вперёд и наоборот
На вход программе подается строка текста из натуральных чисел. Из элементов строки формируется список чисел. Напишите программу, которая меняет местами соседние элементы списка ($a[0]$ c $a[1]$, $a[2]$ c $a[3]$ и т.д.). Если в списке нечетное количество элементов, то последний остается на своем месте.

**Формат входных данных:** <br>
На вход программе подается строка текста, содержащая натуральные числа, разделенные пробелами.

**Формат выходных данных:** <br>
Программа должна вывести измененный список, разделяя его элементы одним пробелом.

## sample_input:
1 2 3 4 5 <br>

## sample_output:
2 1 4 3 5 <br>

```python
# date: 08.01.2023

num_str = input().split()

if len(num_str) % 2 == 0:
    num_list = []
    for i in range(0, len(num_str), 2):
        num_list.append(num_str[i + 1])
        num_list.append(num_str[i])
    
elif len(num_str) % 2 != 0:
    num_list = []
    for i in range(0, len(num_str) - 1, 2):
        num_list.append(num_str[i + 1])
        num_list.append(num_str[i])
    
    num_list.append(num_str[-1])

print(*num_list)
```

# step_06 - Произведение чисел
Напишите программу для определения, является ли число произведением двух чисел из данного набора, выводящую результат в виде ответа «ДА» или «НЕТ».

**Формат входных данных:** <br>
В первой строке подаётся число $n$, ($0 < n < 1000$) – количество чисел в наборе. В последующих $n$ строках вводятся целые числа, составляющие набор (могут повторяться). Затем следует целое число, которое является или не является произведением двух каких-то чисел из набора.

**Формат выходных данных:** <br>
Программа должна вывести *«ДА»* или *«НЕТ»* в соответствии с условием задачи.

**Примечание №1** <br>
Само на себя число из набора умножиться не может, другими словами, два множителя должны иметь разные индексы в наборе.

**Примечание №2** <br>
Для решения задачи используйте вложенные циклы.

## sample_input:
3 <br>
33 <br>
17 <br>
35 <br>
999 <br>

## sample_output:
НЕТ <br>

```python
# date: 08.01.2023

from itertools import combinations

arr = [int(input()) for _ in range(int(input()))]
result = int(input())

flag = False
for i in combinations(arr, 2):
    if i[0] * i[1] == result:
        flag = True
        break

if flag:
    print("ДА")
else:
    print("НЕТ")
```

# step_09 - Орел и решка
Дана строка текста, состоящая из букв русского алфавита "О" и "Р". Буква "О" – соответствует выпадению Орла, а буква "Р" – соответствует выпадению Решки. Напишите программу, которая подсчитывает наибольшее количество подряд выпавших Решек.

**Формат входных данных:** <br>
На вход программе подается строка текста, состоящая из букв русского алфавита "О" и "Р".

**Формат выходных данных:** <br>
Программа должна вывести наибольшее количество подряд выпавших Решек.

**Примечание №1:** <br>
Если выпавших Решек нет, то необходимо вывести число 0.

## sample_input:
ОРРОРОРООРРРО <br>

## sample_output:
3 <br>

```python
# date: 08.01.2023

string = input().split("О")

res = 0
for i in range(len(string)):
    if len(string[i]) > res:
        res = len(string[i])
    else:
        continue

print(res)
```
