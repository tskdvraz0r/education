# step_02 - На easy
На вход программе подаются два целых числа $a$ и $b$. Напишите программу, которая выводит:
* сумму чисел $a$ и $b$;
* разность чисел $a$ и $b$;
* произведение чисел $a$ и $b$;
* частное чисел $a$ и $b$;
* целую часть от деления числа $a$ на $b$;
* остаток от деления числа $a$ на $b$;
* корень квадратный из суммы их 10-х степеней: $\sqrt{a^{10} + b^{10}}$.

**Формат входных данных:** <br>
На вход программе подаются два целых числа $a$ и $b$, ($b \neq 0$), каждое на отдельной строке.

**Формат выходных данных:** <br>
Программа должна вывести результаты математических операций в соответствии с условием задачи.

## sample_input:
3 <br>
8 <br>

## sample_output:
11 <br>
-5 <br>
24 <br>
0.375 <br>
0 <br>
3 <br>
32768.90100384814 <br>

```python
# date: 08.01.2023

import math as mt
a, b = int(input()), int(input())
if b != 0:
    print(a + b, a - b, a * b, a / b, a // b, a % b, mt.sqrt(a ** 10 + b ** 10), sep='\n')
```

# step_03 - Индекс массы тела
Напишите программу для вычисления и оценки индекса массы тела (ИМТ) человека. ИМТ показывает, весит человек больше или меньше нормы для своего роста. ИМТ человека рассчитывают по формуле:

$ИМТ = \dfrac{масса, (кг)}{рост(м) \times рост(м)}$, <br>

где масса измеряется в килограммах, а рост — в метрах.

Масса человека считается оптимальной, если его ИМТ находится между *18.5* и *25*. Если ИМТ меньше *18.5*, то считается, что человек весит ниже нормы. Если значение ИМТ больше *25*, то считается, что человек весит больше нормы.

Программа должна вывести "Оптимальная масса", если ИМТ находится между *18.5* и *25* (включительно). *"Недостаточная масса"*, если ИМТ меньше *18.5* и *"Избыточная масса"*, если значение ИМТ больше *25*.

**Формат входных данных:** <br>
На вход программе подается два числа: масса и рост человека, каждое на отдельной строке. Все входные числа являются вещественными, используйте для них тип данных float.

**Формат выходных данных:** <br>
Программа должна вывести текст в соответствии с условием задачи.

## sample_input:
65 <br>
1.75 <br>

## sample_output:
Оптимальная масса <br>

```python
# date: 08.01.2023

w, h = float(input()), float(input())
ibm = w / (h * h)

if ibm < 18.5:
    print('Недостаточная масса')
elif 18.5 <= ibm <= 25:
    print('Оптимальная масса')
elif ibm > 25:
    print('Избыточная масса')
```

# step_07 - Переворот числа
Дано пятизначное или шестизначное натуральное число. Напишите программу, которая изменит порядок его последних пяти цифр на обратный.

**Формат входных данных:** <br>
На вход программе подается одно натуральное пятизначное или шестизначное число.

**Формат выходных данных:** <br>
Программа должна вывести число, которое получится в результате разворота, указанного в условии задачи. Число нужно выводить без незначащих нулей.

## sample_input:
12345 <br>

## sample_output:
54321 <br>

```python
# date: 08.01.2023

num = input()
if len(num) == 5:
    print(int(num[::-1]))
else:
    print(f'{num[0]}{num[5:0:-1]}')
```

# step_08 - Standard American Convention
На вход программе подаётся натуральное число. Напишите программу, которая вставляет в заданное число запятые в соответствии со стандартным американским соглашением о запятых в больших числах.

**Формат входных данных:** <br>
На вход программе подаётся натуральное число $n$, ($0 < n < 10^{100}$).

**Формат выходных данных:** <br>
Программа должна вывести число с запятыми в соответствии с условием задачи.

## sample_input:
1000000 <br>

## sample_output:
1,000,000 <br>

```python
# date: 08.01.2023

num = int(input())

# Проверить, входит ли число в заданный диапазон;
if 0 < num < 10 ** 100:

    # Проверить строковую длинну числа, если равно или меньше 3 -- вывести на печать
    if len(str(num)) <= 3:
        print(num)

    # Если число больше или равно 4 -- разложить на "элементы" в Список
    else:
        num_list = list()
        for i in range(len(str(num))):
            num_list.append(str(num)[i])

        index = -3
        # Проверить, делится ли строковая длина числа на 3 без остатка;
        if len(str(num)) % 3 == 0:
            for j in range((len(str(num)) // 3) - 1):
                num_list.insert(index, ',')
                index -= 4

        # Если число нечётное --
        elif len(str(num)) % 3 != 0:
            for j in range((len(str(num)) // 3)):
                num_list.insert(index, ',')
                index -= 4

        # Собрать Строку в Число
        num_total = str()
        for k in range(len(num_list)):
            num_total = ''.join(num_list)

        print(num_total)

else:
    print('Вне диапазона!')
```

# step_10 - Задача Иосифа Флавия 🌶️🌶️
$n$ человек, пронумерованных числами от 1 до $n$, стоят в кругу. Они начинают считаться, каждый $k$-й по счету человек выбывает из круга, после чего счет продолжается со следующего за ним человека. Напишите программу, определяющую номер человека, который останется в кругу последним.

**Формат входных данных:** <br>
На вход программе подаются два числа $n$ и $k$, записанные на отдельных строках.

**Формат выходных данных:** <br>
Программа должна вывести одно число – номер человека, который останется в кругу последним.

**Примечание №1** <br>
Подробнее ознакомиться с классической задачей Иосифа Флавия можно <a href="https://ru.wikipedia.org/wiki/Задача_Иосифа_Флавия">тут</a>.

**Примечание №2** <br>
Визуализацию работы алгоритма можно посмотреть <a href="https://www.geogebra.org/m/ExvvrBbR">тут</a>.

## sample_input:
5 <br>
2 <br>

## sample_output:
3 <br>

```python
# date: 08.01.2023

n, k = int(input()), int(input())
last = 0

for i in range(1 , n + 1):
    last = (last + k) % i

print(last + 1)
```
