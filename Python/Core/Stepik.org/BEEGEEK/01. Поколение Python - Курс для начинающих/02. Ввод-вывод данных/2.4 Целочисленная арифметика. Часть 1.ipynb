# step_07 - Куб
Напишите программу, вычисляющую объём куба и площадь его полной поверхности, по введённому значению длины ребра.

**Формат входных данных:** <br>
На вход программе подается одно целое число – длина ребра.

**Формат выходных данных:** <br>
Программа должна вывести текст и числа в соответствии с условием задачи.

**Примечание №1:** <br>
Объём куба и площадь полной поверхности можно вычислить по формулам $V = a^3$, $S = 6a^2$.

## sample_input:
25

## sample_output:
Объем = 15625 <br>
Площадь полной поверхности = 3750 <br>

```python
# date: 05.09.2021

a = int(input())
print('Объем =', a**3)
print('Площадь полной поверхности =', 6*a**2)
```

```python
# date: 12.11.2022

print(f'Объем = {(num := int(input())) ** 3}',
      f'Площадь полной поверхности = {6 * num ** 2}',
      sep = '\n'
      )
```

```python
def cube_volume(rib_length: float) -> float:
      """
      :param rib_length: cube rib length
      :return: f(x) = rib_length ** 3  # cube volume
      """

      return rib_length ** 3

def cube_square(rib_length: float) -> float:
      """
      :param rib_length: cube rib length
      :return: f(x) = 6 * rib_length ** 2  # cube square
      """

      return 6 * rib_length ** 2
```

```python
# date: 12.11.2022

print(f'Объём = {cube_volume(25)}',
      f'Площадь полной поверхности = {cube_square(25)}',
      sep = '\n')
```

# step_12 - Арифметическая прогрессия
Арифметической прогрессией называется последовательность чисел $a_1, a_2, ..., a_n$, каждое из которых, начиная с $a_2$, получается из предыдущего, прибавлением к нему одного и того же постоянного числа $d$ (разность прогрессии), то есть: $a_n = a_(n − 1) + d$

Если известен первый член прогрессии и её разность, то $n$-ый член арифметической прогрессии находится по формуле: $a_n = a_1 + d(n - 1)$

**Формат входных данных:** <br>
На вход программе подаётся три целых числа: $a_1$, $d$ и $n$, каждое на отдельной строке.

**Формат выходных данных:** <br>
Программа должна вывести $n$-ый член арифметической прогрессии.

## sample_input:
1 <br>
1 <br>
10 <br>

## sample_output:
10

```python
# date: 05.09.2021

a1 = int(input())
d = int(input())
n = int(input())
print(a1 + d * (n - 1))
```

```python
# date: 12.11.2022

print(int(input()) + int(input()) * (int(input()) - 1))
```

```python
def arithmetic_progression(a_1: float, diff: float, a_n: float) -> float:
      """
      An arithmetic progression or arithmetic sequence (AP) is a sequence of numbers such that the difference between the consecutive terms is constant. For instance, the sequence 5, 7, 9, 11, 13, 15, . . . is an arithmetic progression with a common difference of 2.

      :param a_1: the first member of the progression
      :param diff: progression difference
      :param a_n: the last member of the progression
      :return: arithmetic progression (a_1 + diff * (a_n - 1))
      """

      return a_1 + diff * (a_n - 1)
```

```python
print(arithmetic_progression(1, 1, 10))
```
