# step_06 - Кастомный разделитель
Напишите программу, которая считывает строку-разделитель и три строки, а затем выводит указанные строки через разделитель.

**Формат входных данных:** <br>
На вход программе подаётся строка-разделитель и три строки, каждая на отдельной строке.

**Формат выходных данных:** <br>
Программа должна вывести введённые три строки через разделитель.

## sample_input:
\* <br>
Раз <br>
Два <br>
Три <br>

## sample_output:
Раз*Два*Три

```python
# date: 05.09.2021

namesep= input()
name1= input()
name2= input()
name3= input()
print(name1, name2, name3, sep=namesep)
```

```python
# date: 12.11.2022

separator: str = input()
word_0: str = input()
word_1: str = input()
word_2: str = input()

print(word_0,
      word_1,
      word_2,
      sep = separator)
```

```python
# date: 12.11.2022

print(input().join([input() for _ in range(3)]))
```

# step_07 - Приветствие
Напишите программу, которая приветствует пользователя, выводя слово «Привет» (без кавычек), после которого должна стоять запятая и пробел, а затем введенное имя и восклицательный знак.

**Формат входных данных:** <br>
На вход программе подаётся одна строка — имя пользователя.

**Формат выходных данных:** <br>
Программа должна вывести текст в соответствии с условием задачи.

**Примечание №1:** <br>
Перед восклицательным знаком не должно быть пробелов.

**Примечание №2:** <br>
Используйте необязательный параметр end.

## sample_input:
Тирион

## sample_output:
Привет, Тирион!

-----

```python
# date: 05.09.2021

name= input()
print('Привет,', name, end='!')
```

```python
# date: 12.11.2022

name: str = input()

print(f'Привет, {name}',
      end = '!')
```

```python
# date: 12.11.2022

print(f'Привет, {input()}', end = '!')
```

```python
def hello_to_2(name: str) -> None:
    print(f'Привет, {name}',
          end = '!')

hello_to_2('Тирион')
```
