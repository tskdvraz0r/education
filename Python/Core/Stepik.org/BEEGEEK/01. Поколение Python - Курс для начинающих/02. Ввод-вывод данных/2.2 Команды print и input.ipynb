# step_13 - Повторяй за мной
Напишите программу, которая считывает три строки по очереди, а затем выводит их в той же последовательности, каждую на отдельной строчке.

**Формат входных данных:** <br>
На вход программе подаются три строки, каждая на отдельной строке.

**Формат выходных данных:** <br>
Программа должна вывести введенные строки в той же последовательности, каждую на отдельной строке.

**Примечание №1:** <br>
Для считывания текста используйте команду *input()*, для печати текста на экране используйте команду *print()*.

## sample_input:
I was <br>
born <br>
this way

## sample_output:
I was <br>
born <br>
this way

````python
# date: 05.09.2021

name1= input()
name2= input()
name3= input()
print(name1)
print(name2)
print(name3)
````

# step_14 - Повторяй за мной 2
Напишите программу, которая считывает три строки по очереди, а затем выводит их в обратной последовательности, каждую на отдельной строчке.

**Формат входных данных:** <br>
На вход программе подается три строки, каждая на отдельной строке.

**Формат выходных данных:** <br>
Программа должна вывести введенные строки в обратной последовательности, каждую на отдельной строке.

**Примечание №1:** <br>
Используйте 3 переменные для сохранения введённых строк текста.

## sample_input:
Hello <br>
it's <br>
me <br>

## sample_output:
me <br>
it's <br>
Hello <br>

````python
# date: 05.09.2021

name1= input()
name2= input()
name3= input()
print(name3)
print(name2)
print(name1)
````

````python
# date: 12.11.2022

word_0, word_1, word_2 = input(), input(), input()

print(word_2,
      word_1,
      word_0,
      sep = '\n')
````
