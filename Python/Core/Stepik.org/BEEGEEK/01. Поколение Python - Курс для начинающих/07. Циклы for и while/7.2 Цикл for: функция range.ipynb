# step_07 - Последовательность чисел 1
Даны два целых числа $m$ и $n$. Напишите программу, которая выводит все числа от $m$ до $n$ включительно.

**Формат входных данных:** <br>
На вход программе подаются два целых числа $m$ и $n$, каждое на отдельной строке.

**Формат выходных данных:** <br>
Программа должна вывести числа в соответствии с условием задачи.

## sample_input:
1 <br>
9 <br>

## sample_output:
1 <br>
2 <br>
3 <br>
4 <br>
5 <br>
6 <br>
7 <br>
8 <br>
9 <br>

```python
# date: 01.10.2021

m, n = int(input()), int(input())
for i in range(m, n+1):
    print(i)
```

```python
# date: 20.11.2022

print(*[i for i in range(int(input()), int(input()) + 1)], sep = '\n')
```

# step_08 - Последовательность чисел 2 🌶️
Даны два целых числа $m$ и $n$. Напишите программу, которая выводит все числа от $m$ до $n$ включительно в порядке возрастания, если $m < n$, или в порядке убывания в противном случае.

**Формат входных данных:** <br>
На вход программе подаются два целых числа mm и nn, каждое на отдельной строке.

**Формат выходных данных:** <br>
Программа должна вывести числа в соответствии с условием задачи.

## sample_input:
1 <br>
5 <br>

## sample_output:
1 <br>
2 <br>
3 <br>
4 <br>
5 <br>

```python
# date: 01.10.2021

m, n = int(input()), int(input())
if m < n:
    for i in range(m, n + 1, 1):
        print(i)
else:
    for j in range(m, n - 1, -1):
        print(j)
```

```python
# date: 20.11.2022

print(*[i for i in range(n, m + 1)] if (n := int(input())) < (m := int(input())) else [i for i in range(n, m - 1, -1)], sep = '\n')
```

# step_09 - Последовательность чисел 3 🌶️
Даны два целых числа $m$ и $n$ ($m > n$). Напишите программу, которая выводит все нечетные числа от $m$ до $n$ включительно в порядке убывания.

**Формат входных данных:** <br>
На вход программе подаются два целых числа $m$ и $n$, каждое на отдельной строке.

**Формат выходных данных:** <br>
Программа должна вывести числа в соответствии с условием задачи.

**Подсказка №1:** <br>
Попробуйте решить задачу двумя способами: с использованием условного оператора *if* и без него.

## sample_input:
77 <br>
62 <br>

## sample_output:
77 <br>
75 <br>
73 <br>
71 <br>
69 <br>
67 <br>
65 <br>
63 <br>

```python
# date: 01.10.2021

m, n = int(input()), int(input())

if m > n:
    if m % 2 != 0:
        for i in range(m, n, -2):
            print(i)
    elif m % 2 == 0:
        for j in range(m - 1, n - 1, -2):
            print(j)
```

```python
# date: 20.11.2022

print(*[i for i in range(m, n - 1, -2)] if (m := int(input())) % 2 != 0 & (n := int(input())) < m else [i for i in range(m - 1, n - 1, -2)], sep = '\n')
```

# step_10 - Последовательность чисел 4
Даны два натуральных числа $m$ и $n$ ($m \le n$). Напишите программу, которая выводит все числа от $m$ до $n$ включительно удовлетворяющие хотя бы одному из условий:
- число кратно 17;
- число оканчивается на 9;
- число кратно 3 и 5 одновременно.

**Формат входных данных:** <br>
На вход программе подаются два натуральных числа $m$ и $n$ (m \le nm≤n), каждое на отдельной строке.

**Формат выходных данных:** <br>
Программа должна вывести числа в соответствии с условием задачи.

**Подсказка №1:** <br>
Если чисел удовлетворяющих условию нет, выводить ничего не надо.

## sample_input:
1 <br>
20 <br>

## sample_output:
9 <br>
15 <br>
17 <br>
19 <br>

```python
# date: 01.10.2021

m: int = int(input())
n: int = int(input())

for i in range(m, n + 1):
    if m <= n and i % 17 == 0 or i == 9 or i % 10 == 9 or (i % 3 == 0 and i % 5 == 0):
        print(i)
```

# step_11 - Таблица умножения
Дано натуральное число $n$. Напишите программу, которая выводит таблицу умножения на $n$.

**Формат входных данных:** <br>
На вход программе подается натуральное число.

**Формат выходных данных:** <br>
Программа должна вывести таблицу умножения на введенное число.

**Подсказка №1:** <br>
В качестве знака умножения используйте английскую букву $x$.

## sample_input:
5

## sample_output:
5 x 1 = 5 <br>
5 x 2 = 10 <br>
5 x 3 = 15 <br>
5 x 4 = 20 <br>
5 x 5 = 25 <br>
5 x 6 = 30 <br>
5 x 7 = 35 <br>
5 x 8 = 40 <br>
5 x 9 = 45 <br>
5 x 10 = 50 <br>

```python
# date: 01.10.2021

n = int(input())
for i in range(1, 11):
    print(n, 'x', i, '=', i*n)
```

```python
# date: 21.11.2022

print(*[f'{n} x {i} = {i * n}' for i in range(1, 11)] if (n := int(input())) > 0 else '-', sep = '\n')
```
