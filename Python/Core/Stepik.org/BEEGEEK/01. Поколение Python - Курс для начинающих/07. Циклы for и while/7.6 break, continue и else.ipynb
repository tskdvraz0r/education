# step_07 - Наименьший делитель
На вход программе подается число $n > 1$. Напишите программу, которая выводит его наименьший отличный от 1 делитель.

**Формат входных данных:** <br>
На вход программе подается одно натуральное число nn.

**Формат выходных данных:** <br>
Программа должна вывести наименьший делитель отличный от 11.

**Примечание №1:** <br>
Используйте оператор break при обнаружении делителя.

## sample_input:
15 <br>

## sample_output:
3 <br>

```python
# date: 26.12.2021

num = int(input())
for i in range(num + 1):
    if i != 0 and i != 1:
        if num % i == 0:
            print(i)
            break
```

# step_08 - Следуй правилам
На вход программе подается натуральное число nn. Напишите программу, которая выводит числа от 11 до nn включительно за исключением:
* чисел от 55 до 99 включительно;
* чисел от 1717 до 3737 включительно;
* чисел от 7878 до 8787 включительно.

**Формат входных данных:** <br>
На вход программе подается одно натуральное число nn.

**Формат выходных данных:** <br>
Программа должна вывести числа в соответствии с условием задачи, каждое на отдельной строке.

**Примечание №1:** <br>
Используйте оператор *continue*.

## sample_input:
10 <br>

## sample_output:
1 <br>
2 <br>
3 <br>
4 <br>
10 <br>

```python
# date: 26.12.2021

num = int(input())

for i in range(1, num + 1):
    if i in range(5, 10) or i in range(17, 38) or i in range(78, 88):
        continue
    print(i)
```
