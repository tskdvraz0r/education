# step_05 - Таблица-1
Дано натуральное число $n$, ($n \le 9$). Напишите программу, которая печатает таблицу размером $n \times 3$ состоящую из данного числа (числа отделены одним пробелом).

**Формат входных данных:** <br>
На вход программе подается одно натуральное число.

**Формат выходных данных:** <br>
Программа должна вывести таблицу размером $n \times 3$ состоящую из данного числа.

**Примечание №1:** <br>
В конце строки может быть пробел.

## sample_input:
8 <br>

## sample_output:
8 8 8 <br>
8 8 8 <br>
8 8 8 <br>
8 8 8 <br>
8 8 8 <br>
8 8 8 <br>
8 8 8 <br>
8 8 8 <br>

````python
# date: 26.12.2021

num = int(input())

for i in range(num):
    for j in range(3):
        print(num, end=' ')
    print()
````

# step_06 - Таблица-2
Дано натуральное число $n$, ($n \le 9$). Напишите программу, которая печатает таблицу размером $n \times 5$, где в $i$-ой строке указано число $i$ (числа отделены одним пробелом).

**Формат входных данных:** <br>
На вход программе подается одно натуральное число.

**Формат выходных данных:** <br>
Программа должна вывести таблицу размером n \times 5n×5 в соответствии с условием.

**Примечание №1:** <br>
В конце строки может быть пробел.

## sample_input:
3 <br>

## sample_output:
1 1 1 1 1 <br>
2 2 2 2 2 <br>
3 3 3 3 3 <br>

````python
# date: 26.12.2021

num = int(input())

for i in range(num):
    for j in range(5):
        print(i + 1, end=' ')
    print()
````

# step_07 - Таблица-3
Дано натуральное число $n$, ($n \le 9$). Напишите программу, которая печатает таблицу сложения для всех чисел от 1 до $n$ в соответствии с примером.

**Формат входных данных:** <br>
На вход программе подается одно натуральное число.

**Формат выходных данных:** <br>
Программа должна вывести таблицу сложения для всех чисел от 11 до nn.

**Примечание №1:** <br>
В конце строки может быть пробел.

## sample_input:
1 <br>

## sample_output:
1 + 1 = 2 <br>
1 + 2 = 3 <br>
1 + 3 = 4 <br>
1 + 4 = 5 <br>
1 + 5 = 6 <br>
1 + 6 = 7 <br>
1 + 7 = 8 <br>
1 + 8 = 9 <br>
1 + 9 = 10 <br>

-----

````python
# date: 26.12.2021

num = int(input())

for i in range(num):
    for j in range(9):
        print(i + 1, '+', (j + 1), '=', (i + 1) + (j + 1))
    print()
````

# step_08 - Звездный треугольник 🌶️🌶️
Дано нечетное натуральное число $n$. Напишите программу, которая печатает равнобедренный звездный треугольник с основанием, равным $n$ в соответствии с примером:
```
*
**
***
****
***
**
*
```

**Формат входных данных:** <br>
На вход программе подается одно нечетное натуральное число.

**Формат выходных данных:** <br>
Программа должна вывести треугольник в соответствии с условием.

**Примечание №1:** <br>
Используйте вложенный цикл for.

## sample_input:
3 <br>

## sample_output:
```
*
**
*
```

````python
# date: 26.12.2021

num = int(input())

if num % 2 != 0:
    for i in range(num):
        for j in range(i + 1):
            print('*', end='')
        if i == (num // 2):
            print()
            break
        print()
    for i in range(num):
        for j in range(i + 1, (num // 2) + 1):
            print('*', end='')
        print()
````

# step_09 - Численный треугольник 1
Дано натуральное число $n$. Напишите программу, которая печатает численный треугольник в соответствии с примером:

1 <br>
22 <br>
333 <br>
4444 <br>
55555 <br>

**Формат входных данных:** <br>
На вход программе подается одно натуральное число.

**Формат выходных данных:** <br>
Программа должна вывести треугольник в соответствии с условием.

**Примечание №1:**
Используйте вложенный цикл *for*.

## sample_input:
5 <br>

## sample_output:
1 <br>
22 <br>
333 <br>
4444 <br>
55555 <br>

````python
# date: 26.12.2021

num = int(input())

for i in range(num + 1):
    for j in range(i):
        print(i, end='')
    print()
````
