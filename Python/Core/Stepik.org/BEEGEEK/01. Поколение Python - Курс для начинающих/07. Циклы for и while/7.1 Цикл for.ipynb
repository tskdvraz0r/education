*Формат входных данных:* <br>
На вход программе подается натуральное число $n$.

*Формат выходных данных:* <br>
Программа должна вывести текст в соответствии с условием задачи.

## sample_input:

Квадрат числа 3 равен 9

# step_09 - Звездный треугольник
На вход программе подается натуральное число $n$, ($n \ge 2$) – катет прямоугольного равнобедренного треугольника.

Напишите программу, которая выводит звездный треугольник в соответствии с примером.

*Формат входных данных:* <br>
На вход программе подается одно натуральное число $n$, ($n \ge 2$).

*Формат выходных данных:* <br>
Программа должна вывести треугольник в соответствии с условием задачи.

## sample_input:
3

## sample_output:
```
***
**
*
```

````python
# date: 19.11.2022
num: int = int(input())
print(*[(num - i) * '*' for i in range(num)], sep = '\n')
````

# step_10 - Популяция
На вход программе подается три натуральных числа $m$, $p$, $n$:
* m: стартовое количество организмов;
* p: среднесуточное увеличение в %;
* n: количество дней для размножения.

На вход программе подается три натуральных числа.

*Формат выходных данных:* <br>

## sample_input:
10
50
6

## sample_output:

2 15.0
3 22.5
4 33.75
5 50.625
6 75.9375
