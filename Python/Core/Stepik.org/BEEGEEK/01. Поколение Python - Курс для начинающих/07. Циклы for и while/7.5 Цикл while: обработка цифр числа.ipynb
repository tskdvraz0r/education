# step_04 - Обратный порядок 1
Дано натуральное число. Напишите программу, которая выводит его цифры в столбик в обратном порядке.

Формат входных данных
На вход программе подается одно натуральное число.

Формат выходных данных
Программа должна вывести цифры введенного числа в столбик в обратном порядке.

## sample_input:
12345

## sample_output:
5
4
3
2
1

```python
# date:

num = int(input())
while num != 0:
    last_num = num % 10
    print(last_num)
    num = num // 10
```

# step_06 - max и min
Дано натуральное число n, \, (n \ge 10)n,(n≥10). Напишите программу, которая определяет его максимальную и минимальную цифры.

Формат входных данных
На вход программе подается одно натуральное число.

Формат выходных данных
Программа должна вывести максимальную и минимальную цифры введенного числа (с поясняющей надписью).

## sample_input:
26670

## sample_output:
Максимальная цифра равна 7
Минимальная цифра равна 0

-----

```python
# date:

num = int(input())
num_min = num % 10
num_max = num % 10

while num != 0:
    last_digital = num % 10
    if last_digital > num_max:
        num_max = last_digital
    elif last_digital < num_min:
        num_min = last_digital
    num = num // 10

print('Максимальная цифра равна', num_max)
print('Минимальная цифра равна', num_min)
```

# step_07 - Все вместе
Дано натуральное число. Напишите программу, которая вычисляет:

сумму его цифр;
количество цифр в нем;
произведение его цифр;
среднее арифметическое его цифр;
его первую цифру;
сумму его первой и последней цифры.
Формат входных данных
На вход программе подается одно натуральное число.

Формат выходных данных
Программа должна вывести значения указанных величин в указанном порядке.

## sample_input:
5678

## sample_output:
26
4
1680
6.5
5
13

```python
# date:

num = int(input())
sum_num = 0
cnt_num = 0
pro_num = 1
last_num = num % 10
revers = str()

while num != 0:
    last_digital = num % 10
    sum_num += last_digital
    cnt_num += 1
    pro_num *= last_digital
    revers = revers + str(last_digital)
    num = num // 10

print(sum_num)  # сумма
print(cnt_num)  # количество
print(pro_num)  # произведение
print(sum_num / cnt_num)  # среднее
print(int(revers) % 10)  # первая
print(last_num + (int(revers) % 10))  # сумма первой и последней
```

# step_09 - Одинаковые цифры
Дано натуральное число. Напишите программу, которая определяет, состоит ли указанное число из одинаковых цифр.

Формат входных данных
На вход программе подается одно натуральное число.

Формат выходных данных
Программа должна вывести «YES» если число состоит из одинаковых цифр и «NO» в противном случае.

## sample_input:
11111

## sample_output:
YES

```python
# date:

num = int(input())  # enter num
last_num = num % 10  # last number
same_num = False  # flag

while num != 0:
    last_digital = num % 10
    if last_digital == last_num:
        same_num = True
    else:
        same_num = False
        break

    last_num = last_digital
    num = num // 10

if same_num == True:
    print('YES')
else:
    print('NO')
```

# step_10 - Упорядоченные цифры 🌶️
Дано натуральное число. Напишите программу, которая определяет, является ли последовательность его цифр при просмотре справа налево упорядоченной по неубыванию.

Формат входных данных
На вход программе подается одно натуральное число.

Формат выходных данных
Программа должна вывести «YES» если последовательность его цифр при просмотре справа налево является упорядоченной по неубыванию и «NO» в противном случае.

## sample_input:
5321

## sample_output:
YES

```python
# date:

n=int(input())
flag=True
n1=n%10
n=n//10
while n!=0 and flag==True:
    if n1<=n%10:
        n1=n%10
        n//=10
    else:
        flag=False
if  flag == True:
    print('YES')
else:
    print('NO')
```
