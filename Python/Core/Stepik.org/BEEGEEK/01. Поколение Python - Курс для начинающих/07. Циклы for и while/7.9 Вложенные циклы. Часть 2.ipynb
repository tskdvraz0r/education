# step_01 - Численный треугольник 3
Дано натуральное число $n$. Напишите программу, которая печатает численный треугольник с высотой равной $n$, в соответствии с примером:
1 <br>
2 3 <br>
4 5 6 <br>
7 8 9 10 <br>
11 12 13 14 15 <br>
16 17 18 19 20 21 <br>

**Формат входных данных:** <br>
На вход программе подается одно натуральное число.

**Формат выходных данных:** <br>
Программа должна вывести треугольник в соответствии с условием.

**Примечание №1:** <br>
Используйте вложенный цикл for.

## sample_input:
3 <br>

## sample_output:
1 <br>
2 3 <br>
4 5 6 <br>

```python
# date: 26.12.2021

num = int(input())
cnt = 0
for i in range(num + 1):
    for j in range(i):
        print(cnt + 1, end=' ')
        cnt += 1
    print()
```

# step_02 - Численный треугольник 4
Дано натуральное число $n$. Напишите программу, которая печатает численный треугольник с высотой равной $n$, в соответствии с примером:
1 <br>
121 <br>
12321 <br>
1234321 <br>
123454321 <br>

**Формат входных данных:** <br>
На вход программе подается одно натуральное число.

**Формат выходных данных:** <br>
Программа должна вывести треугольник в соответствии с условием.

**Примечание №1:** <br>
Используйте вложенный цикл for.

## sample_input:
5 <br>

## sample_output:
1 <br>
121 <br>
12321 <br>
1234321 <br>
123454321 <br>

```python
# date: 26.12.2021

n = int(input())
for i in range(1, n + 1):
    for j in range(1, 2 * i):
        print(min(j, 2 * i - j), end='')
    print()
```

# step_03 - Делители-1 🌶️
На вход программе подается два натуральных числа $a$ и $b$ ($a < b$). Напишите программу, которая находит натуральное число из отрезка $[a;b]$ с максимальной суммой делителей.

**Формат входных данных:** <br>
На вход программе подаются два числа, каждое на отдельной строке.

**Формат выходных данных:** <br>
Программа должна вывести два числа на одной строке, разделенных пробелом: число с максимальной суммой делителей и сумму его делителей.

**Примечание №1:** <br>
Если таких чисел несколько, то выведите наибольшее из них.

## sample_input:
1 <br>
10 <br>

## sample_output:
10 18 <br>

-----

```python
# date: 26.12.2021

a, b = int(input()), int(input())
total_maximum = 0
digital = 0

for i in range(a, b + 1):
    maximum = 0
    for j in range(1, i + 1):
        if i % j == 0:
            maximum += j
        if maximum >= total_maximum:
            total_maximum = maximum
            digital = j

print(digital, total_maximum)
```

# step_04 - Делители-2
На вход программе подается натуральное число $n$. Напишите программу, выводящую графическое изображение делимости чисел от 1 до $n$ включительно. В каждой строке надо напечатать очередное число и столько символов «+», сколько делителей у этого числа.

**Формат входных данных:** <br>
На вход программе подается одно натуральное число.

**Формат выходных данных:** <br>
Программа должна вывести графическое изображение чисел от 1 до $n$, каждое на отдельной строке.

## sample_input:
5 <br>

## sample_output:
1+ <br>
2++ <br>
3++ <br>
4+++ <br>
5++ <br>

```python
# date: 26.12.2021

n = int(input())
for i in range(1, n + 1):
    print(i, end='')
    for j in range(1, i + 1):
        if i % j == 0:
            print('+', end='')
    print()
```

# step_05 - Цифровой корень
На вход программе подается натуральное число $n$. Напишите программу, которая находит цифровой корень данного числа. Цифровой корень числа $n$ получается следующим образом: если сложить все цифры этого числа, затем все цифры найденной суммы и повторить этот процесс, то в результате будет получено однозначное число (цифра), которое и называется цифровым корнем данного числа.

**Формат входных данных:** <br>
На вход программе подается одно натуральное число.

**Формат выходных данных:** <br>
Программа должна вывести цифровой корень введенного числа.

**Примечание №1:** <br>
Используйте вложенные циклы while.

## sample_input:
192

## sample_output:
3

```python
# date: 26.12.2021

num = int(input())
total_sum = 0

while num != 0:
    last_digital = num % 10
    total_sum += last_digital
    num //= 10

num = total_sum
digital_sum = 0

while num > 0:
     last_digital = num % 10
     digital_sum += last_digital
     num //= 10

if digital_sum <= 9:
    print(digital_sum)
else:
    total_sum2 = 0
    while digital_sum != 0:
        last_digital = digital_sum % 10
        total_sum2 += last_digital
        digital_sum //= 10
    print(total_sum2)
```

# step_06 - Сумма факториалов
Дано натуральное число $n$. Напишите программу, которая выводит значение суммы $1!+2!+3!+\ldots+n!$.

**Формат входных данных:** <br>
На вход программе подается одно натуральное число.

**Формат выходных данных:** <br>
Программа должна вывести значение суммы $1!+2!+3!+\ldots+n!$.

**Примечание №1:** <br>
Факториалом натурального числа $n$, называется произведение всех натуральных чисел от 1 до $n$, то есть <br>
$$n!=1\cdot2\cdot3\cdot…\cdot n$$

## sample_input:
3

## sample_output:
9

```python
# date: 26.12.2021

import math
a = int(input())
fuck_num = 0
total_fuck = 0

for i in range(a):
    fuck_num = math.factorial(i + 1)
    total_fuck += fuck_num

print(total_fuck)
```

# step_07 - Простые числа
На вход программе подается два натуральных числа $a$ и $b$ ($a < b$). Напишите программу, которая находит все простые числа от $a$ до $b$ включительно.

**Формат входных данных:** <br>
На вход программе подаются два числа, каждое на отдельной строке.

**Формат выходных данных:** <br>
Программа должна вывести все простые числа от $a$ до $b$ включительно, каждое на отдельной строке.

**Примечание №1:** <br>
Число 1 простым не является.

## sample_input:
2 <br>
15 <br>

## sample_output:
2 <br>
3 <br>
5 <br>
7 <br>
11 <br>
13 <br>

```python
# date: 26.12.2021

a, b, = int(input()), int(input())
for i in range(a, b + 1):
    if i == 1:
        continue
    for j in range(2, i):
        if i % j == 0:
            break
    else:
        print(i)
```
