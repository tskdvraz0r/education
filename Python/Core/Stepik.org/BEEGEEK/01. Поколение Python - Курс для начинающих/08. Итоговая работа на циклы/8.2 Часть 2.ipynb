Программист торопился и написал программу неправильно. Найдите все ошибки в этой программе (их может быть одна или несколько). Известно, что каждая ошибка затрагивает только одну строку и может быть исправлена без изменения других строк.

**Примечание №1:** <br>

````python
# date: 27.12.2021

n = input()
s = 0
while n > 10:
    if n % 2 == 1:
        s = n % 10
    n //= 10
print(s)
````

# step_02 - Ревью кода - 8 🌶️
На обработку поступает последовательность из 8 целых чисел. Известно, что вводимые числа по абсолютной величине не превышают $10^{12}$. Нужно написать программу, которая выводит на экран количество делящихся нацело на 4 чисел в исходной последовательности и максимальное делящееся нацело на 4 число. Если делящихся нацело на 4 чисел нет, требуется на экран вывести *«NO»*. Программист торопился и написал программу неправильно. Найдите все ошибки в этой программе (их может быть одна или несколько). Известно, что каждая ошибка затрагивает только одну строку и может быть исправлена без изменения других строк.

**Примечание №1:** <br>
Обратите внимание, что требуется найти ошибки в имеющейся программе, а не написать свою, возможно, использующую другой алгоритм решения.


````python
# date: 27.12.2021

n = 7
count = 0
maximum = 1000
for i in range(1, n + 1):
    x = int(input())
    if x // 4 == 0:
        count += 1
        if x < maximum:
            maximum = x
if count > 0:
    print(count)
    print(maximum)
else:
    print('NO')
````

# step_04 - Звездная рамка
На вход программе подается натуральное число $n$. Напишите программу, которая печатает звездную рамку размерами $n \times 19$.

**Формат входных данных:** <br>
На вход программе подаётся натуральное число $n \in [3;19]$ — высота звездной рамки.

**Формат выходных данных:** <br>
Программа должна вывести звездную рамку размерами $n \times 19$.

**Примечание №1:** <br>
Для печати звездной линии используйте умножение строки на число.

## sample_input:
5 <br>

## sample_output:
```
*******************
*                 *
*                 *
*                 *
*******************
```

````python
# date: 27.12.2021

num = int(input())

print('*' * 19)

if 3 <= num <= 19:
    for i in range(num - 2):
        print('*', ' ' * 17, '*', sep='')

print('*' * 19)
````

# step_05 - Третья цифра
Дано натуральное число $n, ($n > 99$). Напишите программу, которая определяет его третью (с начала) цифру.

**Формат входных данных:** <br>
На вход программе подается одно натуральное число, состоящее как минимум из трех цифр.

**Формат выходных данных:** <br>
Программа должна вывести его третью (с начала) цифру.

## sample_input:
100_000_000 <br>

## sample_output:
0 <br>

````python
# date: 27.12.2021

num = int(input())

while num > 1000:
    num //= 10

last_digital = 0
if 99 < num < 999:
    last_digital = num % 10
print(last_digital)
````

# step_06 - Все вместе 2
Дано натуральное число. Напишите программу, которая вычисляет:
* количество цифр 3 в нем;
* сколько раз в нем встречается последняя цифра;
* количество четных цифр;
* сумму его цифр, больших пяти;
* произведение цифр, больших семи (если цифр больших семи нет, то вывести 1, если такая цифра одна, то вывести ее);
* сколько раз в нем встречается цифры 0 и 5 (всего суммарно).

**Формат входных данных:** <br>
На вход программе подается одно натуральное число.

**Формат выходных данных:** <br>
Программа должна вывести значения указанных величин в указанном порядке.

## sample_input:
56639 <br>

## sample_output:
1 <br>
1 <br>
2 <br>
21 <br>
9 <br>
1 <br>

````python
# date: 27.12.2021

num = int(input())
last_digital = num % 10
count_3 = 0
count_ld = 0
count_del2 = 0
sum_digital_5 = 0
mult_digital_7 = 1
count_digital_05 = 0

while num != 0:

    # считаем количество цифр 3
    if num % 10 == 3:
        count_3 += 1
    # считаем сколько раз встречается последняя цифра
    if num % 10 == last_digital:
        count_ld += 1
    # считаем количество чётных цифр
    if num % 2 == 0:
        count_del2 += 1
    # считаем сумму цифр больше 5
    if num % 10 > 5:
        num5 = num % 10
        sum_digital_5 += num5
    # считаем сумму цифр больше 7
    if num % 10 > 7:
        num7 = num % 10
        mult_digital_7 *= num7
    #
    if num % 10 == 5 or num % 10 == 0:
        num05 = num % 10
        count_digital_05 += 1
    # отсекаем последюю цифру
    num //= 10

print(count_3, count_ld, count_del2, sum_digital_5, mult_digital_7, count_digital_05, sep='\n')
````
