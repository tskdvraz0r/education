# step_08 - Арифметическая прогрессия
Напишите программу, которая определяет, являются ли три заданных числа (в указанном порядке) последовательными членами <a href="https://ru.wikipedia.org/wiki/Арифметическая_прогрессия">арифметической прогрессии</a>.

**Формат входных данных:** <br>
На вход программе подаются три числа, каждое на отдельной строке.

**Формат выходных данных:** <br>
Программа должна вывести *«YES»* или *«NO»* (без кавычек) в соответствии с условием задачи.

## sample_input:
1 <br>
2 <br>
3 <br>

## sample_output:
YES

```python
# date: 08.09.2021

num1= int(input())
num2= int(input())
num3= int(input())

if num2-num1 == num3-num2:
    print('YES')
else:
    print('NO')
```

```python
# date: 27.11.2022

def is_progression(num_list: list) -> bool:
    first_num: float = num_list[0]
    diff: float = num_list[-1] - num_list[-2]
    list_length = len(num_list) - 1
    prog_list: list = [first_num]

    while  list_length > 0:
        prog_list.append(first_num + diff)
        first_num = prog_list[-1]
        list_length -= 1

    return num_list == prog_list
```

```python
test = [1, 2, 3]

print('YES'if is_progression(test) else 'NO')
```

# step_11 - Возрастная группа
Напишите программу, которая по введённому возрасту пользователя сообщает, к какой возрастной группе он относится:
* до 13 включительно – детство;
* от 14 до 24 – молодость;
* от 25 до 59 – зрелость;
* от 60 – старость.

**Формат входных данных:** <br>
На вход программе подаётся одно целое число – возраст пользователя.

**Формат выходных данных:** <br>
Программа должна вывести название возрастной группы.

## sample_input:
4

## sample_output:
детство

```python
# date: 08.09.2021

age= int(input())
if age <= 13:
    print('детство')
if 14 <= age <= 24:
    print('молодость')
if 25 <= age <=59:
    print('зрелость')
if age >= 60:
    print('старость')
```

```python
# date: 17.11.2022

age: int = int(input())

match age:
    case age if age <= 13:
        print('детство')
    case age if 14 <= age <= 24:
        print('молодость')
    case age if 25 <= age <= 59:
        print('зрелость')
    case age if age >= 60:
        print('старость')
```
