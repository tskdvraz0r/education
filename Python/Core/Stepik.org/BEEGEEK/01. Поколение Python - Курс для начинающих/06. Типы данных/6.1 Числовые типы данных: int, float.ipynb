# step_12 - Наибольшее и наименьшее
Напишите программу, которая находит наименьшее и наибольшее из пяти чисел.

**Формат входных данных:** <br>
На вход программе подается пять целых чисел, каждое на отдельной строке.

**Формат выходных данных:** <br>
Программа должна вывести наименьшее и наибольшее число с поясняющей надписью.

## sample_input:
1 <br>
2 <br>
3 <br>
4 <br>
5 <br>

## sample_output:
Наименьшее число = 1
Наибольшее число = 5

```python
# date: 18.11.2022

a, b, c, d, e = int(input()), int(input()), int(input()), int(input()), int(input())

print(f'Наименьшее число = {min(a, b, c, d, e)}',
      f'Наибольшее число = {max(a, b, c, d, e)}',
      sep = '\n')
```

```python
# date: 28.11.2022

def minimum(num_list: list) -> float:
      """
      :param num_list:
      :return:
      """

      min_num: float = num_list[0]

      for i in num_list:
          min_num = i if min_num > i else min_num

      return min_num

def maximum(num_list: list) -> float:
      """
      :param num_list:
      :return:
      """

      max_num: float = num_list[0]

      for i in num_list:
            max_num = i if max_num < i else max_num

      return max_num

def minimum_and_maximum(num_list: list) -> list:
      """
      :param num_list:
      :return:
      """

      min_num: float = num_list[0]
      max_num: float = num_list[0]

      for i in num_list:
          min_num = i if min_num > i else min_num
          max_num = i if max_num < i else max_num

      return [min_num, max_num]
```

```python
a = [1, 2, 3, 4, 5]

print(minimum(a),
      maximum(a),
      minimum_and_maximum(a),
      sep = '\n')
```

# step_13 - Сортировка трёх 🌶️
Напишите программу, которая упорядочивает три числа от большего к меньшему.

*Формат входных данных:* <br>
На вход программе подается три целых числа, каждое на отдельной строке.

*Формат выходных данных:* <br>
Программа должна вывести три числа, каждое на отдельной строке, упорядоченных от большего к меньшему.

## sample_input:
132
129
135

## sample_output:
135
132
129

```python
# date: 18.11.2022

a, b, c = int(input()), int(input()), int(input())

print(max(a, b, c,),
      (a + b + c) - max(a, b, c) - min(a, b ,c),
      min(a, b ,c),
      sep = '\n')
```

```python
def insertion_sort(num_list: list) -> list:
      """
      :param num_list:
      :return:
      """
      itr = len(num_list) // 2

      while itr > 0:
            for i in range(len(num_list) - 1):
                  if num_list[i] > num_list[i + 1]:
                        num_list[i], num_list[i + 1] = num_list[i + 1], num_list[i]

            itr -= 1

      return num_list
```

```python
test_list: list = [512, 1, 256, 2, 128, 4, 64, 8, 32, 16]

insertion_sort(test_list)
```
