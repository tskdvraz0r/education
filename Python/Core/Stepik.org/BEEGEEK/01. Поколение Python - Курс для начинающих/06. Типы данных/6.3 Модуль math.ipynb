# step_02 - Евклидово расстояние
На плоскости евклидово расстояние между двумя точками ($x_1; y_1$) и ($x_2}; y_2$) определяется так $\rho = \sqrt(x_1 - x_2)^2 + (y_1 - y_2)^2$.

Напишите программу, определяющую евклидово расстояние между двумя точками, координаты которых заданы.

*Формат входных данных:* <br>
На вход программе подается четыре вещественных числа, каждое на отдельной строке – $x_1, y_1, x_2, y_2$.

*Формат выходных данных:* <br>
Программа должна вывести одно число – евклидово расстояние.

![img_02](https://ucarecdn.com/26c7ba17-bee9-4f8f-9764-3b68b2ede790/)

## sample_input:
2.0
2.5
44.155
100.50

## sample_output:
106.68197610187018

```python
# date: 19.11.2022

import math

x1, y1, x2, y2 = float(input()), float(input()), float(input()), float(input())
print(math.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2))
```

# step_04 - Средние значения
В математике выделяют следующие средние значения:
* среднее арифметическое чисел $a$ и $b$: $\dfrac{a+b}{2}$;
* среднее геометрическое чисел $a$ и $b$: $\sqrt{a\cdot b}$;
* среднее гармоническое чисел $a$ и $b$: $\dfrac{2ab}{a+b}$;
* среднее квадратичное чисел $a$ и $b$: $\sqrt{\dfrac{a^2+b^2}{2}}$.

*Формат входных данных:* <br>
На вход программе подается два вещественных числа aa и bb​, каждое на отдельной строке.

*Формат выходных данных:* <br>
Программа должна вывести 4 числа – среднее арифметическое, геометрическое, гармоническое и квадратичное.

## sample_input:
1.0
2.0

## sample_output:
1.5
1.4142135623730951
1.3333333333333333
1.5811388300841898

```python
# date: 19.11.2022

from math import sqrt

a, b = float(input()), float(input())

print((a + b) / 2,
      sqrt(a * b),
      (2*a*b)/(a + b),
      sqrt((a**2 + b**2)/2),
      sep = '\n')
```

# step_05 - Тригонометрическое выражение
Напишите программу, вычисляющую значение тригонометрического выражения $\sin x + \cos x + \tan^2x$ по заданному числу градусов $x$.

*Формат входных данных:* <br>
На вход программе подается одно вещественное число $x$ измеряемое в градусах.

*Формат выходных данных:* <br>
Программа должна вывести одно число – значение тригонометрического выражения.

*Примечание №1:*
Тригонометрические функции принимают аргумент в радианах. Чтобы перевести градусы в радианы, воспользуйтесь формулой $r = {{x \cdot \pi } \over 180}$

*Примечание №2:*
Модуль math содержит встроенную функцию radians(), которая переводит угол из градусов в угол в радианах.

## sample_input:
30.0

## sample_output:
1.6993587371177719

```python
# date:

import math

x1 = math.radians(float(input()))
print(math.sin(x1) + math.cos(x1) + math.tan(x1) * math.tan(x1))
```

# step_07 - Квадратное уравнение 🌶️🌶️
Даны три вещественных числа $a$, $b$, $c$. Напишите программу, которая находит вещественные корни квадратного уравнения $ax^2 + bx + c = 0$.

*Формат входных данных:* <br>
На вход программе подается три вещественных числа $a \neq 0, b, c$ каждое на отдельной строке.

*Формат выходных данных:* <br>
Программа должна вывести вещественные корни уравнения если они существуют или текст «Нет корней» в противном случае.

*Примечание №1:*
Если уравнение имеет два корня, то следует вывести их в порядке возрастания.

## sample_input:
1
2
1

## sample_output:
-1.0

```python
# date: 19.11.2022

import math

a, b, c = float(input()), float(input()), float(input())

if a != 0:
    D = b ** 2 - 4 * a * c
    if D < 0:
        print('Нет корней')
    elif D > 0:
        x1 = ((-b) - math.sqrt(D)) / (2 * a)
        x2 = ((-b) + math.sqrt(D)) / (2 * a)
        print(min(x1, x2),
              max(x1, x2),
              sep = '\n')
    elif D == 0:
        print((-b) / (2 * a))
```

# step_08 - Правильный многоугольник
Правильный многоугольник — выпуклый многоугольник, у которого равны все стороны и все углы между смежными сторонами. Площадь правильного многоугольника с длиной стороны $a$ и количеством сторон $n$ вычисляется по формуле:
$S = \dfrac{n \cdot a^2}{4 tg \left(\dfrac{\pi}{n}\right)}$

Даны два числа: натуральное число $n$ и вещественное число $a$. Напишите программу, которая находит площадь указанного правильного многоугольника.

*Формат входных данных:* <br>
На вход программе подается два числа nn и aa, каждое на отдельной строке.

*Формат выходных данных:* <br>
Программа должна вывести вещественное число – площадь многоугольника.

## sample_input:
3
2.0

## sample_output:
1.7320508075688779

```python
# date: 19.11.2022

import math

n, a = int(input()), float(input())
print((n * a**2)/(4 * math.tan(math.pi / n)))
```
