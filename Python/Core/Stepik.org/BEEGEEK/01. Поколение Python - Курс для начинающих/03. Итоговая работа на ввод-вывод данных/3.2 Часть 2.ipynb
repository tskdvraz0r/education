# step_02 - Сумма квадратов VS квадрат суммы
Напишите программу, которая считывает два целых числа $a$ и $b$ и выводит на экран квадрат суммы $(a + b)^2$ и сумму квадратов $a^2 + b^2$ этих чисел.

**Формат входных данных:** <br>
На вход программе подаётся два целых числа, каждое на отдельной строке.

**Формат выходных данных:** <br>
Программа должна вывести текст в соответствии с условием.

## sample_input:
3 <br>
2 <br>

## sample_output:
Квадрат суммы 3 и 2 равен 25 <br>
Сумма квадратов 3 и 2 равна 13 <br>

````python
# date: 06.09.2021

a=int(input())
b=int(input())
print('Квадрат суммы', a, 'и', b, 'равен', (a+b)**2)
print('Сумма квадратов', a, 'и', b, 'равна', (a**2+b**2))
````

````python
# date: 12.11.2022

print(f'Квадрат суммы {int(num_a := int(input()))} и {int(num_b := int(input()))} равен {(num_a + num_b) ** 2}',
      f'Сумма квадратов {num_a} и {num_b} равна {num_a ** 2 + num_b ** 2}',
      sep = '\n'
      )
````

````python
def sum_of_squares(a: float, b: float) -> float:
      """
      :param a: 1st number
      :param b: 2nd number
      :return: f(x) = a ** 2 + b ** 2
      """

      return a ** 2 + b ** 2

def square_of_sum(a: float, b: float) -> float:
      """
      :param a: 1st number
      :param b: 2nd number
      :return: f(x) = (a + b) ** 2
      """

      return (a + b) ** 2
````

````python
# date: 12.11.2022

print(f'Квадрат суммы {int(num_a := int(input()))} и {int(num_b := int(input()))} равен {square_of_sum(num_a, num_b)}',
      f'Сумма квадратов {num_a} и {num_b} равна {sum_of_squares(num_a, num_b)}',
      sep = '\n'
      )
````

# step_03 - Большое число
Как известно, целые числа в языке Python не имеют ограничений, которые встречаются в других языках программирования. Напишите программу, которая считывает четыре целых положительных числа $a, b, c, d$ и выводит на экран значение выражения $a^b + c^d$.

**Формат входных данных:** <br>
На вход программе подаётся четыре целых положительных числа $a, b, c, d$, каждое на отдельной строке в указанном порядке.

**Формат выходных данных:** <br>
Программа должна вывести значение $a^b + c^d$.

## sample_input:
9 <br>
29 <br>
7 <br>
27 <br>

## sample_output:
4710194409608608369201743232

````python
# date: 06.09.2021

a = int(input())
b = int(input())
c = int(input())
d = int(input())
print(a**b+c**d)
````

````python
# date: 12.11.2022

print(int(input()) ** int(input()) + int(input()) ** int(input()))
````

# step_04 - Размножение n-ок
Напишите программу, которая считывает целое положительное число $n$, $n \in [1:9]$ и выводит значение числа $n + \overline{nn} + \overline{nnn}$.

**Формат входных данных:** <br>
На вход программе подаётся одно целое положительное число $n$, $n \in [1:9]$.

**Формат выходных данных:** <br>
Программа должна вывести число $n + \overline{nn} + \overline{nnn}$.

**Примечание №1:** <br>
Для первого теста 1 + 11 + 111 = 123

## sample_input:
1

## sample_output:
123

````python
# date: 06.09.2021

a = input()
b = a + a
c = a + a + a
print(int(a) + int(b) + int(c))
````

````python
# date: 12.11.2022

print(int(number := input()) + int(number + number) + int(number + number + number))
````
