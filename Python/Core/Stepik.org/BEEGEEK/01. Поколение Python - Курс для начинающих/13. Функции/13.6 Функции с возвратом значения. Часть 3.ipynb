# step_03 - Середина отрезка
Напишите функцию *get_middle_point(x1, y1, x2, y2)*, которая принимает в качестве аргументов координаты концов отрезка ($x_1$; $y_1$) и ($x_2$; $y_2$) и возвращает координаты точки являющейся серединой данного отрезка.

**Примечание №1:** <br>
Координаты середины отрезка с концами в точках ($x_1$; $y_1$); и ($x_2$; $y_2$) вычисляются по формуле:
$$\left(\dfrac{x_1+x_2}{2}; \dfrac{y_1+y_2}{2}\right)$$

```python
# date: 13.12.2022

def get_middle_point(x_1: float, y_1: float, x_2: float, y_2: float) -> float:
    mid_point_x: float = ((x_1 + x_2) / 2)
    mid_point_y: float = ((y_1 + y_2) / 2)

    print(mid_point_x, mid_point_y)

get_middle_point(float(input()), float(input()), float(input()), float(input()))
```

# step_04 - Площадь и длина
Напишите функцию *get_circle(radius)*, которая принимает в качестве аргумента радиус окружности и возвращает два значения: длину окружности и площадь круга, ограниченного данной окружностью.

**Примечание №1:** <br>
Длина окружности и площадь круга радиуса rr вычисляются по формулам:
$$С = 2 \pi r, S = \pi r^2$$

**Примечание №2:** <br>
Для числа $$\piπ$$ используйте глобальную константу из модуля *math*.

## sample_input:
1.5 <br>

## sample_output:
9.42477796076938 7.0685834705770345 <br>

```python
# date: 13.12.2022

def get_circle(radius):
    from math import pi
    c = 2 * pi * radius
    s = pi * radius ** 2

    return c, s

# считываем данные
r = float(input())

# вызываем функцию
length, square = get_circle(r)
print(length, square)
```

# step_05 - Корни уравнения 🌶️🌶️
Напишите функцию *solve(a, b, c)*, которая принимает в качестве аргументов три целых числа *a*, *b*, *c* – коэффициенты квадратного уравнения $ax^2+bx+c = 0$ и возвращает его корни в порядке возрастания.

**Примечание №1:** <br>
С подобной задачей мы уже сталкивались.

**Примечание №2:** <br>
Гарантируется, что квадратное уравнение имеет корни.

## sample_input:
-2 <br>
7 <br>
-5 <br>

## sample_output:
1.0 2.5 <br>

```python
# date: 13.12.2022

def solve(a, b, c):
    d = (b**2) -4*a*c
    x1 = ((-1*b)-d**0.5)/(2*a)
    x2 = ((-1*b)+d**0.5)/(2*a)

    return min(x1,x2), max(x1,x2)

a, b, c = int(input()), int(input()), int(input())

x1, x2 = solve(a, b, c)
print(x1, x2)
```
