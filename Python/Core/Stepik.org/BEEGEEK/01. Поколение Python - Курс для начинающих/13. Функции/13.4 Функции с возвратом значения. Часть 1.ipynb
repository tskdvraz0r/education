# step_06 - Конвертер километров
Напишите функцию *convert_to_miles(km)*, которая принимает в качестве аргумента расстояние в километрах и возвращает расстояние в милях. Формула для преобразования: мили = километры * 0.6214.

## sample_input:
5 <br>

## sample_output:
3.107 <br>

```python
# date: 12.11.2021

# объявление функции
def convert_to_miles(km):
    return km * 0.6214

# считываем данные
num = int(input())

# вызываем функцию
print(round(convert_to_miles(num), 3))
```

# step_07 - Количество дней
Напишите функцию *get_days(month)*, которая принимает в качестве аргумента номер месяца и возвращает количество дней в данном месяце.

**Примечание №1:** <br>
Гарантируется, что передаваемый аргумент находится в диапазоне от 1 до 12.

**Примечание №2:** <br>
Считайте, что год является невисокосным.

## sample_input:
9 <br>

## sample_output:
30 <br>

```python
# date: 12.11.2021

# объявление функции
def get_days(month):
    result = -1
    if month in (1, 3, 5, 7, 8, 10, 12):
        result = 31
    elif month in (4, 6, 9, 11):
        result = 30
    elif month == 2:
        result = 28
    return result

# считываем данные
num = int(input())

# вызываем функцию
print(get_days(num))
```

# step_08 - Делители 1
Напишите функцию *get_factors(num)*, принимающую в качестве аргумента натуральное число и возвращающую список всех делителей данного числа.

## sample_input:
5 <br>

## sample_output:
[1, 5] <br>

```python
# date: 12.11.2021

# объявление функции
def get_factors(num):
    result = []
    for i in range(1, num + 1):
        if num % i == 0:
            result.append(i)
    return result

# считываем данные
n = int(input())

# вызываем функцию
print(get_factors(n))
```

# step_10 - Найти всех
Напомним, что строковый метод *find('a')* возвращает местоположение первого вхождения символа *a* в строке. Проблема заключается в том, что данный метод не находит местоположение всех символов *а*.

Напишите функцию с именем *find_all(target, symbol)*, которая принимает два аргумента: строку *target* и символ *symbol* и возвращает список, содержащий все местоположения этого символа в строке.

**Примечание №1:** <br>
Если указанный символ не встречается в строке, то следует вернуть пустой список.

## sample_input:
abcdabcaaa <br>
a <br>

## sample_output:
[0, 4, 7, 8, 9] <br>

```python
# date: 12.11.2021

# объявление функции
def find_all(target, symbol):
    result = []
    for i in range(len(target)):
        if target[i] == symbol:
            result.append(i)

    return result

# считываем данные
s = input()
char = input()

# вызываем функцию
print(find_all(s, char))
```

# step_11 - Merge lists 1
Напишите функцию *merge(list1, list2)*, которая принимает в качестве аргументов два отсортированных по возрастанию списка, состоящих из целых чисел, и объединяет их в один отсортированный список.

**Примечание №1:** <br>
Списки *list1* и *list2* могут иметь разную длину.

**Примечание №2:** <br>
Можно использовать списочный метод *sort()*, а можно обойтись и без него 😎.

## sample_input:
[1, 7, 10, 16] <br>
[5, 6, 13, 20] <br>

## sample_output:
[1, 5, 6, 7, 10, 13, 16, 20] <br>

```python
# date: 12.11.2021

# объявление функции
def merge(list1, list2):
    l3 = list1 + list2
    l3.sort()
    return l3

# считываем данные
numbers1 = [int(c) for c in input().split()]
numbers2 = [int(c) for c in input().split()]

# вызываем функцию
print(merge(numbers1, numbers2))
```

# step_13 - Merge lists 2
На вход программе подается число $n$, а затем $n$ строк, содержащих целые числа в порядке возрастания. Из данных строк формируются списки чисел. Напишите программу, которая объединяет указанные списки в один отсортированный список с помощью функции *quick_merge()*, а затем выводит его.

**Формат входных данных:** <br>
На вход программе подается натуральное число $n$, а затем $n$ строк, содержащих целые числа в порядке возрастания, разделенные символом пробела.

**Формат выходных данных:** <br>
Программа должна вывести текст в соответствии с условием задачи.

## sample_input:
3 <br>
1 2 3 4 <br>
5 6 7 <br>
10 11 17 <br>

## sample_output:
1 2 3 4 5 6 7 10 11 17 <br>

```python
# date: 12.11.2021

total_list = []
for i in range(int(input())):
    total_list.append([int(c) for c in input().split()])

total_list2 = []
for j in total_list:
    for k in j:
        total_list2.append(k)

total_list2.sort()

print(*total_list2)
```
