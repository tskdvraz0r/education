# step_07 - Список строк
На вход программе подается натуральное число $n$, а затем $n$ строк. Напишите программу, которая создает из указанных строк список и выводит его.

**Формат входных данных:** <br>

# step_10 - Список делителей
На вход программе подается натуральное число $n$. Напишите программу, которая создает список состоящий из делителей введенного числа.

**Формат входных данных:** <br>
На вход программе подается натуральное число $n$.

**Формат выходных данных:** <br>
Программа должна вывести список, состоящий из делителей введенного числа.

## sample_input:
17 <br>

## sample_output:
[1, 17] <br>

```python
# date: 29.10.2021

n = int(input())
mylist = list()

for i in range(1, n + 1):
    if n % i == 0:
        mylist.append(i)

print(mylist)
```

# step_11 - Суммы двух
На вход программе подается натуральное число $n \ge 2$, а затем $n$ целых чисел. Напишите программу, которая создает из указанных чисел список, состоящий из сумм соседних чисел (0 и 1, 1 и 2, 2 и 3 и т.д.).

**Формат входных данных:** <br>
На вход программе подаются натуральное число $n$, а затем $n$ целых чисел, каждое на отдельной строке.

**Формат выходных данных:** <br>
Программа должна вывести список, состоящий из сумм соседних чисел.

## sample_input:
5 <br>
1 <br>
2 <br>
3 <br>
4 <br>
5 <br>

## sample_output:
[3, 5, 7, 9] <br>

```python
# date: 29.10.2021

n = int(input())
num1 = int(input())
mylist = []

for i in range(n - 1):
    num2 = int(input())
    mylist.append(num1 + num2)
    num1 = num2

print(mylist)
```

# step_12 - Удалите нечетные индексы
На вход программе подается натуральное число $n$, а затем $n$ целых чисел. Напишите программу, которая создает из указанных чисел список, затем удаляет все элементы стоящие по нечетным индексам, а затем выводит полученный список.

**Формат входных данных:** <br>
На вход программе подаются натуральное число $n$, а затем $n$ целых чисел, каждое на отдельной строке.

**Формат выходных данных:** <br>
Программа должна вывести список в соответствии с условием задачи.

**Примечание №1:** <br>
Используйте оператор *del*.

## sample_input:
10 <br>
0 <br>
1 <br>
2 <br>
3 <br>
4 <br>
5 <br>
6 <br>
7 <br>
8 <br>
9 <br>

## sample_output:
[0, 2, 4, 6, 8] <br>

```python
# date: 29.10.2021

n = int(input())
num_list = []

if n == 1:
    num = int(input())
    num_list.append(num)

else:
    for i in range(n):
        n = int(input())
        num_list.append(n)

    del num_list[1::2]

print(num_list)
```

# step_13 - k-ая буква слова 🌶️🌶️
На вход программе подается натуральное число $n$ и $n$ строк, а затем число $k$. Напишите программу, которая выводит $k$-ую букву из введенных строк на одной строке без пробелов.

**Формат входных данных:** <br>
На вход программе подается натуральное число $n$, далее $n$ строк, каждая на отдельной строке. В конце вводится натуральное число $k$ – номер буквы (нумерация начинается с единицы).

**Формат выходных данных:** <br>
Программа должна вывести текст в соответствии с условием задачи.

**Примечание №1:** <br>
Если некоторые строки слишком короткие, и в них нет символа с заданным номером, то такие строки при выводе нужно игнорировать.

## sample_input:
5 <br>
abcdef <br>
bcdefg <br>
cdefgh <br>
defghi <br>
efghij <br>
2 <br>

## sample_output:
bcdef <br>

```python
# date: 29.10.2021

mylist = []
text = str()

for i in range(int(input())):
    s = input()
    mylist.append(s)

k = int(input())

for j in range(len(mylist)):
    if len(mylist[j]) >= k:
        text += mylist[j][k - 1]

print(text)
```

# step_14 - Символы всех строк
На вход программе подается натуральное число $n$, а затем $n$ строк. Напишите программу, которая создает список из символов всех строк, а затем выводит его.

**Формат входных данных:** <br>
На вход программе подаются натуральное число $n$, а затем $n$ строк, каждая на отдельной строке.

**Формат выходных данных:** <br>
Программа должна вывести список состоящий из символов всех введенных строк.

**Примечание №1:** <br>
В результирующем списке могут содержаться одинаковые символы.

## sample_input:
3 <br>
abc <br>
def <br>
ghi <br>

## sample_output:
['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i'] <br>

```python
# date: 29.10.2021

mylist = []

for i in range(int(input())):
    s = input()
    mylist.extend(s)

print(mylist)
```
