# step_03 - Значение функции
На вход программе подается натуральное число $n$, а затем $n$ целых чисел. Напишите программу, которая для каждого введенного числа xx выводит значение функции $f(x) = x^2 + 2x + 1$, каждое на отдельной строке.

**Формат входных данных:** <br>
На вход программе подаются натуральное число $n$, а затем $n$ целых чисел, каждое на отдельной строке.

**Формат выходных данных:** <br>
Программа должна вывести сначала введенные числа, затем пустую строку, а затем соответствующие значения функции.

**Примечание №1:** <br>
Для первого теста имеем:
* $f(1) = 1^2 + 2\cdot 1 + 1 = 4$, $f(2) = 2^2 + 2\cdot 2 + 1 = 9$, $f(3) = 3^2 + 2 \cdot 3 + 1 = 16$, $\ldots$

## sample_input:
5 <br>
1 <br>
2 <br>
3 <br>
4 <br>
5 <br>

## sample_output:
1 <br>
2 <br>
3 <br>
4 <br>
5 <br>
 <br>
4 <br>
9 <br>
16 <br>
25 <br>
36 <br>

```python
# date: 31.10.2021

n = int(input())
num_list = []

for i in range(n):
    num = int(input())
    num_list.append(num)

print(*num_list, sep='\n')
print()

for j in num_list:
    f = j ** 2 + 2 * j + 1
    print(f)
```

# step_04 - Remove outliers
При анализе данных, собранных в рамках научного эксперимента, бывает полезно удалить самое большое и самое маленькое значение.

На вход программе подается натуральное число $n$, а затем $n$ различных натуральных чисел. Напишите программу, которая удаляет наименьшее и наибольшее значение из указанных чисел, а затем выводит оставшиеся числа каждое на отдельной строке, не меняя их порядок.

**Формат входных данных:** <br>
На вход программе подаются натуральное число $n$, а затем $n$ различных натуральных чисел, каждое на отдельной строке.

**Формат выходных данных:** <br>
Программа должна вывести текст в соответствии с условием задачи.

## sample_input:
10 <br>
9 <br>
17 <br>
189 <br>
3 <br>
55 <br>
78 <br>
11 <br>
7 <br>
888 <br>
160 <br>

## sample_output:
9 <br>
17 <br>
189 <br>
55 <br>
78 <br>
11 <br>
7 <br>
160 <br>

```python
# date: 31.10.2021

n = int(input())
num_list = []
# --------------------------------------------------
for i in range(n):
    num_list.append(int(input()))
# --------------------------------------------------
num_list.remove(min(num_list))
num_list.remove(max(num_list))

print(*num_list, sep='\n')
```

# step_05 - Без дубликатов
На вход программе подается натуральное число $n$, а затем $n$ строк. Напишите программу, которая выводит только уникальные строки, в том же порядке, в котором они были введены.

**Формат входных данных:** <br>
На вход программе подаются натуральное число $n$, а затем $n$ строк, каждая на отдельной строке.

**Формат выходных данных:** <br>
Программа должна вывести текст в соответствии с условием задачи.

**Примечание №1:** <br>
Считайте, что все строки состоят из строчных символов.

## sample_input:
5 <br>
first <br>
second <br>
first <br>
third <br>
second <br>

## sample_output:
first <br>
second <br>
third <br>

```python
# date: 31.10.2021

string_list = []  # создаём пустой список для сбора информации
# --------------------------------------------------
for i in range(int(input())):  # запускаем цикл n-ное количество раз (число вводится пользователем)
    new_string = input()  # ввод строки пользователем
    string_list.append(new_string)  # добавление новой строки в список
# --------------------------------------------------
total_string_list = []  # создаём спимок для фильтрации дублей
for j in range(len(string_list)):  # запускаем цикл
    if string_list[j] not in total_string_list:  # определяем наличие элемента (по индексу) в списке (без дублей)
        total_string_list.append(string_list[j])  # добавление элемента при отсутствии оного
# --------------------------------------------------
print(*total_string_list, sep='\n')  # распаковка и вывод построчно
# --------------------------------------------------
```

# step_06 - Google search - 1
На вход программе подается натуральное число $n$, затем $n$ строк, затем еще одна строка — поисковый запрос. Напишите программу, которая выводит все введенные строки, в которых встречается поисковый запрос.

**Формат входных данных:** <br>
На вход программе подаются натуральное число $n$ — количество строк, затем сами строки в указанном количестве, затем один поисковый запрос.

**Формат выходных данных:** <br>
Программа должна вывести все введенные строки, в которых встречается поисковый запрос.

**Примечание №1:** <br>
Поиск не должен быть чувствителен к регистру символов.

## sample_input:
5 <br>
Язык Python прекрасен <br>
C# - отличный язык программирования <br>
Stepik - отличная платформа <br>
BEEGEEK FOREVER! <br>
язык Python появился 20 февраля 1991 <br>
язык <br>

## sample_output:
Язык Python прекрасен <br>
C# - отличный язык программирования <br>
язык Python появился 20 февраля 1991 <br>

```python
# date: 31.10.2021

string_list = []
# --------------------------------------------------
for i in range(int(input())):
    new_string = input()
    string_list.append(new_string)

search_string = input()
# --------------------------------------------------
search_result = []

for i in range(len(string_list)):
    if search_string.lower() in string_list[i].lower():
        search_result.append(string_list[i])
# --------------------------------------------------
print(*search_result, sep='\n')
```

# step_07 - Google search - 2 🌶️🌶️
На вход программе подается натуральное число $n$, затем $n$ строк, затем число $k$ — количество поисковых запросов, затем $k$ строк — поисковые запросы. Напишите программу, которая выводит все введенные строки, в которых встречаются все поисковые запросы.

**Формат входных данных:** <br>
На вход программе подаются натуральное число $n$ — количество строк, затем сами строки в указанном количестве, затем число $k$, затем сами поисковые запросы.

**Формат выходных данных:** <br>
Программа должна вывести все введенные строки, в которых встречаются все поисковые запросы.

**Примечание №1:** <br>
Поиск не должен быть чувствителен к регистру символов.

## sample_input:
5 <br>
Язык Python прекрасен <br>
C# - отличный язык программирования <br>
Stepik - отличная платформа <br>
BEEGEEK FOREVER! <br>
язык Python появился 20 февраля 1991 <br>
2 <br>
язык <br>
python <br>

## sample_output:
Язык Python прекрасен <br>
язык Python появился 20 февраля 1991 <br>

```python
# date: 31.10.2021

string_list = []  # список строк
# --------------------------------------------------
for i in range(int(input())):  # запускаем цикл (n-раз задаётся пользователем)
    new_string = input()  # считываем строковые данные
    string_list.append(new_string)  # добавляем данные в список
# --------------------------------------------------
search_list = []  #
# --------------------------------------------------
for j in range(int(input())):  # запускаем цикл (n-раз задаётся пользователем)
    new_search_string = input()  # считываем строковые данные запросов
    search_list.append(new_search_string)  # добавление данных запросов в список
# --------------------------------------------------
search_result = []  # список из запросов
# --------------------------------------------------
for string in range(len(string_list)):  # цикл перебирает "строки" из списка
    flag = True
    for search in range(len(search_list)):  # цикл перебирает "запросы" из списка
        if search_list[search].lower() not in string_list[string].lower():  # ищем "запрос" в "поисковой строке"
            flag = False  # если "запроса" нет, флаг - False
    if flag == True:  # если флаг, после поиска всех "запросов", остался - True
        search_result.append(string_list[string])  # добавляем строку в список к выводу
# --------------------------------------------------
print(*search_result, sep='\n')  # распаковываем список и выводим результат построчно
```

# step_08 - Negatives, Zeros and Positives
На вход программе подается натуральное число $n$, а затем $n$ целых чисел. Напишите программу, которая сначала выводит все отрицательные числа, затем нули, а затем все положительные числа, каждое на отдельной строке. Числа должны быть выведены в том же порядке, в котором они были введены.

**Формат входных данных:** <br>
На вход программе подаются натуральное число $n$, а затем $n$ целых чисел, каждое на отдельной строке.

**Формат выходных данных:** <br>
Программа должна вывести текст в соответствии с условием задачи.

## sample_input:
7 <br>
3 <br>
-4 <br>
1 <br>
0 <br>
-1 <br>
0 <br>
-2 <br>

## sample_output:
-4 <br>
-1 <br>
-2 <br>
0 <br>
0 <br>
3 <br>
1 <br>

```python
# date: 31.10.2021

negative_num_list = []
zero_num_list = []
positive_num_list = []

for i in range(int(input())):
    n = int(input())
    if n < 0:
        negative_num_list.append(n)
    elif n == 0:
        zero_num_list.append(n)
    else:
        positive_num_list.append(n)

print(*negative_num_list, sep='\n')
print(*zero_num_list, sep='\n')
print(*positive_num_list, sep='\n')
```
