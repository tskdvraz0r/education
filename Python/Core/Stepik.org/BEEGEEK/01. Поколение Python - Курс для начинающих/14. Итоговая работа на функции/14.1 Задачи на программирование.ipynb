# step_04 - Звездный треугольник 🌶️
Напишите функцию *draw_triangle()*, которая выводит звездный равнобедренный треугольник с основанием и высотой равными 15 и 8 соответственно:
```python
       *
      ***
     *****
    *******
   *********
  ***********
 *************
***************
```

**Примечание №1:** <br>
Для вывода треугольника используйте цикл for.

**Примечание №2:** <br>
Справа от звездочек пробелов нет.

````python
# date: 23.11.2022

def draw_triangle():
    n = 15
    for i in range(1, n + 1, 2):
        print(' ' * ((n - i) // 2) + '*' * i)

draw_triangle()
````

# step_05 - Калькулятор доставки
Интернет магазин осуществляет экспресс доставку для своих товаров по цене 10001000 рублей за первый товар и 120 рублей за каждый последующий товар. Напишите функцию *get_shipping_cost(quantity)*, которая принимает в качестве аргумента натуральное число *quantity* – количество товаров в заказе и возвращает стоимость доставки.

## sample_input:
print(get_shipping_cost(3))

## sample_output:
1240

````python
# date: 23.11.2022

# объявление функции
def get_shipping_cost(quantity):
    return 1000 + ((quantity - 1) * 120)

# считываем данные
n = int(input())

# вызываем функцию
print(get_shipping_cost(n))
````

# step_06 - Биномиальный коэффициент 🌶️
Напишите функцию *compute_binom(n, k)*, которая принимает в качестве аргументов два натуральных числа $n$ и $k$ и возвращает значение биномиального коэффициента, равного $\dfrac{n!}{k! (n-k)!}$.

**Примечание №1:** <br>
Факториалом натурального числа $n$, называется произведение всех натуральных чисел от 1 до $n$, то есть $n!=1\cdot2\cdot3\cdot…\cdot$

**Примечание №2:** <br>
Реализуйте вспомогательную функцию *factorial(n)*, вычисляющую факториал числа или воспользуйтесь уже готовой функцией из модуля *math*.

-----

````python
# date: 23.11.2022

# объявление функции
def compute_binom(n, k):
    from math import factorial

    return int(factorial(n) / (factorial(k) * factorial(n - k)))

# считываем данные
n: int = int(input())
k: int = int(input())

# вызываем функцию
print(compute_binom(n, k))
````

# step_07 -  Число словами 🌶️
Напишите функцию *number_to_words(num)*, которая принимает в качестве аргумента натуральное число num и возвращает его словесное описание на русском языке.

**Примечание №1:** <br>
Считайте, что число $1 \le num \le 99$.

````python
# date: 23.11.2022

def number_to_words(num):
    s = ["один", "два", "три", "четыре", "пять", "шесть", "семь", "восемь", "девять", "десять", "одиннадцать", "двенадцать", "тринадцать", "четырнадцать", "пятнадцать", "шестнадцать", "семнадцать", "восемнадцать", "девятнадцать", "двадцать", "тридцать", "сорок", "пятьдесят", "шестьдесят", "семьдесят", "восемьдесят", "девяносто", ""]

    if num <= 20:
        return s[num - 1]

    else:
        return s[num // 10 - 1 + 18] + " " + s[num % 10 - 1]

n = int(input())

print(number_to_words(n))
````

# step_08 - Искомый месяц
Напишите функцию *get_month(language, number)*, которая принимает на вход два аргумента *language* – язык ru или en и *number* – номер месяца (от 1 до 12) и возвращает название месяца на русском или английском языке.

````python
# date: 23.11.2022

# объявление функции
def get_month(language, number):
    list1 = ['', 'январь', 'февраль', 'март', 'апрель', 'май', 'июнь', 'июль', 'август', 'сентябрь', 'октябрь', 'ноябрь', 'декабрь']
    list2 = ['', 'january', 'february', 'march', 'april', 'may', 'june', 'july', 'august', 'september', 'october', 'november', 'december']

    return list1[number] if language == 'ru' else list2[number]


# считываем данные
lan = input()
num = int(input())

# вызываем функцию
print(get_month(lan, num))
````

# step_09 - Магические даты
Магическая дата – это дата, когда день, умноженный на месяц, равен числу образованному последними двумя цифрами года.

Напишите функцию, *is_magic(date)* которая принимает в качестве аргумента строковое представление корректной даты и возвращает значение *True* если дата является магической и *False* в противном случае.

````python
# date: 23.11.2022

# объявление функции
def is_magic(date):
    day, month, year = date.split('.')
    return int(day) * int(month) == int(year) % 100

# считываем данные
date = input()

# вызываем функцию
print(is_magic(date))
````

# step_10 - Панграммы
Панграмма – это фраза, содержащая в себе все буквы алфавита. Обычно панграммы используют для презентации шрифтов, чтобы можно было в одной фразе рассмотреть все глифы.

Напишите функцию, *is_pangram(text)* которая принимает в качестве аргумента строку текста на английском языке и возвращает значение *True*, если текст является панграммой и *False* в противном случае.

**Примечание №1:** <br>
Гарантируется, что введенная строка содержит только буквы английского алфавита.

````python
# date:

# объявление функции
def is_pangram(text):
    letters = ['q', 'w', 'e', 'r', 't', 'y', 'u', 'i', 'o', 'p', 'a', 's', 'd', 'f', 'g', 'h', 'j', 'k', 'l', 'z', 'x', 'c', 'v', 'b', 'n', 'm']

    for letter in text.lower():
        if letter in letters:
            letters.remove(letter)

    return len(letters) == 0

# считываем данные
text = input()

# вызываем функцию
print(is_pangram(text))
````
