# step_01 - Список четных
На вход программе подается четное число $n$, $n \ge 2$. Напишите программу, которая выводит список четных чисел *[2, 4, 6, ..., n]*.

**Формат входных данных:** <br>
На вход программе подается четное натуральное число.

**Формат выходных данных:** <br>
Программа должна вывести текст в соответствии с условием задачи.

## sample_input:
2 <br>

## sample_output:
[2] <br>

```python
# date: 11.11.2021

print([i for i in range(int(input()) + 1) if i > 1 and i % 2 == 0])
```

# step_02 - Сумма двух списков
На вход программе подаются две строки текста, содержащие целые числа. Из данных строк формируются списки чисел *L* и *M*. Напишите программу, которая создает третий список, элементами которого являются суммы соответствующих элементов списков *L* и *M*. Далее программа должна вывести каждый элемент полученного списка на одной строке через 1 пробел.

**Формат входных данных:** <br>
На вход программе подаются две строки текста, содержащие целые числа, разделенные символом пробела.

**Формат выходных данных:** <br>
Программа должна вывести текст в соответствии с условием задачи.

**Примечание №1:** <br>
Количество чисел в обеих строках одинаковое.

## sample_input:
3 1 4 <br>
1 5 9 <br>

## sample_output:
4 6 13 <br>

```python
# date: 11.11.2021

s1 = input().split()
s2 = input().split()
n1 = []
n2 = []
n3 = []

for i in s1:
    n1.append(int(i))

for j in s2:
    n2.append(int(j))

for k in range(len(s1)):
    n3.append(n1[k] + n2[k])

print(*n3)
```

# step_03 - Сумма чисел
На вход программе подается строка текста, содержащая натуральные числа. Напишите программу, которая вставляет между каждым числом знак +, а затем вычисляет сумму полученных чисел.

**Формат входных данных:** <br>
На вход программе подается строка текста, содержащая натуральные числа, разделенные символом пробела.

**Формат выходных данных:** <br>
Программа должна вывести текст в соответствии с условием задачи.

**Примечание №1:** <br>
Строковый метод *join()* работает только со списком строк.

## sample_input:
2 5 11 33 55 <br>

## sample_output:
2+5+11+33+55=106 <br>

```python
# date: 11.11.2021

s = input().split()
t = str()
n = 0

for i in s:
    t += i + '+'
    n += int(i)

print(t[:-1], '=', n, sep='')
```

# step_04 - Валидный номер 🌶️🌶️
На вход программе подается строка текста. Напишите программу, которая определяет являться ли введенная строка корректным телефонным номером. Строка текста является корректным телефонным номером если она имеет формат:
*abc-def-hijk* или *7-abc-def-hijk* <br>
где *a, b, c, d, e, f, h, i, j, k* – цифры от 0 до 9.

**Формат входных данных:** <br>
На вход программе подается строка текста.

**Формат выходных данных:** <br>
Программа должна вывести *«YES»* если строка является корректным телефонным номером и *«NO»* в противном случае.

**Примечание №1:** <br>
Телефонный номер должен содержать только цифры и символ *-*, а количество цифр в каждой группе должны быть правильным.

## sample_input:
7-301-447-5820 <br>

## sample_output:
YES <br>

```python
# date: 11.11.2021

s = input().split('-')
flag = True

if len(s) == 4 and len(s[0]) == 1 and s[0] == '7' and len(s[1]) == 3 and len(s[2]) == 3 and len(s[3]) == 4:
    for i in range(len(s)):
        if s[i].isdigit():
            flag = True
        else:
            flag = False
            break

elif len(s) == 3 and len(s[0]) == 3 and len(s[1]) == 3 and len(s[2]) == 4:
    for i in range(len(s)):
        if s[i].isdigit():
            flag = True
        else:
            flag = False
            break

else:
    flag = False

if flag == True:
    print('YES')

else:
    print('NO')
```

На вход программе подается строка текста. Напишите программу, использующую списочное выражение, которая находит длину самого длинного слова.

## sample_input:

```python
# date: 11.11.2021

print(max([len(i) for i in input().split()]))
```

# step_06 - Молодежный жаргон
На вход программе подается строка текста. Напишите программу, использующую списочное выражение, которая преобразует каждое слово введенного текста в *"молодежный жаргон"* по следующему правилу:
* первая буква каждого слова удаляется и ставится в конец слова;
* затем в конец слова добавляется слог "ки".

**Формат входных данных:** <br>
На вход программе подается строка текста на русском языке.

**Формат выходных данных:** <br>
Программа должна вывести текст в соответствии с условием задачи.

## sample_input:
проспал почти всю ночь <br>

## sample_output:
роспалпки очтипки сювки очьнки <br>

```python
# date: 11.11.2021

print(*[i[1:] + i[0] + 'ки' for i in input().split()])
```
