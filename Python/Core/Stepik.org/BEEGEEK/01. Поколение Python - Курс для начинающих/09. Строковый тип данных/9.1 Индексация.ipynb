# step_07 - В столбик 1
На вход программе подается одна строка. Напишите программу, которая выводит элементы строки с индексами $0$, $2$, $4$, ... в столбик.

**Формат входных данных:** <br>
На вход программе подается одна строка.

**Формат выходных данных:** <br>
Программа должна вывести элементы строки с индексами $0$, $2$, $4$, ..., каждое на отдельной строке.

## sample_input:
abcdefghijklmnop <br>

## sample_output:
a <br>
c <br>
e <br>
g <br>
i <br>
k <br>
m <br>
o <br>

```python
# date: 27.11.2021

string = input()
for i in range(0, len(string), 2):
    print(string[i])
```

# step_10 - Цифра 1
На вход программе подается одна строка состоящая из цифр. Напишите программу, которая считает сумму цифр данной строки.

**Формат входных данных:** <br>
На вход программе подается одна строка состоящая из цифр.

**Формат выходных данных:** <br>
Программа должна вывести сумму цифр данной строки.

2514 <br>

# step_11 - Цифра 2
На вход программе подается одна строка. Напишите программу, которая выводит сообщение *«Цифра»* (без кавычек), если строка содержит цифру. В противном случае вывести сообщение *«Цифр нет»*8 (без кавычек).

**Формат входных данных:** <br>
На вход программе подается одна строка.

**Формат выходных данных:** <br>
Программа должна вывести текст в соответствии с условием задачи.

## sample_input:
Hi Python

## sample_output:
Цифр нет

```python
# date: 27.11.2021

string = input()
flag = True

for i in range(len(string)):
    if string[i] in '1234567890':
        flag = False

if flag == True:
    print('Цифр нет')
else:
    print('Цифра')
```

# step_12 - Сколько раз?
На вход программе подается одна строка. Напишите программу, которая определяет сколько раз в строке встречаются символы + и *.

**Формат входных данных:** <br>
На вход программе подается одна строка.

**Формат выходных данных:** <br>
Программа должна вывести сколько раз встречаются символы + и * в строке.

## sample_input:
bcd+a++++**31415 <br>

## sample_output:
Символ + встречается 5 раз <br>
Символ * встречается 2 раз <br>

```python
# date: 27.11.2021

string = input()
count_mult = 0
count_plus = 0

for i in range(len(string)):

    if string[i] == '*':
        count_mult += 1
    if string[i] == '+':
        count_plus += 1

print('Символ + встречается', count_plus, 'раз')
print('Символ * встречается', count_mult, 'раз')
```

# step_13 - Одинаковые соседи
На вход программе подается одна строка. Напишите программу, которая определяет сколько в ней одинаковых соседних символов.

**Формат входных данных:** <br>
На вход программе подается одна строка.

**Формат выходных данных:** <br>
Программа должна вывести количество одинаковых соседних символов.

## sample_input:
abcd <br>

## sample_output:
0 <br>

```python
# date: 27.11.2021

string = input()
first_char = str()
second_char = str()
count = 0

for i in range(len(string)):
    second_char = string[i]
    if second_char == first_char:
        count += 1
    first_char = second_char

print(count)
```

# step_14 - Гласные и согласные
На вход программе подается одна строка с буквами русского языка. Напишите программу, которая определяет количество гласных и согласных букв.

**Формат входных данных:** <br>
На вход программе подается одна строка.

**Формат выходных данных:** <br>
Программа должна вывести количество гласных и согласных букв.

**Примечание:** <br>
В русском языке 10 гласных букв (а, у, о, ы, и, э, я, ю, ё, е) и 21 согласная буква (б, в, г, д, ж, з, й, к, л, м, н, п, р, с, т, ф, х, ц, ч, ш, щ).

## sample_input:
Вдохновение – это умение приводить себя в рабочее состояние! <br>

## sample_output:
Количество гласных букв равно 25 <br>
Количество согласных букв равно 24 <br>

```python
# date: 27.11.2021

string = input()
count_gl = 0
count_slg = 0

for i in range(len(string)):
    if string[i] in 'АаУуОоЫыИиЭэЯяЮюЁёЕе':
        count_gl += 1
    if string[i] in 'БбВвГгДдЖжЗзЙйКкЛлМмНнПпРрСсТтФфХхЦцЧчШшЩщ':
        count_slg += 1

print('Количество гласных букв равно', count_gl)
print('Количество согласных букв равно', count_slg)
```

# step_15 - Decimal to Binary
На вход программе подается натуральное число, записанное в десятичной системе счисления. Напишите программу, которая переводит данное число в двоичную систему счисления.

**Формат входных данных:** <br>
На вход программе подается одно натуральное число.

**Формат выходных данных:** <br>
Программа должна вывести число записанное в двоичной системе счисления.

## sample_input:
5 <br>

## sample_output:
101 <br>

```python
# date: 27.11.2021

print(bin(int(input()))[2:])
```
