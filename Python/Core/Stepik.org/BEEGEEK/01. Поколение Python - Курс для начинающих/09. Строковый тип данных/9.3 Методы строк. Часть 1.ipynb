# step_08 - Заглавные буквы
На вход программе подается строка состоящая из имени и фамилии человека, разделенных одним пробелом. Напишите программу, которая проверяет, что имя и фамилия начинаются с заглавной буквы.

**Формат входных данных:** <br>
На вход программе подается строка.

**Формат выходных данных:** <br>
Программа должна вывести *«YES»* если имя и фамилия начинаются с заглавной буквы и *«NO»* в противном случае.

**Примечание №1:** <br>
Строка содержит только буквы.

## sample_input:
chris alan <br>

## sample_output:
NO <br>

```python
# date: 28.10.2022

string = input()

if string == string.title():
    print('YES')
else:
    print('NO')
```

# step_09 - sWAP cASE
На вход программе подается строка. Напишите программу, которая меняет регистр символов, другими словами замените все строчные символы заглавными и наоборот.

**Формат входных данных:** <br>
На вход программе подается строка.

**Формат выходных данных:** <br>

## sample_input:
Swap Case <br>

sWAP cASE <br>

# step_10 - Хороший оттенок
На вход программе подается строка текста. Напишите программу, которая определяет являться ли оттенок текста хорошим или нет. Текст имеет хороший оттенок, если содержит подстроку «хорош» во всевозможных регистрах.

**Формат входных данных:** <br>
На вход программе подается строка текста.

**Формат выходных данных:** <br>
Программа должна вывести *«YES»* если текст имеет хороший оттенок и *«NO»* в противном случае.

**Примечание №1:** <br>
Текст содержащий хорош, ХОРОШ, Хорош, хОРОШ и т.д. имеет хороший оттенок.

## sample_input:
я очень хороший текст =) <br>

## sample_output:
YES <br>

-----

```python
# date: 28.10.2022

string = input()
string = string.lower()

if 'хорош' in string:
    print('YES')
else:
    print('NO')
```

# step_11 - Нижний регистр
На вход программе подается строка. Напишите программу, которая подсчитывает количество буквенных символов в нижнем регистре.

**Формат входных данных:** <br>
На вход программе подается строка.

**Формат выходных данных:** <br>
Программа должна вывести количество буквенных символов в нижнем регистре.

## sample_input:
abcABCD12345 <br>

## sample_output:
3 <br>

```python
# date: 28.10.2022

string = input()
abc = 'qwertyuiopasdfghjklzxcvbnm'
count_abc = 0

for i in range(len(string)):
    if string[i] in abc:
        count_abc += 1

print(count_abc)
```
