# step_03
Реализуйте математическую функцию sign(x). Она определяется так:

$sign(x)=\begin{cases}
 & \text{1 if } x>0 \\
 & \text{0 if } x=0 \\
 & \text{-1 if } x<0
\end{cases}$

Значения х подаются на вход (подразумевается, что они подставляются системой из заложенных тестовых чисел). Результаты sign(x) вывести через print.

## sample_input:
2

## sample_output:
1

```python
# date: 13.11.2022

x: float = float(input())

if x > 0:
    print(1)

elif x == 0:
    print(0)

else:
    print(-1)
```

# step_04
Пусть координаты $x_1$, $y_1$ первой точки подаются на вход (подразумевается, что они подставляются системой из заложенных тестовых чисел). Пускай координаты $x_2$, $y_2$ второй точки заданы значениями 3, 6 соответственно.

Необходимо найти евклидово расстояние между двумя точками. Если расстояние строго больше epsilon, вывести на экран "Вне кластера" (с помощью print), иначе вывести "В кластере".

Евклидово расстояние задаем формулой r = $\sqrt{(x_1 - x_2)^2 + (y_1 - y_2)^2}$

Проверка на вхождение точки в эпсилон окрестность используется, например, в алгоритме DBSCAN, подробнее можно прочитать <a href = "https://habr.com/ru/post/322034/">тут</a>.

## sample_input:
3 6

## sample_output:
В кластере

```python
# date: 13.11.2022

x1, y1 = map(float, input().split())
x2, y2 = 3, 6
epsilon: int = 1
r: float = ((x1 - x2) ** 2 + (y1 - y2) ** 2) ** 0.5

print('Вне кластера' if r > epsilon else 'В кластере')
```

# step_05
Пусть a, b - переменные, подаваемые на вход (подразумевается, что они подставляются системой из заложенных тестовых чисел). Значения base_threshold, left_threshold, right_threshold заданы значениями 1, 2, 3 соответственно.

* В случае, если значение переменной a строго больше base_threshold - вывести на экран фразу "Правое ребро", иначе вывести фразу "Левое ребро" (- использовать print("Левое ребро") ).
* Если до этого была выведена фраза "Правое ребро", сравнить является ли переменная b строго больше right_threshold. Если да, вывести фразу "Правый лист", в противном случае вывести фразу "Левый лист".
* Если до этого была выведена фраза "Левое ребро", сравнить является ли переменная b строго больше left_threshold. Если да, вывести фразу "Правый лист", в противном случае вывести фразу "Левый лист".

Так выглядит этап прогноза в алгоритме решающего дерева, подробнее про него можно прочесть тут или в рамках соответствующего курса. Тут a, b - признаковые описания данных, в зависимости от выбранных признаков (и порогов) на каждом этапе разделения мы попадаем в тот или иной лист. Каждый лист имеет свой ответ, в зависимости от попавших в него обучающих данных. Приведенная задача, естественно, является упрощенной версией для понимания этапа прогноза обученного алгоритма.

## sample_input:
3 0

## sample_output:
Правое ребро
Левый лист

```python
# date: 13.11.2022

a, b = map(int, input().split())

base_threshold: int = 1
left_threshold: int  = 2
right_threshold: int = 3

if a > base_threshold:
    print('Правое ребро',
          'Правый лист' if b > right_threshold else 'Левый лист',
          sep = '\n')

else:
    print('Левое ребро',
          'Правый лист' if b > left_threshold else 'Левый лист',
          sep = '\n')
```

# step_06 - Прогноз логистической регрессией в задаче бинарной классификации
В среде инициализированы переменные a, b, в которые поступают числа из заложенного набора тестов. Кроме этого, вещественными числами 0.3, 0.1, -0.5 заданы переменные w0, w1, w2 соответственно. В машинном обучении W обычно называют весами (англ. weights).

Реализуйте формулу, которая вычисляет значение следующего (скалярного) произведения.
x = ${w_0 + w_1* a + w_2* b}$

Полученное произведение подайте на вход функции сигмоиды, определенной ниже. В качестве ee возьмите число 2.7182818

$\sigma (x)= {\frac{1}{1+e^{-x}}}$

Запишите результат сравнения $\sigma(x) > 0.5σ(x)>0.5$ в переменную prediction.



То, что вы сейчас реализовали - прогноз логистической регрессией в задаче бинарной классификации. Это один из важнейших алгоритмов машинного обучения. Он же является одним нейроном в нейронной сети (важно уточнить - с сигмоидальной активацией. Подробнее про физический смысл логистической регрессии можно почитать <a href="https://habr.com/ru/company/io/blog/265007/">тут</a> или <a href="https://habr.com/ru/company/ods/blog/323890/">тут</a> или в рамках соответствующего курса.

Реализация подобных алгоритмов на простых примерах с двумя числами - важный момент в освоении алгоритмов машинного обучения и анализа данных. Это позволяет лучше понять, как работают готовые реализации алгоритмов и какие проблемы с ними могут возникнуть при неправильной подготовке данных.

## sample_input:
2 -3

## sample_output:
True

```python
# date: 14.11.2022

a, b = map(int, input().split())
e = 2.7182818
w0, w1, w2 = 0.3, 0.1, -0.5

x = w0 + w1 * a + w2 * b
sigma = 1 / (1 + e ** (-x))
prediction = sigma > 0.5

print(prediction)
```
