В этом задании предлагается познакомиться с системой проверки.

Чтобы не вводить переменные с клавиатуры ( функция input() ) платформа сама подставляет значения в переменные. В задании в переменные a, b будут подставлены целые числа из заранее заданного набора тестов. Вам необходимо сложить эти два числа, результат записать в переменную c. Не удаляйте print(c) - так платформа может сверить ваш вывод с ответами, заложенными в наборах тестов.

Чуть более подробно, что происходит: текст '8 11' разбивается по пробелам с помощью .split(), к каждому элементу полученного списка с помощью map применяется функция явного приведения типов int(), полученные таким образом два целых числа сохраняется в a и b. Все эти методы и функции будут детально рассмотрены в следующих частях курса.

## sample_input:
8 11
## sample_output:
19

```python
# date: 13.11.2022

'''
Не меняйте этот код! Он необходим для автоматической проверки решений.
Целые числа из заложенного набора тестов подставляются в переменные а и b.
Вам необходимо произвести с ними операции (в задаче - сложения).
'''
a, b = map(int, input().split())

# Изменяйте только этот код (вместо ... )
c = a + b

# Не меняйте этот код! Он необходим для автоматической проверки решений.
print(c)
```

# step_06
Пусть a, b, c - стороны треугольника. Реализуйте формулу Герона для нахождения площади треугольника.

S = $\frac{{\sqrt{(a+b+c)(a+b-c)(a-b+c)(-a+b+c)}}}{4}$

## sample_input:
3 4 5

## sample_output:
6.0

```python
# date: 13.11.2022

'''
Не меняйте этот код! Он необходим для автоматической проверки решений.
'''
a, b, c = map(int, input().split())

# Изменяйте только этот код (вместо ... )
S = ((a + b + c) * (a + b - c) * (a - b + c) * (- a + b + c)) ** 0.5 / 4

# Не меняйте этот код! Он необходим для автоматической проверки решений.
print(S)
```
