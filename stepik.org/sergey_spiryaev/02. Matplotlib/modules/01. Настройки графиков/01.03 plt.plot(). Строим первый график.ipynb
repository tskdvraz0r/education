```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
values: dict[str, list[int]] = {
    "x": [2, 4, 8, 12, 15, 20, 25],
    "y": [0, 5, 4, 11, 12, 10, 9]
}

dataframe = pd.DataFrame(data=values)

display(dataframe)
```

```python
plt.plot()
```

```python
# Создание простого графика;
plt.plot([1, 2, 3, 4, 5])
plt.show()
```

```python
# Построение простого графика на основе дата фрейма;
plt.plot(
    dataframe["x"],
    dataframe["y"]
)
plt.show()
```
