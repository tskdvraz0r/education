```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Создание дата фрейма;
values: dict[str, list[int]] = {
    "x": [2, 4, 8, 12, 15, 20, 25],
    "y": [0, 5, 4, 11, 12, 10, 9]
}

dataframe = pd.DataFrame(
    data=values
)
```

```python
plt.plot(
    dataframe["x"],
    dataframe["y"]
)
plt.grid(
    color="gray",
    linestyle="--",
    linewidth=0.5
)
plt.title(
    label="График зависимостей переменных x и y",
    fontsize=16,
    color="red",
    alpha=0.75
)
plt.xlabel(
    xlabel="X",
    fontsize=14,
    color="red"
)
plt.ylabel(
    ylabel="Y",
    fontsize=14,
    color="red",
    rotation=0
)

plt.xticks(
    ticks=[0, 10, 20, 30]
)
plt.yticks(
    ticks=[0, 5, 10, 15]
)

plt.show()
```

```python
plt.plot(
    dataframe["x"],
    dataframe["y"]
)

plt.grid(
    color="gray",
    linestyle="--",
    linewidth=0.5
)

plt.title(
    label="График зависимостей переменных x и y",
    color="red",
    fontsize=16,
    alpha=1
)

plt.xlabel(
    xlabel="X",
    color="red",
    fontsize=14
)
plt.ylabel(
    ylabel="Y",
    color="red",
    fontsize=14,
    rotation=0
)

plt.xticks(
    ticks=range(0, 26)
)
plt.yticks(
    ticks=[0, 1, 2, 3, 4, 5, 10]
)

plt.show()
```

```python
plt.plot(
    dataframe["x"],
    dataframe["y"]
)

plt.grid(
    color="black",
    linestyle=":",
    linewidth=0.25
)

plt.title(
    label="График зависимостей переменных x и y",
    color="green",
    fontsize=16,
    alpha=0.75
)

plt.xlabel(
    xlabel="X",
    color="green",
    fontsize=12,
    labelpad=10,
    rotation=0,
)
plt.ylabel(
    ylabel="Y",
    color="green",
    fontsize=12,
    labelpad=10,
    rotation=0
)

plt.xticks(
    ticks=range(0, dataframe["x"].max()),
    rotation=45
)
plt.yticks(
    ticks=range(0, dataframe["y"].max()),
    rotation=45
)

plt.show()
```
