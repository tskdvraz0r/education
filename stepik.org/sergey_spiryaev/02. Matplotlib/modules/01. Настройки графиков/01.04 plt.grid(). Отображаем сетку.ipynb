```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
values: dict[str, list[int]] = {
    "x": [2, 4, 8, 12, 15, 20, 25],
    "y": [0, 5, 4, 11, 12, 10, 9]
}

dataframe = pd.DataFrame(
    data=values
)
```

```python
plt.plot(
    dataframe["x"],
    dataframe["y"]
)
plt.grid()

plt.show()
```

```python
plt.plot(
    dataframe["x"],
    dataframe["y"]
)
plt.grid(
    color="green",
    linestyle="--",
    linewidth=1
)

plt.show()
```
