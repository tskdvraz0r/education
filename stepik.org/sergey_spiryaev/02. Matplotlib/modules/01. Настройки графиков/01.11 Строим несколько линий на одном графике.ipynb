```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
values: dict[str, list[int]] = {
    "x1": [2, 4, 8, 12, 15, 20, 25],
    "x2": [1, 3, 7, 10, 13, 18, 22],
    "y1": [0, 5, 4, 11, 12, 10, 9],
    "y2": [3, 10, 15, 6, 5, 2, 1]
}

dataframe = pd.DataFrame(data=values)

dataframe
```

```python
plt.plot(
    dataframe["x1"],
    dataframe["y1"],
    dataframe["x2"],
    dataframe["y2"],
    linewidth=1,
    marker="."
)
```

```python
plt.plot(
    dataframe["x1"],
    dataframe["y1"],
    color="darkred",
    linewidth=1,
    marker="."
)

plt.show()

plt.plot(
    dataframe["x2"],
    dataframe["y2"],
    color="darkgreen",
    linewidth=1,
    marker="."
)

plt.show()
```
