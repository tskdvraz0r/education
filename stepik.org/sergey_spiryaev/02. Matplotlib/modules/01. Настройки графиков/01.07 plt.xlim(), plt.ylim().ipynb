```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
values: dict[str, list[int]] = {
    "x": [2, 4, 8, 12, 15, 20, 25],
    "y": [0, 5, 4, 11, 12, 10, 9]
}

dataframe = pd.DataFrame(
    data=values
)
```

```python
plt.plot(
    dataframe["x"],
    dataframe["y"],
)
plt.grid(
    color="gray",
    linestyle="--",
    linewidth=0.5
)
plt.title(
    label="График зафисимости признаков x и y",
    fontsize=16,
    color="black",
    alpha=0.75
)
plt.xlabel(
    xlabel=dataframe.columns[0],
    fontsize=12,
    color="black",
    alpha=0.75,
    labelpad=10
)
plt.ylabel(
    ylabel=dataframe.columns[1],
    fontsize=12,
    color="black",
    alpha=0.75,
    labelpad=10
)
plt.xlim(0, 26)
plt.ylim(0, 15)

plt.show()
```
