```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Создание дата фрейма;
values: dict[str, list[int]] = {
    "x": [2, 4, 8, 12, 15, 20, 25],
    "y": [0, 5, 4, 11, 12, 10, 9]
}

dataframe = pd.DataFrame(
    data=values
)
```

```python
# Построение простого графика на основе дата фрейма;
plt.plot(
    dataframe["x"],
    dataframe["y"]
)

# Добавление сетки с настройками на график;
plt.grid(
    color="gray",
    linestyle="--",
    linewidth=0.5
)

# Добавление наименования с настройками на график;
plt.title(
    label="График зависимостей x и y",
    fontsize=15,
    color="black",
    alpha=0.75
)

# Добавление наименований осей;
plt.xlabel(xlabel="X")
plt.ylabel(ylabel="Y")

plt.show()
```

```python
# Построение простого графика на основе дата фрейма;
plt.plot(
    dataframe["x"],
    dataframe["y"]
)

# Добавление сетки с настройками на график;
plt.grid(
    color="gray",
    linestyle="--",
    linewidth=0.5
)

# Добавление наименования с настройками на график;
plt.title(
    label="График зависимостей x и y",
    fontsize=15,
    color="black",
    alpha=0.75
)

# Добавление наименований осей;
plt.xlabel(dataframe.columns[0])
plt.ylabel(dataframe.columns[1])

plt.show()
```

```python
# Построение простого графика на основе дата фрейма;
plt.plot(
    dataframe["x"],
    dataframe["y"]
)

# Добавление сетки с настройками на график;
plt.grid(
    color="gray",
    linestyle="--",
    linewidth=0.5
)

# Добавление наименования с настройками на график;
plt.title(
    label="График зависимостей x и y",
    fontsize=15,
    color="black",
    alpha=0.75
)

# Добавление наименований осей с настройками;
plt.xlabel(
    xlabel=dataframe.columns[0],
    labelpad=10
)
plt.ylabel(
    ylabel=dataframe.columns[1],
    labelpad=10,
    rotation=0
)

plt.show()
```
