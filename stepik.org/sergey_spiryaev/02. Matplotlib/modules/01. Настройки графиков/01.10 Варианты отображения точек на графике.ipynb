```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
values: dict[str, list[int]] = {
    "x": [2, 4, 8, 12, 15, 20, 25],
    "y": [0, 5, 4, 11, 12, 10, 9]
}

dataframe = pd.DataFrame(
    data=values
)
```

```python
plt.plot(
    dataframe["x"],
    dataframe["y"],
    color="darkgreen",
    linewidth=1,
    marker=".",
    markersize=10
)

plt.show()
```

```python
plt.plot(
    dataframe["x"],
    dataframe["y"],
    color="darkgreen",
    linewidth=1,
    marker=".",
    markersize=10,
    markeredgecolor="red"
)

plt.show()
```

```python
plt.plot(
    dataframe["x"],
    dataframe["y"],
    color="darkgreen",
    linewidth=1,
    marker="o",
    markersize=10,
    markerfacecolor="red",
    markeredgecolor="black",
    markeredgewidth=1
)

plt.show()
```

```python
plt.plot(
    dataframe["x"],
    dataframe["y"],
    color="darkgreen",
    linestyle="None",
    linewidth=1,
    marker="o",
    markersize=10,
    markerfacecolor="red",
    markeredgecolor="black",
    markeredgewidth=1
)

plt.show()
```
