# Lesson 1

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
values: dict[str, list[int]] = {
    "x1": [2, 4, 8, 12, 15, 20, 25],
    "x2": [1, 3, 7, 10, 13, 18, 22],
    "y1": [0, 5, 4, 11, 12, 10, 9],
    "y2": [3, 10, 15, 6, 5, 2, 1]
}

dataframe = pd.DataFrame(data=values)

display(dataframe)
```

```python
plt.plot(
    dataframe["x1"],
    dataframe["y1"],
    label="x1_y1",
    color="darkgreen",
    linewidth=1,
    marker="o",
    markerfacecolor="green",
    markersize=5,
    markeredgewidth=1,
    markeredgecolor="black",
)

plt.plot(
    dataframe["x2"],
    dataframe["y2"],
    label="x2_y2",
    color="darkred",
    linewidth=1,
    marker="o",
    markerfacecolor="red",
    markersize=5,
    markeredgewidth=1,
    markeredgecolor="black",
)

plt.legend(
    ["x1_y1", "x2_y2"],
    loc="center"
)
plt.show()
```

```python
plt.plot(
    dataframe["x1"],
    dataframe["y1"],
    color="darkgreen",
    linewidth=1,
    marker="o",
    markerfacecolor="green",
    markersize=5,
    markeredgewidth=1,
    markeredgecolor="black",
)

plt.plot(
    dataframe["x2"],
    dataframe["y2"],
    color="darkred",
    linewidth=1,
    marker="o",
    markerfacecolor="red",
    markersize=5,
    markeredgewidth=1,
    markeredgecolor="black",
)

plt.legend(
    ["x1_y1", "x2_y2"],
    loc=(0.4, -0.25)
)
plt.show()
```

```python
for style in ("Solarize_Light2", "ggplot", "seaborn-v0_8"):
    plt.plot(
        dataframe["x1"],
        dataframe["y1"],
        color="darkgreen",
        linewidth=1,
        marker="o",
        markerfacecolor="green",
        markersize=5,
        markeredgewidth=1,
        markeredgecolor="black",
    )

    plt.plot(
        dataframe["x2"],
        dataframe["y2"],
        color="darkred",
        linewidth=1,
        marker="o",
        markerfacecolor="red",
        markersize=5,
        markeredgewidth=1,
        markeredgecolor="black",
    )

    plt.legend(
        ["x1_y1", "x2_y2"],
        loc=(0.4, -0.25)
    )

    plt.style.use(style=style)

    plt.show()
```
