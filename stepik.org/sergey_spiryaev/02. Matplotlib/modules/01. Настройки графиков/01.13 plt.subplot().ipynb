```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
values: np.ndarray = np.linspace(start=0, stop=10, num=5)
y1: np.ndarray = values ** 2
y2: np.ndarray = values ** 3
y3: np.ndarray = values ** 0.5
y4: np.ndarray = np.sin(values ** 2)
```

```python
plt.subplot(2, 2, 1)
plt.plot(
    values,
    y1,
    color="darkgreen",
    linewidth=1,
    marker="o",
    markerfacecolor="green",
    markersize=5,
    markeredgewidth=1,
    markeredgecolor="black",
)
plt.title(label="y1")

plt.subplot(2, 2, 2)
plt.plot(
    values,
    y2,
    color="darkred",
    linewidth=1,
    marker="o",
    markerfacecolor="red",
    markersize=5,
    markeredgewidth=1,
    markeredgecolor="black",
)
plt.title(label="y2")

plt.subplot(2, 2, 3)
plt.plot(
    values,
    y3,
    color="yellow",
    linewidth=1,
    marker="o",
    markerfacecolor="orange",
    markersize=5,
    markeredgewidth=1,
    markeredgecolor="black",
)
plt.title(label="y3")

plt.subplot(2, 2, 4)
plt.plot(
    values,
    y4,
    color="blue",
    linewidth=1,
    marker="o",
    markerfacecolor="royalblue",
    markersize=5,
    markeredgewidth=1,
    markeredgecolor="black",
)
plt.title(label="y4")

plt.suptitle('Графики y1-4')
plt.tight_layout()

plt.show()
```
