```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
values: dict[str, list[int]] = {
    "x": [2, 4, 8, 12, 15, 20, 25],
    "y": [0, 5, 4, 11, 12, 10, 9]
}

dataframe = pd.DataFrame(
    data=values
)
```

```python
plt.plot(
    dataframe["x"],
    dataframe["y"],
    color="darkgreen",
    linewidth=3,
    alpha=0.5
)

plt.grid(
    color="gray",
    linestyle="--",
    linewidth=0.5
)

plt.title(
    label="График зависимостей переменных x и y;",
    color="red",
    fontsize=16
)

plt.xlabel(
    xlabel="X",
    color="red",
    fontsize=14
)

plt.ylabel(
    ylabel="Y",
    color="red",
    fontsize=14
)

plt.show()
```

```python
plt.plot(
    dataframe["x"],
    dataframe["y"],
    color="darkgreen",
    linestyle="-",
    linewidth=3,
    alpha=0.5
)

plt.plot(
    dataframe["x"],
    dataframe["y"] + 2,
    color="darkred",
    linestyle="--",
    linewidth=3,
    alpha=0.5
)

plt.plot(
    dataframe["x"],
    dataframe["y"] + 4,
    linestyle="-.",
    color="royalblue",
    linewidth=3,
    alpha=0.5
)

plt.plot(
    dataframe["x"],
    dataframe["y"] + 6,
    color="coral",
    linestyle=":",
    linewidth=3,
    alpha=0.5
)

plt.grid(
    color="gray",
    linestyle="--",
    linewidth=0.5
)

plt.title(
    label="График зависимостей переменных x и y;",
    color="red",
    fontsize=16
)

plt.xlabel(
    xlabel="X",
    color="red",
    fontsize=14
)

plt.ylabel(
    ylabel="Y",
    color="red",
    fontsize=14
)

plt.show()
```
