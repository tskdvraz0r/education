```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator
```

```python
x = np.linspace(-np.pi, np.pi, 20)
cos_x = np.cos(x)
sin_x = np.sin(x)

print(
    x,
    cos_x,
    sin_x,
    sep="\n\n"
)
```

```python
fig = plt.figure(figsize=(21, 9))
ax = fig.add_axes([0, 0, 1, 1])

ax1 = ax.plot(
    x,
    cos_x,
    color="red",
    linestyle="--",
    linewidth=3,
    marker="o",
    markersize=10,
    markerfacecolor="darkred",
    markeredgecolor="black",
    markeredgewidth=0.5,
    label="x_cosx"
)
ax2 = ax.plot(
    x,
    sin_x,
    color="green",
    linestyle=":",
    linewidth=3,
    marker="p",
    markersize=10,
    markerfacecolor="darkgreen",
    markeredgecolor="black",
    markeredgewidth=0.5,
    label="x_sinx"
)

ax.grid(
    color="black",
    which="both"
)

ax.set_title(
    label="Графики тригонометрических функций",
    fontsize=32,
    color="darkred"
)

ax.set_xlabel(
    xlabel="X",
    color="darkred",
    fontsize=32,
    labelpad=16
)
ax.set_ylabel(
    ylabel="Y",
    fontsize=32,
    color="darkred",
    rotation=0,
    labelpad=16
)

ax.set_xlim(-4, 4)
ax.set_ylim(-1.5, 1.5)

ax.set_xticks([-4, -2, 0, 2, 4])
ax.xaxis.set_minor_locator(AutoMinorLocator(4))
ax.yaxis.set_minor_locator(AutoMinorLocator(2))
ax.tick_params(
    labelrotation=45,
    axis="x",
    labelsize=16,
    pad=16,
    direction="inout",
    length=16,
    width=2,
    which="both"
)
ax.tick_params(
    axis="y",
    labelcolor="red",
    labelsize=16,
    pad=16,
    direction="inout",
    length=16,
    width=2,
    color="darkred",
    which="both"
)

ax.tick_params(which="minor", labelsize=8, grid_alpha=0.1)
ax.xaxis.set_minor_formatter("{x:.1f}")
ax.yaxis.set_minor_formatter("{x:.2f}")

ax.legend(
    facecolor="lightgray",
    edgecolor="black",
    fontsize=16
)


plt.show()
```
