```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-np.pi, np.pi, 20)
cos_x = np.cos(x)
sin_x = np.sin(x)

print(
    x,
    cos_x,
    sin_x,
    sep="\n\n"
)
```

# Step 1 - Lesson

```python
fig = plt.figure(figsize=(21, 9))
ax = fig.add_axes([0, 0, 1, 1])

ax1 = ax.plot(
    x,
    cos_x,
    color="red",
    linestyle="--",
    linewidth=3,
    marker="o",
    markersize=10,
    markerfacecolor="darkred",
    markeredgecolor="black",
    markeredgewidth=0.5,
    label="x_cosx"
)
ax2 = ax.plot(
    x,
    sin_x,
    color="green",
    linestyle=":",
    linewidth=3,
    marker="p",
    markersize=10,
    markerfacecolor="darkgreen",
    markeredgecolor="black",
    markeredgewidth=0.5,
    label="x_sinx"
)

ax.grid(color="lightgray")
ax.set_title(
    label="Графики тригонометрических функций",
    fontsize=32,
    color="darkred"
)
ax.set_xlabel(
    xlabel="X",
    color="darkred",
    fontsize=32
)
ax.set_ylabel(
    ylabel="Y",
    fontsize=32,
    color="darkred",
    rotation=0
)

ax.legend(
    facecolor="lightgray",
    edgecolor="black",
    fontsize=16
)


plt.show()
```
