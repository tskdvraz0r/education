```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-np.pi, np.pi, 20)
cos_x = np.cos(x)
sin_x = np.sin(x)

print(
    x,
    cos_x,
    sin_x,
    sep="\n\n"
)
```

```python
fig = plt.figure(figsize=(21, 9), facecolor="lightgray")
ax = fig.add_axes([0, 0, 1, 1])

ax1 = ax.plot(x, sin_x)
ax2 = ax.plot(x, cos_x)

ax.legend(
    labels=["x_sinx", "x_cosx"],
    loc="upper right",
    facecolor="lightgray",
    edgecolor="black",
    fontsize=16
)

plt.show()
```

# Step 2 - Lesson

```python
fig = plt.figure(figsize=(21, 9))
ax = fig.add_axes([0, 0, 1, 1])

ax1 = ax.plot(
    x,
    sin_x,
    color="cyan",
    marker="o",
    markerfacecolor="blue",
    label="x_sinx"
)
ax2 = ax.plot(
    x,
    cos_x,
    color="red",
    marker="o",
    markerfacecolor="darkred",
    label="x_cosx"
)

ax.grid(color="gray")
ax.legend(
    loc="upper right",
    facecolor="lightgray",
    edgecolor="black",
    fontsize=16
)
ax.set_facecolor(color="#F3FCCC")

plt.show()
```
