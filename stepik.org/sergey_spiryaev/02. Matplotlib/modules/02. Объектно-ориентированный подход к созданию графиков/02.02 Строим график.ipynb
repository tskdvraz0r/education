```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-np.pi, np.pi, 20)
cos_x = np.cos(x)
sin_x = np.sin(x)

print(
    x,
    cos_x,
    sin_x,
    sep="\n\n"
)
```

```python
fig = plt.figure(figsize=(21, 9))
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, cos_x)
plt.show()
```

```python
fig = plt.figure(figsize=(21, 9))
ax = fig.add_axes([0, 0, 1, 1])

ax1 = ax.plot(x, cos_x)
ax2 = ax.plot(x, sin_x)

plt.show()
```
