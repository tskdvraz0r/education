# Step 2

```python
%%ipytest -v

def count_words_in_string(string: str) -> int:
    logger.debug(f"{string=}, {type(string)}")

    if not isinstance(string, str):
        error_message: str = "Некорректный тип данных;"
        logger.error(error_message)
        raise TypeError(error_message)

    return len(string.split())

@pytest.mark.parametrize(
    ["test_string", "expected_result"],
    [
        # Good cases
        ("Hello world", 2),
        ("Hello world!", 2),
        ("Hello world! How are you?", 5),
        ("Hello world! How are you? I am fine. Thank you!", 10),

        # Edge cases
        ("", 0),
        (" ", 0),
        ("\n", 0),
        ("\n\n", 0),

        # Error cases
        (None, TypeError),
        (1, TypeError),
        (True, TypeError),
    ]
)
def test_count_words_in_string(
        test_string: str,
        expected_result: int,
) -> None:
    if test_string in {None, 1, True}:
        with pytest.raises(expected_exception=expected_result):
            count_words_in_string(string=test_string)
    else:
        assert count_words_in_string(string=test_string) == expected_result

if __name__ == "__main__":
    try:
        logger.info("Запуск функции 'count_words_in_string';")
        print(count_words_in_string(string="some text"))
        logger.info("Функция 'count_words_in_string' успешно завершена;")
    except Exception:
        logger.error("Ошибка выполнения в функции 'count_words_in_string';")
```

```python
%%ipytest -v

def startwith_mam(string: str) -> bool:
    logger.debug(f"{string=}, {type(string)};")

    if not isinstance(string, str):
        error_message: str = "Некорректный тип данных;"
        logger.error(error_message)
        raise TypeError(error_message)

    return (
        string
        .lower()
        .startswith("mam")
    )

@pytest.mark.parametrize(
    ["test_string", "expected_result"],
    [
        # Good cases
        ("MaMba", True),
        ("Marka", False),
        ("mama", True),

        # Edge cases
        ("МАМА", False),
        ("", False),
        (" ", False),

        # Error cases
        (None, TypeError),
        (1, TypeError),
        (True, TypeError),
    ]
)
def test_startwith_mam(
        test_string: str,
        expected_result: bool,
) -> None:
    if test_string in {None, 1, True}:
        with pytest.raises(expected_exception=expected_result):
            startwith_mam(string=test_string)
    else:
        assert startwith_mam(string=test_string) == expected_result

if __name__ == "__main__":
    try:
        logger.info("Запуск функции 'startwith_mam';")
        print(startwith_mam(string="mam text"))
        logger.info("Функция 'startwith_mam' успешно завершена;")
    except Exception:
        logger.error("Ошибка выполнения в функции 'startwith_mam';")
```

# Step 6

```python
%%ipytest -v

def string_endswith(string_and_postfix: list[str]) -> bool:
    logger.debug(f"{string_and_postfix=}, {type(string_and_postfix)};")

    if not isinstance(string_and_postfix, list):
        error_message: str = "Некорректный тип данных;"
        logger.error(error_message)
        raise TypeError(error_message)

    if not all(isinstance(value, str) for value in string_and_postfix):
        error_message: str = "Некорректный тип данных;"
        logger.error(error_message)
        raise TypeError(error_message)

    string, postfix = string_and_postfix

    return (
        string
        .strip()
        .endswith(postfix)
    )

@pytest.mark.parametrize(
    ["test_string_and_postfix", "expected_result"],
    [
        # Good cases
        (["Привет, как дела", "дела"], True),
        (["Привет, Как Дела", "дела"], False),
        (["Eggs and bakery", "kery"], True),

        # Edge cases

        # Error cases
        (123, TypeError),
        (None, TypeError)
    ]
)
def test_string_endswith(
        test_string_and_postfix: list[str],
        expected_result
) -> None:
    if test_string_and_postfix in [123, None]:
        with pytest.raises(expected_exception=expected_result):
            string_endswith(string_and_postfix=test_string_and_postfix)
    else:
        assert string_endswith(string_and_postfix=test_string_and_postfix) == expected_result

if __name__ == "__main__":
    try:
        logger.info("Запуск функции 'string_endswith';")
        print(string_endswith(string_and_postfix=["Привет, как дела", "дела"]))
        logger.info("Функция 'string_endswith' успешно завершена;")
    except Exception:
        logger.error("Ошибка выполнения в функции 'string_endswith';")
```

# Step 7

```python
%%ipytest -v

def is_startswith_and_endswith(string_and_prepostfix: list[str, str, str]) -> bool:
    logger.debug(f"{string_and_prepostfix=}, {type(string_and_prepostfix)};")

    if not isinstance(string_and_prepostfix, list):
        type_error_message: str = "Некорректный тип данных;"
        logger.error(type_error_message)
        raise TypeError(type_error_message)

    if len(string_and_prepostfix) != 3:
        value_error_message: str = "Некорректное количество элементов;"
        logger.error(value_error_message)
        raise ValueError(value_error_message)

    if not all(isinstance(x, str) for x in string_and_prepostfix):
        type_error_message: str = "Некорректный тип данных;"
        logger.error(type_error_message)
        raise TypeError(type_error_message)

    string, prefix, postfix = string_and_prepostfix

    return string.strip().startswith(prefix.strip()) and string.strip().endswith(postfix.strip())

@pytest.mark.parametrize(
    ["test_strings", "expected_result"],
    [
        # Good cases
        (["translate russian to english", "translate", "english"], True),
        (["translate russian to english", "translate", "SH"], False),

        # Edge cases

        # Error cases
        (None, TypeError),
        (123, TypeError),
        (["1 2", "1", 33], TypeError),
        (["1 2 3", "1", "2", "3"], ValueError),
    ]
)
def test_is_startswith_and_endswith(
        test_strings: list[str, str, str],
        expected_result
) -> None:
    if test_strings in [123, None, ["1 2", "1", 33], ["1 2 3", "1", "2", "3"]]:
        with pytest.raises(expected_exception=expected_result):
            is_startswith_and_endswith(string_and_prepostfix=test_strings)
    else:
        assert is_startswith_and_endswith(string_and_prepostfix=test_strings) == expected_result

if __name__ == "__main__":
    try:
        logger.info("Запуск функции 'is_startswith_and_endswith';")
        print(is_startswith_and_endswith(string_and_prepostfix=["Привет, как дела", "Привет", "дела"]))
        logger.info("Функция 'is_startswith_and_endswith' успешно завершена;")
    except Exception:
        logger.error("Ошибка выполнения в функции 'is_startswith_and_endswith';")
```

```python
%%ipytest -v

def zero_prefix_to_len_ten(number: int) -> str:
    logger.debug(f"INPUT: {number=}, {type(number)}")

    if not isinstance(number, int):
        type_error_message: str = "Некорректный тип данных;"
        logger.error(type_error_message)
        raise TypeError(type_error_message)

    if number >= 10 ** 9:
        value_error_message: str = "Некорректное значение;"
        logger.error(value_error_message)
        raise ValueError(value_error_message)

    result: str = str(number).rjust(10, "0")
    logger.debug(f"OUTPUT: {result=}, {type(result)}")

    return result

@pytest.mark.parametrize(
    ["test_number", "expected_result"],
    [
        # Good cases
        (123, "0000000123"),
        (12345, "0000012345"),
        (1234567, "0001234567"),

        # Edge cases
        (0, "0000000000"),

        # Error cases
        (None, TypeError),
        (10 ** 9, ValueError)

    ]
)
def test_zero_prefix_to_len_ten(
        test_number: int,
        expected_result
) -> None:
    if test_number in {None, 10 ** 9}:
        with pytest.raises(expected_exception=expected_result):
            zero_prefix_to_len_ten(number=test_number)
    else:
        assert zero_prefix_to_len_ten(number=test_number) == expected_result

if __name__ == "__main__":
    try:
        logger.info("Запуск функции 'zero_prefix_to_len_ten';")
        print(zero_prefix_to_len_ten(number=123))
        logger.info("Функция 'zero_prefix_to_len_ten' успешно завершена;")

    except Exception:
        logger.error("Ошибка выполнения в функции 'zero_prefix_to_len_ten';")
```
