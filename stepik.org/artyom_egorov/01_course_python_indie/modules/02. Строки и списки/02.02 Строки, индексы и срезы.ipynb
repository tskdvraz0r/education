```python
import pytest
import ipytest
from loguru import logger
```

```python
ipytest.autoconfig()

logger.add(
    sink='../../logs/logs.json',
    format='{time} {level} {message}',
    level='DEBUG',
    rotation='1 Mb',
    retention='1 day',
    compression='zip',
    backtrace=True,
    diagnose=True,
    enqueue=True,
    serialize=True,
    catch=True,
    colorize=True
)
```

# Step 3

```python
%%ipytest

@logger.catch()
def first_element(string: str) -> str:
    if not isinstance(string, str):
        error_msg: str = "Некорректный тип данных;"
        logger.error(error_msg)
        raise TypeError(error_msg)

    return string[0]

@logger.catch()
@pytest.mark.parametrize(
    ["test_string", "expected_result"],
    [
        ("String", "S"),
        ("Test", "T"),
        ("Python", "P")
    ]
)
def test_first_element(
        test_string: str,
        expected_result: str
) -> None:
    assert first_element(string=test_string) == expected_result

if __name__ == "__main__":
    print(
        first_element(string="МФТИ")
    )
```
