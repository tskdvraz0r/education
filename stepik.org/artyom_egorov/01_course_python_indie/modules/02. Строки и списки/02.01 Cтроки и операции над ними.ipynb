```python
import datetime as dt

import pytest
import ipytest
from loguru import logger
```

```python
ipytest.autoconfig()

logger.add(
    sink=rf'..\..\logs\logs_{str(dt.date.today()).replace("-", "_")}.json',
    format='{time} {level} {message}',
    level='DEBUG',
    rotation='10 KB',
    retention='1 day',
    compression='zip',
    backtrace=True,
    diagnose=True,
    enqueue=True,
    serialize=True,
    catch=True,
    colorize=True
)
```

# Step 3

```python
%%ipytest

def i_will_be_great_programmer() -> tuple[str]:
    return ("Я стану крутым программистом!", ) * 3

@pytest.mark.parametrize(
    ["expected_result"],
    [
        (("Я стану крутым программистом!", ) * 3, ), 
    ]
)
def test_i_will_be_great_programmer(expected_result: tuple[str]) -> None:
    assert i_will_be_great_programmer() == expected_result

if __name__ == "__main__":
    print(
        *i_will_be_great_programmer(),
        sep="\n"
    )
```

# Step 6

```python
%%ipytest

def return_same_strings(
        strings: list[str],
        is_stdin: bool = False
) -> list[str]:
    for value, value_type in zip((strings, is_stdin), (list, bool)):
        if not isinstance(value, value_type):
            error_msg: str = "Некорректный тип данных;"
            logger.error(error_msg)
            raise TypeError(error_msg)
    
    if is_stdin:
        strings = [x.strip() for x in strings]
    
    return strings

@pytest.mark.parametrize(
    ["test_strings", "test_is_stdin", "expected_result"],
    [
        (["Hello", "World"], False, ["Hello", "World"]),
        (["Hello", "Hi", "!"], False, ["Hello", "Hi", "!"]),
        (["Привет", "Пока"], False, ["Привет", "Пока"]),
        (["Hello\n", "World"], True, ["Hello", "World"]),
        (["Hello\n", "Hi", "!"], True, ["Hello", "Hi", "!"]),
        (["Привет\n", "Пока"], True, ["Привет", "Пока"])
    ]
)
def test_return_same_strings(
        test_strings: list[str],
        test_is_stdin: bool,
        expected_result: list[str]
) -> None:
    assert return_same_strings(
        strings=test_strings,
        is_stdin=test_is_stdin
    ) == expected_result

if __name__ == "__main__":
    print(
        *return_same_strings(
            strings=["Привет", "Пока"],
            is_stdin=False
        ),
        sep="\n"
    )
```

# Step 7

```python
%%ipytest

def return_reverse_strings(
        strings: list[str],
        is_stdin: bool = False
) -> list[str]:
    for value, value_type in zip((strings, is_stdin), (list, bool)):
        if not isinstance(value, value_type):
            error_msg: str = "Некорректный тип данных;"
            logger.error(error_msg)
            raise TypeError(error_msg)
    
    if is_stdin:
        strings = [x.strip() for x in strings]
    
    return strings[::-1]

@pytest.mark.parametrize(
    ["test_strings", "test_is_stdin", "expected_result"],
    [
        (["Hello", "World"], False, ["World", "Hello"]),
        (["Hello", "Hi", "!"], False, ["!", "Hi", "Hello"]),
        (["Привет", "Пока"], False, ["Пока", "Привет"]),
        (["Hello\n", "World"], True, ["World", "Hello"]),
        (["Hello\n", "Hi", "!"], True, ["!", "Hi", "Hello"]),
        (["Привет\n", "Пока"], True, ["Пока", "Привет"])
    ]
)
def test_return_reverse_strings(
        test_strings: list[str],
        test_is_stdin: bool,
        expected_result: list[str]
) -> None:
    assert return_reverse_strings(
        strings=test_strings,
        is_stdin=test_is_stdin
    ) == expected_result

if __name__ == "__main__":
    print(
        *return_reverse_strings(
            strings=["Привет", "Пока"],
            is_stdin=False
        ),
        sep="\n"
    )
```
