# Step 6

```python
%%ipytest

def return_same_strings(
        strings: list[str],
        is_stdin: bool = False
) -> list[str]:
    for value, value_type in zip((strings, is_stdin), (list, bool)):
        if not isinstance(value, value_type):
            error_msg: str = "Некорректный тип данных;"
            logger.error(error_msg)
            raise TypeError(error_msg)
    
    if is_stdin:
        strings = [x.strip() for x in strings]
    
    return strings

@pytest.mark.parametrize(
    ["test_strings", "test_is_stdin", "expected_result"],
    [
        (["Hello", "World"], False, ["Hello", "World"]),
        (["Hello", "Hi", "!"], False, ["Hello", "Hi", "!"]),
        (["Привет", "Пока"], False, ["Привет", "Пока"]),
        (["Hello\n", "World"], True, ["Hello", "World"]),
        (["Hello\n", "Hi", "!"], True, ["Hello", "Hi", "!"]),
        (["Привет\n", "Пока"], True, ["Привет", "Пока"])
    ]
)
def test_return_same_strings(
        test_strings: list[str],
        test_is_stdin: bool,
        expected_result: list[str]
) -> None:
    assert return_same_strings(
        strings=test_strings,
        is_stdin=test_is_stdin
    ) == expected_result

if __name__ == "__main__":
    print(
        *return_same_strings(
            strings=["Привет", "Пока"],
            is_stdin=False
        ),
        sep="\n"
    )
```

```python
%%ipytest

def return_reverse_strings(
        strings: list[str],
        is_stdin: bool = False
) -> list[str]:
    for value, value_type in zip((strings, is_stdin), (list, bool)):
        if not isinstance(value, value_type):
            error_msg: str = "Некорректный тип данных;"
            logger.error(error_msg)
            raise TypeError(error_msg)
    
    if is_stdin:
        strings = [x.strip() for x in strings]
    
    return strings[::-1]

@pytest.mark.parametrize(
    ["test_strings", "test_is_stdin", "expected_result"],
    [
        (["Hello", "World"], False, ["World", "Hello"]),
        (["Hello", "Hi", "!"], False, ["!", "Hi", "Hello"]),
        (["Привет", "Пока"], False, ["Пока", "Привет"]),
        (["Hello\n", "World"], True, ["World", "Hello"]),
        (["Hello\n", "Hi", "!"], True, ["!", "Hi", "Hello"]),
        (["Привет\n", "Пока"], True, ["Пока", "Привет"])
    ]
)
def test_return_reverse_strings(
        test_strings: list[str],
        test_is_stdin: bool,
        expected_result: list[str]
) -> None:
    assert return_reverse_strings(
        strings=test_strings,
        is_stdin=test_is_stdin
    ) == expected_result

if __name__ == "__main__":
    print(
        *return_reverse_strings(
            strings=["Привет", "Пока"],
            is_stdin=False
        ),
        sep="\n"
    )
```

# Step 8

```python
%%ipytest

@logger.catch()
def repeat_four_times(string: str) -> str:
    if not isinstance(string, str):
        error_msg: str = "Некорректный тип данных;"
        logger.error(error_msg)
        raise TypeError(error_msg)
    
    return " ".join([string] * 4)

@logger.catch()
@pytest.mark.parametrize(
    ["test_string", "expected_result"],
    [
        ("Hello", "Hello Hello Hello Hello"),
        ("World", "World World World World"),
        ("Привет", "Привет Привет Привет Привет"),
        ("Пока", "Пока Пока Пока Пока")
    ]
)
def test_repeat_four_times(
        test_string: str,
        expected_result: str
) -> None:
    assert repeat_four_times(test_string) == expected_result

if __name__ == "__main__":
    try:
        logger.info("Вызов функции 'repeat_four_times';")
        print(repeat_four_times("Поехали!"))
        logger.info("Успешное завершение функции 'repeat_four_times';")
    
    except Exception as err:
        logger.error("Ошибка выполнения функции 'repeat_four_times';")
```

```python
%%ipytest

@logger.catch()
def reverse_concat(
        strings: list[str],
        is_stdin: bool = False
) -> str:
    if not isinstance(strings, list):
        error_msg: str = "Некорректный тип данных;"
        logger.error(error_msg)
        raise TypeError(error_msg)
    
    if is_stdin:
        strings = [x.strip() for x in strings]
    
    return "".join(strings[::-1])

@logger.catch()
@pytest.mark.parametrize(
    ["test_strings", "test_is_stdin", "expected_result"],
    [
        (["Hello", "World"], False, "WorldHello"),
        (["Hello", "Hi", "!"], False, "!HiHello"),
        (["Привет", "Пока"], False, "ПокаПривет"),
        (["Hello\n", "World"], True, "WorldHello"),
        (["Hello\n", "Hi", "!"], True, "!HiHello"),
        (["Привет\n", "Пока"], True, "ПокаПривет"),
    ]
)
def test_reverse_concat(
        test_strings: list[str],
        test_is_stdin: bool,
        expected_result: str
) -> None:
    assert reverse_concat(
        strings=test_strings,
        is_stdin=test_is_stdin
    ) == expected_result

if __name__ == "__main__":
    try:
        logger.info("Вызов функции 'reverse_concat';")
        print(
            reverse_concat(
                strings=["Привет", "Пока"],
                is_stdin=False
            )
        )
        logger.info("Успешное завершение функции 'reverse_concat';")
    
    except Exception as err:
        logger.error("Ошибка выполнения функции 'reverse_concat';")
```

```python
%%ipytest

@logger.catch()
def symbol_code(symbols: str) -> tuple[str]:
    if not isinstance(symbols, str):
        error_msg: str = "Некорректный тип данных;"
        logger.error(error_msg)
        raise TypeError(error_msg)
    
    return tuple(f"Simvol code {x} is {ord(x)}." for x in symbols.split())

@logger.catch()
@pytest.mark.parametrize(
    ["test_symbols", "expected_result"],
    [
        ("H e l", ("Simvol code H is 72.", "Simvol code e is 101.", "Simvol code l is 108.")),
        ("W r d", ("Simvol code W is 87.", "Simvol code r is 114.", "Simvol code d is 100.")),
    ]
)
def test_symbol_code(
        test_symbols: str,
        expected_result: tuple[str]
) -> None:
    assert symbol_code(symbols=test_symbols) == expected_result

if __name__ == "__main__":
    try:
        logger.info("Вызов функции 'symbol_code';")
        print(
            *symbol_code(symbols="a b c"),
            sep="\n"
        )
        logger.info("Успешное завершение функции 'symbol_code';")
    
    except Exception as err:
        logger.error("Ошибка выполнения функции 'symbol_code';")
```
