# Step 5

```python
%%ipytest -v

def is_strings_equal(
        strings: list[str],
        is_stdin: bool = False,
) -> bool:
    logger.debug(f"{strings=}: {type(strings)}")
    logger.debug(f"{is_stdin=}: {type(is_stdin)}")

    # Проверка типов переданных аргументов;
    for value, value_type in zip(
            (strings, is_stdin),
            (list, bool)
    ):
        if not isinstance(value, value_type):
            error_message: str = "Некорректный тип данных;"
            logger.error(error_message)
            raise TypeError(error_message)

    # Проверка количества элементов в списке;
    if len(strings) !=  2:
        error_message: str = "Некорректное количество элементов;"
        logger.error(error_message)
        raise ValueError(error_message)

    # Проверка типов элементов списка;
    if not all(isinstance(x, str) for x in strings):
        error_message: str = "Некорректный тип данных элемента;"
        logger.error(error_message)
        raise TypeError(error_message)

    return strings[0].strip().lower() == strings[1].strip().lower()

@pytest.mark.parametrize(
    ["test_strings", "test_is_stdin", "expected_result"],
    [
        (["test_string", "test_string"], False, True),
        (["pswrd1", "PSWd1"], False, False),
        (["test_string\n", "test_string"], True, True),
        (["pswrd1\n", "PSWd1"], True, False),
    ]
)
def test_is_strings_equal(
        test_strings: list[str],
        test_is_stdin: bool,
        expected_result: bool,
) -> None:
    assert is_strings_equal(
        strings=test_strings,
        is_stdin=test_is_stdin,
    ) == expected_result

if __name__ == "__main__":
    try:
        logger.info("Запуск функции 'is_strings_equal';")
        print(
            is_strings_equal(
                strings=["password", "passWord"],
                is_stdin=False
            )
        )
        logger.info("Функция 'is_strings_equal' успешно завершена;")
    except Exception as e:
        error_message: str = "Ошибка выполнения в функции 'is_strings_equal';"
        logger.error(error_message)
```
