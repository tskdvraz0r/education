```python
import pytest
import ipytest

ipytest.autoconfig()
```

# Step 4

```python
%%ipytest

def older_by_year(number: int) -> int:
    return number + 1

@pytest.mark.parametrize(
    ["test_number", "expected_result"],
    [
        (1, 2),
        (13, 14),
        (23, 24)
    ]
)
def test_older_by_year(
    test_number,
    expected_result
) -> None:
    assert older_by_year(number=test_number) == expected_result

if __name__ == "__main__":
    print(
        older_by_year(
            number=33
        )
    )
```

# Step 5

```python
%%ipytest

def bigger_and_lower(number: int) -> tuple[int]:
    return (
        number * 2,
        number / 2
    )

@pytest.mark.parametrize(
    ["test_number", "expected_result"],
    [
        (8, (16, 4.0)),
        (5, (10, 2.5)),
        (100, (200, 50.0)),
    ]
)
def test_bigger_and_lower(
    test_number,
    expected_result
) -> None:
    assert bigger_and_lower(number=test_number) == expected_result

if __name__ == "__main__":
    print(
        *bigger_and_lower(number=13),
        sep="\n"
    )
```

# Step 7

```python
%%ipytest

def sum_of_nums(
        numbers: list[int | str],
        is_stdin: bool = False
) -> int:
    if is_stdin:
        numbers = [int(i) for i in numbers]
    
    return sum(numbers)

@pytest.mark.parametrize(
    ["test_numbers", "test_is_stdin", "expected_result"],
    [
        ([1, 2], False, 3),
        ([13, 3], False, 16),
        ([10, -5], False, 5),
        (["1\n", "2"], True, 3),
        (["13\n", "3"], True, 16),
        (["10\n", "-5"], True, 5),
    ]
)
def test_sum_of_nums(
        test_numbers,
        test_is_stdin,
        expected_result
) -> None:
    assert sum_of_nums(
        numbers=test_numbers,
        is_stdin=test_is_stdin
    ) == expected_result

if __name__ == "__main__":
    print(
        sum_of_nums(
            numbers=[13, 13],
            is_stdin=False
        )
    )
```

# Step 8

```python
%%ipytest

def rectangle_area_and_perimeter(
        length: list[str | int | float],
        is_stdin: bool = False
) -> tuple[float]:
    if is_stdin:
        length = [int(i) if "." not in i else float(i) for i in length]
    
    a: int | float = length[0]
    b: int | float = length[1]
    
    return (
        float(a * b),
        float(2 * (a + b))
    )

@pytest.mark.parametrize(
    ["test_length", "test_is_stdin", "expected_result"],
    [
        ([7, 2], False, (14.0, 18.0)),
        ([1, 0.5], False, (0.5, 3.0)),
        ([5, 0.1], False, (0.5, 10.2)),
        (["7\n", "2"], True, (14.0, 18.0)),
        (["1\n", "0.5"], True, (0.5, 3.0)),
        (["5\n", "0.1"], True, (0.5, 10.2)),
    ]
)
def test_rectangle_area_and_perimeter(
        test_length,
        test_is_stdin,
        expected_result
) -> None:
    assert rectangle_area_and_perimeter(
        length=test_length,
        is_stdin=test_is_stdin
    ) == expected_result

if __name__ == "__main__":
    print(
        *rectangle_area_and_perimeter(
            length=[3, 13],
            is_stdin=False
        )
    )
```

```python
%%ipytest

def fahrenheit_to_celsius(f: int | float) -> float:
    return ((f - 32) * 5) / 9

@pytest.mark.parametrize(
    ["test_f", "expected_result"],
    [
        (68, 20.0),
        (32.0, 0.0),
        (86, 30.0)
    ]
)
def test_fahrenheit_to_celsius(
    test_f,
    expected_result
) -> None:
    assert fahrenheit_to_celsius(f=test_f) == expected_result

if __name__ == "__main__":
    print(
        fahrenheit_to_celsius(f=13)
    )
```

# Step 11

```python
%%ipytest

def absulute_sum(
        numbers: list[int | str],
        is_stdin: False = bool
) -> int:
    if is_stdin:
        numbers = [int(i) for i in numbers]
    
    return sum(map(abs, numbers))

@pytest.mark.parametrize(
    ["test_numbers", "test_is_stdin", "expected_result"],
    [
        ([1, -2], False, 3),
        ([1, -2, 3, -4], False, 10),
        ([1, -2, 3, -4, 5, -6], False, 21),
        (["1\n", "-2"], True, 3),
        (["1\n", "-2\n", "3\n", "-4"], True, 10),
        (["1\n", "-2\n", "3\n", "-4\n", "5\n", "-6"], True, 21),
    ]
)
def test_absulute_sum(
        test_numbers,
        test_is_stdin,
        expected_result
) -> None:
    assert absulute_sum(
        numbers=test_numbers,
        is_stdin=test_is_stdin
    ) == expected_result

if __name__ == "__main__":
    print(
        absulute_sum(
            numbers=[-13, 3, -7],
            is_stdin=False
        )
    )
```

```python
%%ipytest

def distance_between_points(
        points: list[str | float | int],
        is_stdin: bool = False
) -> float:
    if is_stdin:
        points = [float(i) for i in points]
    
    return max(points) - min(points)

@pytest.mark.parametrize(
    ["test_points", "test_is_stdin", "expected_result"],
    [
        ([-2, 6.5], False, 8.5),
        ([3, -9], False, 12.0),
        ([10.5, 8], False, 2.5),
        (["-2\n", "6.5"], True, 8.5),
        (["3\n", "-9"], True, 12.0),
        (["10.5\n", "8"], True, 2.5),
    ]
)
def test_distance_between_points(
        test_points,
        test_is_stdin,
        expected_result
) -> None:
    assert distance_between_points(
        points=test_points,
        is_stdin=test_is_stdin
    ) == expected_result

if __name__ == "__main__":
    print(
        distance_between_points(
            points=[13.3, -13.3],
            is_stdin=False
        )
    )
```

# Step 14

```python
%%ipytest

def seconds_between_times(
        times: list[int | str],
        is_stdin: bool = False
) -> int:
    import datetime as dt
    
    if is_stdin:
        times = [int(i) for i in times]
    
    first_time: dt.datetime = dt.datetime(
        year=2024,
        month=6,
        day=28,
        hour=times[0],
        minute=times[1],
        second=times[2]
    )
    
    second_time: dt.datetime = dt.datetime(
        year=2024,
        month=6,
        day=28,
        hour=times[3],
        minute=times[4],
        second=times[5]
    )
    
    return (second_time - first_time).seconds

@pytest.mark.parametrize(
    ["test_times", "test_is_stdin", "expected_result"],
    [
        ([1, 1, 1, 2, 2, 2], False, 3661),
        ([1, 2, 30, 1, 3, 20], False, 50),
        ([1, 2, 3, 4, 5, 6], False, 10983),
        (["1\n", "1\n", "1\n", "2\n", "2\n", "2"], True, 3661),
        (["1\n", "2\n", "30\n", "1\n", "3\n", "20"], True, 50),
        (["1\n", "2\n", "3\n", "4\n", "5\n", "6"], True, 10983),
    ]
)
def test_seconds_between_times(
    test_times,
    test_is_stdin,
    expected_result
) -> None:
    assert seconds_between_times(
        times=test_times,
        is_stdin=test_is_stdin
    ) == expected_result
    
if __name__ == "__main__":
    print(
        seconds_between_times(
            times=[2, 34, 43, 3, 15, 54],
            is_stdin=False
        )
    )
```

```python
%%ipytest

def expression(
        numbers: list[str | int],
        is_stdin: bool = False
) -> int:
    if is_stdin:
        numbers = [int(i) for i in numbers]
    
    a, b, c = numbers
    
    return max(
        a + b + c,
        a + b * c,
        a * b + c,
        a * b * c,
        (a + b) + c,
        (a + b) * c,
        (a * b) + c,
        a + (b * c),
        a * (b + c)
    )

@pytest.mark.parametrize(
    ["test_numbers", "test_is_stdin", "expected_result"],
    [
        ([1, 2, 3], False, 9),
        ([2, 10, 3], False, 60),
        ([1, 1, 1], False, 3),
        (["1\n", "2\n", "3"], True, 9),
        (["2\n", "10\n", "3"], True, 60),
        (["1\n", "1\n", "1"], True, 3),
    ]
)
def test_expression(
    test_numbers,
    test_is_stdin,
    expected_result
) -> None:
    assert expression(
        numbers=test_numbers,
        is_stdin=test_is_stdin
    ) == expected_result

if __name__ == "__main__":
    print(
        expression(
            numbers=[2, 10, 3],
            is_stdin=False
        )
    )
```
