```python
import pytest
import ipytest

ipytest.autoconfig()
```

# Step 2

```python
%%ipytest

def total_kilos(grams: int) -> int:
    return grams // 1000

@pytest.mark.parametrize(
    ["test_grams", "expected_result"],
    [
        (1000, 1),
        (2345, 2),
        (3456, 3),
    ]
)
def test_total_kilos(
        test_grams,
        expected_result
) -> None:
    assert total_kilos(test_grams) == expected_result

if __name__ == "__main__":
    print(
        total_kilos(grams=1313)
    )
```

# Step 3

```python
%%ipytest

def apple_dividing_1(
        schoolboys_and_apples: list[int | str],
        is_stdin: bool = False
) -> int:
    if is_stdin:
        schoolboys_and_apples = [int(x) for x in schoolboys_and_apples]

    schoolboys, apples = schoolboys_and_apples

    return  apples // schoolboys

@pytest.mark.parametrize(
    ["test_schoolboys_and_apples", "test_is_stdin", "expected_result"],
    [
        ([5, 11], False, 2),
        ([6, 17], False, 2),
        ([3, 42], False, 14),
        (["5\n", "11"], True, 2),
        (["6\n", "17"], True, 2),
        (["3\n", "42"], True, 14)
    ]
)
def test_apple_dividing_1(
        test_schoolboys_and_apples,
        test_is_stdin,
        expected_result
) -> None:
    assert apple_dividing_1(
        schoolboys_and_apples=test_schoolboys_and_apples,
        is_stdin=test_is_stdin
    ) == expected_result

if __name__ == "__main__":
    print(
        apple_dividing_1(
            schoolboys_and_apples=[2, 9],
            is_stdin=False
        )
    )
```

# Step 4

```python
%%ipytest

def pairs_of_shoes(
        depo_and_costs: list[int | str],
        is_stdin: bool  = False
) -> int:
    if is_stdin:
        depo_and_costs = [int(x) for x in depo_and_costs]

    depo, costs = depo_and_costs

    return  depo // costs

@pytest.mark.parametrize(
    ["test_depo_and_costs", "test_is_stdin", "expected_result"],
    [
        ([150, 50], False, 3),
        ([120, 41], False, 2),
        ([200, 33], False, 6),
        (["150\n", "50"], True, 3),
        (["120\n", "41"], True, 2),
        (["200\n", "33"], True, 6)
    ]
)
def test_pairs_of_shoes(
        test_depo_and_costs,
        test_is_stdin,
        expected_result
) -> None:
    assert pairs_of_shoes(
        depo_and_costs=test_depo_and_costs,
        is_stdin=test_is_stdin
    ) == expected_result

if __name__ == "__main__":
    print(
        pairs_of_shoes(
            depo_and_costs=[333, 18],
            is_stdin=False
        )
    )
```

```python
%%ipytest

def last_digit(number: int) -> int:
    return number % 10

@pytest.mark.parametrize(
    ["test_number", "expected_result"],
    [
        (123, 3),
        (234,  4),
        (345, 5),
    ]
)
def test_last_digit(
        test_number,
        expected_result
)  -> None:
    assert last_digit(number=test_number) == expected_result

if __name__ == "__main__":
    print(
        last_digit(number=123)
    )
```

```python
%%ipytest

def category_of_hundreds(number: int) -> int:
    return 0 if number < 100 else number % 1000 // 100

@pytest.mark.parametrize(
    ["test_number", "expected_result"],
    [
        (37, 0),
        (123, 1),
        (234, 2),
        (345, 3),
        (1234, 2),
        (2345, 3),
        (34567, 5),
    ]
)
def test_category_of_hundreds(
        test_number,
        expected_result
) -> None:
    assert category_of_hundreds(number=test_number) == expected_result

if  __name__  ==  "__main__":
    print(
        category_of_hundreds(number=12345)
    )
```

```python
%%ipytest

def sum_of_digits(number: int) -> int:
    return sum(int(i) for i in str(number))

@pytest.mark.parametrize(
    ["test_number", "expected_result"],
    [
        (123, 6),
        (234, 9),
        (345, 12),
        (1234, 10),
        (2345, 14),
        (34567, 25),
    ]
)
def test_sum_of_digits(
        test_number,
        expected_result
) -> None:
    assert sum_of_digits(number=test_number) == expected_result

if __name__ == "__main__":
    print(
        sum_of_digits(number=34567)
    )
```

# Step 8

```python
%%ipytest

def win_the_lottery(deposit: int) -> int:
    nominals: tuple[int] = (100, 20, 10, 5, 1)
    result: int = 0

    for nominal in nominals:
        result += deposit // nominal
        deposit = deposit % nominal

    return result

@pytest.mark.parametrize(
    ["test_deposit", "expected_result"],
    [
        (125, 3),
        (126, 4),
        (127, 5),
        (128, 6),
        (129, 7),
    ]
)
def test_win_the_lottery(
        test_deposit,
        expected_result
) -> None:
    assert win_the_lottery(deposit=test_deposit) == expected_result

if __name__ == "__main__":
    print(
        win_the_lottery(deposit=12345)
    )
```

# Step 9

```python
%%ipytest

def digit_clock(number: int) -> str:
    hour, minute = divmod(number, 60)

    return hour if hour < 24 else hour % 24, minute

@pytest.mark.parametrize(
    ["test_number", "expected_result"],
    [
        (137, (2, 17)),
        (2879, (23, 59)),
        (1511, (1, 11))
    ]
)
def test_digit_clock(test_number, expected_result)  -> None:
    assert digit_clock(number=test_number) == expected_result

if __name__ == "__main__":
    print(
        *digit_clock(number=1511)
    )
```

# Step 10

```python
%%ipytest

def next_even_number(number: int)  -> int:
    return (
        number + (number % 2)
        if number % 2 != 0
        else number + 2
    )

@pytest.mark.parametrize(
    ["test_number", "expected_result"],
    [
        (1, 2),
        (2, 4),
        (3, 4),
        (4, 6),
        (5, 6),
    ]
)
def test_next_even_number(test_number, expected_result) -> None:
    assert next_even_number(number=test_number) == expected_result

if __name__ == "__main__":
    print(
        next_even_number(number=15)
    )
```

# Step 11

```python
%%ipytest

def digit_clock_2(seconds: int) -> str:
    hour: int = seconds // 3600 if seconds // 3600 < 24 else seconds // 3600 % 24
    minute: int = seconds % 3600 // 60
    sec: int = seconds % 3600 % 60

    return f"{hour}:{str(minute).rjust(2, '0')}:{str(sec).rjust(2, '0')}"

@pytest.mark.parametrize(
    ["test_seconds", "expected_result"],
    [
        (3721, "1:02:01"),
        (5000, "1:23:20"),
        (55408, "15:23:28")
    ]
)
def test_digit_clock_2(test_seconds, expected_result)  -> None:
    assert digit_clock_2(seconds=test_seconds) == expected_result

if __name__ == "__main__":
    print(
        digit_clock_2(seconds=5000)
    )
```
