```python
import pytest
import ipytest

ipytest.autoconfig()
```

# Step 4

```python
%%ipytest

def hello_world() -> str:
    return "Hello, world!"

@pytest.mark.parametrize(
    ["expected_result"],
    [
        ("Hello, world!", )
    ]
)
def test_hello_world(expected_result) -> None:
    assert hello_world() == expected_result
    
if __name__ == "__main__":
    print(
        hello_world()
    )
```
