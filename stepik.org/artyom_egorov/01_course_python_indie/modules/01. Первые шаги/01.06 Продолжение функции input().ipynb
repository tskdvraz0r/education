```python
import pytest
import ipytest

ipytest.autoconfig()
```

# Step 3

```python
%%ipytest

def mean_of_nums(numbers: str) -> float:
    numbers = list(map(int, numbers.split()))
    
    return sum(numbers) / len(numbers)

@pytest.mark.parametrize(
    ["test_numbers", "expected_result"],
    [
        ("2 3 4 5", 3.5),
        ("3 3 4 4", 3.5),
        ("5 3 4 5", 4.25),
    ]
)
def test_mean_of_nums(
        test_numbers,
        expected_result
) -> None:
    assert mean_of_nums(numbers=test_numbers) == expected_result

if __name__ == "__main__":
    print(mean_of_nums(numbers="1 2 3 4 5"))
```

# Step 4

```python
%%ipytest

def greatest_estimate(estimates: str) -> int:
    estimates = list(map(int, estimates.split()))
    
    return max(estimates)

@pytest.mark.parametrize(
    ["test_numbers", "expected_result"],
    [
        ("45 80 98 47 81", 98),
        ("1 86 63 98 20", 98),
        ("36 85 92 99 22", 99),
    ]
)
def test_greatest_estimate(
        test_numbers,
        expected_result
) -> None:
    assert greatest_estimate(estimates=test_numbers) == expected_result

if __name__ == "__main__":
    print(greatest_estimate(estimates="1 2 3 4 5"))
```

# Step 5

```python
%%ipytest

def panel_sulfidation(numbers: str) -> int:
    n, a, b = map(int, numbers.split())
    
    return a * b * 2 * n

@pytest.mark.parametrize(
    ["test_numbers", "expected_result"],
    [
        ("5 2 3", 60),
        ("14 23 5", 3220),
        ("13 6 66", 10296)
    ]
)
def test_panel_sulfidation(
        test_numbers,
        expected_result
) -> None:
    assert panel_sulfidation(numbers=test_numbers) == expected_result

if __name__ == "__main__":
    print(
        panel_sulfidation(numbers="3 7 13")
    )
```

# Step 6

```python
%%ipytest

def paper_cranes(number: int) -> tuple[int]:
    return (
        int((number / 3) / 2),
        int((number / 3) * 2),
        int((number / 3) / 2)
    )

@pytest.mark.parametrize(
    ["test_number", "expected_result"],
    [
        (6, (1, 4, 1)),
        (24, (4, 16, 4)),
        (60, (10, 40, 10)),
    ]
)
def test_paper_cranes(
    test_number,
    expected_result
) -> None:
    assert paper_cranes(number=test_number) == expected_result

if __name__ == "__main__":
    print(
        *paper_cranes(number=666)
    )
```

```python
%%ipytest

def stationery_store(stationery: str) -> int:
    cost_pencil: int = 3
    cost_pen: int = cost_pencil + 2
    cost_marker: int = cost_pen + 7
    
    cnt_pencils, cnt_pens, cnt_markers = list(map(int, stationery.split()))
    
    return cnt_pencils * cost_pencil + cnt_pens * cost_pen + cnt_markers * cost_marker

@pytest.mark.parametrize(
    ["test_stationery", "expected_result"],
    [
        ("1 1 1", 20),
        ("23 8 76", 1021),
        ("75 5 72", 1114),
    ]
)
def test_stationery_store(
        test_stationery,
        expected_result
) -> None:
    assert stationery_store(stationery=test_stationery) == expected_result

if __name__ == "__main__":
    print(
        stationery_store(
            stationery="7 13 37"
        )
    )
```

```python
%%ipytest

def two_bandits(shoots: str) -> str:
    x, y = list(map(int, shoots.split()))
    
    return f"{(x + y - 1) - x} {(x + y - 1) - y}"

@pytest.mark.parametrize(
    ["test_shoots", "expected_result"],
    [
        ("4 7", "6 3"),
        ("5 3", "2 4"),
        ("2 7", "6 1"),
    ]
)
def test_two_bandits(
        test_shoots,
        expected_result
) -> None:
    assert two_bandits(shoots=test_shoots) == expected_result

if __name__ == "__main__":
    print(
        two_bandits(shoots="13 7")
    )
```
