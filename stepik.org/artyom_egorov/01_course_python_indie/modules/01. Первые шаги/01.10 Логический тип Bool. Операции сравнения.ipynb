```python
import pytest
import ipytest

ipytest.autoconfig()
```

# Step 2

```python
%%ipytest

def is_positive(number: int) -> bool:
    return number > 0

@pytest.mark.parametrize(
    ["test_number", "expected_result"],
    [
        (1, True),
        (0, False),
        (-1, False)
    ]
)
def test_is_positive(
        test_number: int,
        expected_result: bool,
) -> None:
    assert is_positive(number=test_number) == expected_result

if __name__ == "__main__":
    print(
        is_positive(number=13)
    )
```

```python
%%ipytest

def is_even(number: int) -> bool:
    return number % 2 == 0

@pytest.mark.parametrize(
    ["test_number", "expected_result"],
    [
        (2, True),
        (1, False),
        (0, True)
    ]
)
def test_is_even(
        test_number: int,
        expected_result: bool
) -> None:
    assert is_even(number=test_number) == expected_result

if __name__ == "__main__":
    print(
        is_even(number=13)
    )
```

# Step 8

```python
%%ipytest

def is_both_divided_by_seven(numbers: str) -> bool:
    a, b = list(map(int, numbers.split()))
    return a % 7 == 0 and b % 7 == 0

@pytest.mark.parametrize(
    ["test_numbers", "expected_result"],
    [
        ("77 14", True),
        ("35 10", False),
        ("0 7", True)
    ]
)
def test_is_both_divided_by_seven(
        test_numbers: int,
        expected_result: bool
) -> None:
    assert is_both_divided_by_seven(numbers=test_numbers) == expected_result

if __name__ == "__main__":
    print(
        is_both_divided_by_seven(numbers="7 14")
    )
```

# Step 9

```python
%%ipytest

def is_right_triangle(numbers: str) -> bool:
    a, b, c = list(map(int, numbers.split()))

    return a == b == c

@pytest.mark.parametrize(
    ["test_numbers", "expected_result"],
    [
        ("1 2 3", False),
        ("5 5 5", True),
        ("1 1 1", True)
    ]
)
def test_is_right_triangle(
        test_numbers: int,
        expected_result: bool
) -> None:
    assert is_right_triangle(numbers=test_numbers) == expected_result

if __name__ == "__main__":
    print(
        is_right_triangle(numbers="1 2 3")
    )
```

```python
%%ipytest

def is_number_in_range(number: int) -> bool:
    return 6 <= number <= 19

@pytest.mark.parametrize(
    ["test_number", "expected_result"],
    [
        (6, True),
        (20, False),
        (0, False)
    ]
)
def test_is_number_in_range(
        test_number: int,
        expected_result: bool
) -> None:
    assert is_number_in_range(number=test_number) == expected_result

if __name__ == "__main__":
    print(
        is_number_in_range(number=13)
    )
```

```python
%%ipytest

def is_any_word_awesome(
    words: str | list[str],
    is_stdin: bool = False
) -> bool:
    if is_stdin:
        words = [x.strip() for x in words]
        return "awesome" in words
    else:
        return "awesome" in words.split()

@pytest.mark.parametrize(
    ["test_words", "test_is_stdin", "expected_result"],
    [
        ("cool awesome", False, True),
        ("test sample", False, False),
        ("pythonawesome awesomepython", False, False),

        (["cool\n", "awesome"], True, True),
        (["test\n", "sample"], True, False),
        (["pythonawesome\n", "awesomepython"], True, False)
    ]
)
def test_is_any_word_awesome(
        test_words: str,
        test_is_stdin: bool,
        expected_result: bool
) -> None:
    assert is_any_word_awesome(
        words=test_words,
        is_stdin=test_is_stdin
) == expected_result

if __name__ == "__main__":
    print(
        is_any_word_awesome(
            words="python is awesome",
            is_stdin = False
        )
    )
```
