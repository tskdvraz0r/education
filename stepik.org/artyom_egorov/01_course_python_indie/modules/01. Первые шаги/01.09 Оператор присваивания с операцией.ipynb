```python
import pytest
import ipytest

ipytest.autoconfig()
```

# Step 5

```python
%%ipytest

def increment(number: int) -> int:
    return number + 1

@pytest.mark.parametrize(
    ["test_number", "expected_result"],
    [
        (1, 2),
        (13, 14),
        (999, 1000)
    ]
)
def test_increment(
        test_number: int,
        expected_result: int
) -> None:
    assert increment(number=test_number) == expected_result

if __name__ == "__main__":
    print(
        increment(number=10)
    )
```

# Step 10

```python
%%ipytest

def increase_deposit(deposit: int | float) -> float:
    return deposit * 1.5

@pytest.mark.parametrize(
    ["test_deposit", "expected_result"],
    [
        (100, 150.0),
        (6.3, 9.45),
        (11, 16.5)
    ]
)
def test_increment_deposit(
        test_deposit,
        expected_result
) -> None:
    assert increase_deposit(deposit=test_deposit) == expected_result

if __name__ == "__main__":
    print(
        increase_deposit(deposit=100)
    )
```
