```python
# Date: 28.01.2024
# Step: 4.3.9
# Url: https://stepik.org/lesson/794582/step/9?unit=797335

class Scales:
    def __init__(self):
        self.right_scale = 0
        self.left_scale = 0

    def add_right(self, weight: int):
        self.right_scale += weight

    def add_left(self, weight: int):
        self.left_scale += weight

    def get_result(self):
        if self.left_scale == self.right_scale:
            return "Весы в равновесии"
        elif self.left_scale > self.right_scale:
            return "Левая чаша тяжелее"
        else:
            return "Правая чаша тяжелее"
```

```python
# Date: 28.01.2024
# Step: 4.3.11
# Url: https://stepik.org/lesson/794582/step/11?unit=797335

class Numbers:
    def __init__(self):
        self.numbers = []
    
    def add_number(self, n: int):
        self.numbers.append(n)
    
    def get_even(self):
        return [i for i in self.numbers if i % 2 == 0]
    
    def get_odd(self):
        return [i for i in self.numbers if i % 2 != 0]
```

```python
# Date: 28.01.2024
# Step: 4.3.12
# Url: https://stepik.org/lesson/794582/step/12?unit=797335

class TextHandler:
    def __init__(self):
        self.words = []
    
    def add_words(self, text: str):
        self.words.extend(text.split())
    
    def get_shortest_words(self):
        if len(self.words) != 0:
            min_word_len: int = len(min(self.words, key = len))
            return [i for i in self.words if len(i) == min_word_len]
        else:
            return self.words
    
    def get_longest_words(self):
        if len(self.words) != 0:
            max_word_len: int = len(max(self.words, key = len))
            return [i for i in self.words if len(i) == max_word_len]
        else:
            return self.words
```

```python
# Date: 28.01.2024
# Step: 4.3.12
# Url: https://stepik.org/lesson/794582/step/12?unit=797335

class Todo:
    def __init__(self):
        self.things = []
    
    def add(self, task: str, priority: int) -> None:
        self.things.append((task, priority))
    
    def get_by_priority(self, n: int):
        if len(self.things) == 0:
            return self.things
        return [i[0] for i in self.things if i[1] == n]
    
    def get_low_priority(self):
        if len(self.things) == 0:
            return self.things
        priority: int = min(self.things, key = lambda x: x[1])[1]
        return self.get_by_priority(priority)
    
    def get_high_priority(self):
        if len(self.things) == 0:
            return self.things
        priority: int = max(self.things, key = lambda x: x[1])[1]
        return self.get_by_priority(priority)
```

```python
# Date: 28.01.2024
# Step: 4.3.14
# Url: https://stepik.org/lesson/794582/step/14?unit=797335

class Postman:
    def __init__(self):
        self.delivery_data = []
    
    def add_delivery(self, street: str, house: int, flat: int) -> None:
        self.delivery_data.append((street, house, flat))
    
    def get_houses_for_street(self, street: str) -> list:
        houses: list = []
        for house in [i for i in self.delivery_data if i[0] == street]:
            if house[1] not in houses:
                houses.append(house[1])
        
        return houses
    
    def get_flats_for_house(self, street: str, house: int) -> list:
        flats: list = []
        for flat in [i for i in self.delivery_data if i[0] == street and i[1] == house]:
            if flat[2] not in flats:
                flats.append(flat[2])
        
        return flats
```
