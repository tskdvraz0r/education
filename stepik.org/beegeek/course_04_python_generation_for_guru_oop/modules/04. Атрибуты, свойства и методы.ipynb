```python
# Date: 28.01.2024
# Step: 4.3.12
# Url: https://stepik.org/lesson/794582/step/12?unit=797335

class TextHandler:
    def __init__(self):
        self.words = []
    
    def add_words(self, text: str):
        self.words.extend(text.split())
    
    def get_shortest_words(self):
        if len(self.words) != 0:
            min_word_len: int = len(min(self.words, key = len))
            return [i for i in self.words if len(i) == min_word_len]
        else:
            return self.words
    
    def get_longest_words(self):
        if len(self.words) != 0:
            max_word_len: int = len(max(self.words, key = len))
            return [i for i in self.words if len(i) == max_word_len]
        else:
            return self.words
```

```python
# Date: 28.01.2024
# Step: 4.3.12
# Url: https://stepik.org/lesson/794582/step/12?unit=797335

class Todo:
    def __init__(self):
        self.things = []
    
    def add(self, task: str, priority: int) -> None:
        self.things.append((task, priority))
    
    def get_by_priority(self, n: int):
        if len(self.things) == 0:
            return self.things
        return [i[0] for i in self.things if i[1] == n]
    
    def get_low_priority(self):
        if len(self.things) == 0:
            return self.things
        priority: int = min(self.things, key = lambda x: x[1])[1]
        return self.get_by_priority(priority)
    
    def get_high_priority(self):
        if len(self.things) == 0:
            return self.things
        priority: int = max(self.things, key = lambda x: x[1])[1]
        return self.get_by_priority(priority)
```

```python
# Date: 28.01.2024
# Step: 4.3.14
# Url: https://stepik.org/lesson/794582/step/14?unit=797335

class Postman:
    def __init__(self):
        self.delivery_data = []
    
    def add_delivery(self, street: str, house: int, flat: int) -> None:
        self.delivery_data.append((street, house, flat))
    
    def get_houses_for_street(self, street: str) -> list:
        houses: list = []
        for house in [i for i in self.delivery_data if i[0] == street]:
            if house[1] not in houses:
                houses.append(house[1])
        
        return houses
    
    def get_flats_for_house(self, street: str, house: int) -> list:
        flats: list = []
        for flat in [i for i in self.delivery_data if i[0] == street and i[1] == house]:
            if flat[2] not in flats:
                flats.append(flat[2])
        
        return flats
```

# 4.4 Модификаторы доступа и аксессоры

```python
# Date: 31.01.2024
# Step: 4.4.15
# Url: https://stepik.org/lesson/793864/step/15?unit=796596

class Circle:
    def __init__(self, radius: float):
        self._radius = radius
        self._area = math.pi * radius ** 2
        self._diameter = radius * 2
    
    def get_radius(self) -> float:
        return self._radius
    
    def get_area(self) -> float:
        return self._area
    
    def get_diameter(self) -> float:
        return self._diameter
```

```python
# Date: 31.01.2024
# Step: 4.4.16
# Url: https://stepik.org/lesson/793864/step/16?unit=796596

class BankAccount:
    def __init__(self, balance: int = 0):
        self._balance = balance
    
    def get_balance(self) -> int:
        return self._balance
    
    def deposit(self, amount: int) -> None:
        self._balance += amount
    
    def withdraw(self, amount: int) -> None:
        if amount > self._balance:
            raise ValueError("На счете недостаточно средств")
        
        self._balance -= amount
    
    def transfer(self, account, amount: int) -> None:
        self.withdraw(amount = amount)
        account.deposit(amount = amount)
```

```python
# Date: 31.01.2024
# Step: 4.4.17
# Url: https://stepik.org/lesson/793864/step/17?unit=796596

class User:
    def __init__(self, name: str, age: int) -> None:
        if isinstance(name, str) and name.isalpha():
            self._name = name
        else:
            raise ValueError("Некорректное имя")

        if isinstance(age, int) and 0 <= age <= 110:
            self._age = age
        else:
            raise ValueError("Некорректный возраст")
    
    def get_name(self) -> str:
        return self._name
    
    def set_name(self, name: str) -> None:
        if isinstance(name, str) and name.isalpha():
            self._name = name
        else:
            raise ValueError("Некорректное имя")
    
    def get_age(self) -> int:
        return self._age
    
    def set_age(self, new_age: int) -> None:
        if isinstance(new_age, int) and 0 <= new_age <= 110:
            self._age = new_age
        else:
            raise ValueError("Некорректный возраст")
```

# 4.5 Свойства, функция property()

```python
# Date: 31.01.2024
# Step: 4.5.13
# Url: https://stepik.org/lesson/796462/step/13?unit=799267

class Rectangle:
    def __init__(self, length: int, width: int) -> None:
        self.length = length
        self.width = width
    
    def get_perimeter(self):
        return self.length * 2 + self.width * 2
    
    def get_area(self):
        return self.length * self.width
    
    perimeter = property(get_perimeter)
    area = property(get_area)
```

```python
# Date: 01.02.2024
# Step: 4.5.14
# Url: https://stepik.org/lesson/796462/step/14?unit=799267

class HourClock:
    def __init__(self, hours: int) -> None:
        if isinstance(hours, int) and 1 <= hours <= 12:
            self._hours = hours
        else:
            raise ValueError("Некорректное время")
    
    def get_hours(self) -> int:
        return self._hours
    
    def set_hours(self, new_hours: int) -> None:
        if isinstance(new_hours, int) and 1 <= new_hours <= 12:
            self._hours = new_hours
        else:
            raise ValueError("Некорректное время")
    
    hours = property(get_hours, set_hours)
```

```python
clock = HourClock(13)
```

```python
# Date: 01.02.2024
# Step: 4.5.15
# Url: https://stepik.org/lesson/796462/step/15?unit=799267

class Person:
    def __init__(self, name: str, surname: str) -> None:
        self.name = name
        self.surname = surname
    
    def get_fullname(self) -> str:
        return f"{self.name} {self.surname}"
    
    def set_fullname(self, new_fullname: str) -> None:
        self.name, self.surname = new_fullname.split()
    
    fullname = property(get_fullname, set_fullname)
```

```python
class Cat:
    def __init__(self, breed, name):
        self.breed = breed
        self.name = name

    def get_info(self):
        return self.breed + ' ' + self.name

    def set_info(self, info):
        self.breed, self.name = info.split()

    info = property(get_info, set_info)
```

```python
person = Person('Меган', 'Фокс')

person.name = 'Стефани'
print(person.fullname)
```
