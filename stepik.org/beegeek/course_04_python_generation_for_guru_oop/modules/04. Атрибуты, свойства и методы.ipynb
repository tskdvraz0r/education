```python
# Date: 28.01.2024
# Step: 4.3.12
# Url: https://stepik.org/lesson/794582/step/12?unit=797335

class TextHandler:
    def __init__(self) -> None:
        self.words: list = []
    
    def add_words(self, text: str) -> None:
        self.words.extend(text.split())
    
    def get_shortest_words(self):# -> list:
        if len(self.words) == 0:
            return self.words
        min_word_len: int = len(min(self.words, key = len))
        return [i for i in self.words if len(i) == min_word_len]
    
    def get_longest_words(self):# -> list:
        if len(self.words) == 0:
            return self.words
        max_word_len: int = len(max(self.words, key = len))
        return [i for i in self.words if len(i) == max_word_len]
```

```python
# Date: 28.01.2024
# Step: 4.3.12
# Url: https://stepik.org/lesson/794582/step/12?unit=797335

class Todo:
    def __init__(self) -> None:
        self.things: list = []
    
    def add(self, task: str, priority: int) -> None:
        self.things.append((task, priority))
    
    def get_by_priority(self, n: int):# -> list:
        if len(self.things) == 0:
            return self.things
        return [i[0] for i in self.things if i[1] == n]
    
    def get_low_priority(self):# -> list:
        if len(self.things) == 0:
            return self.things
        priority: int = min(self.things, key = lambda x: x[1])[1]
        return self.get_by_priority(priority)
    
    def get_high_priority(self):# -> list:
        if len(self.things) == 0:
            return self.things
        priority: int = max(self.things, key = lambda x: x[1])[1]
        return self.get_by_priority(n = priority)
```

```python
# Date: 28.01.2024
# Step: 4.3.14
# Url: https://stepik.org/lesson/794582/step/14?unit=797335

class Postman:
    def __init__(self) -> None:
        self.delivery_data: list = []
    
    def add_delivery(self, street: str, house: int, flat: int) -> None:
        self.delivery_data.append((street, house, flat))
    
    def get_houses_for_street(self, street: str) -> list:
        houses: list = []
        for house in [i for i in self.delivery_data if i[0] == street]:
            if house[1] not in houses:
                houses.append(house[1])
        
        return houses
    
    def get_flats_for_house(self, street: str, house: int) -> list:
        flats: list = []
        for flat in [i for i in self.delivery_data if i[0] == street and i[1] == house]:
            if flat[2] not in flats:
                flats.append(flat[2])
        
        return flats
```

```python
# Date: 31.01.2024
# Step: 4.4.16
# Url: https://stepik.org/lesson/793864/step/16?unit=796596

class BankAccount:
    def __init__(self, balance: int = 0) -> None:
        self._balance: int = balance
    
    def get_balance(self) -> int:
        return self._balance
    
    def deposit(self, amount: int) -> None:
        self._balance += amount
    
    def withdraw(self, amount: int) -> None:
        if amount > self._balance:
            raise ValueError("На счете недостаточно средств")
        
        self._balance -= amount
    
    def transfer(self, account, amount: int) -> None:
        self.withdraw(amount = amount)
        account.deposit(amount = amount)
```

# 4.6 Декоратор @property

```python
# Date: 01.02.2024
# Step: 4.6.13
# Url: https://stepik.org/lesson/801862/step/13?unit=804839

class Person:
    def __init__(self, name: str, surname: str) -> None:
        self.name: str = name
        self.surname: str = surname
    
    @property
    def fullname(self) -> str:
        return f"{self.name} {self.surname}"
    
    @fullname.setter
    def fullname(self, new_fullname) -> None:
        self.name, self.surname = new_fullname.split()
```

```python
# Date: 01.02.2024
# Step: 4.6.14
# Url: https://stepik.org/lesson/801862/step/14?unit=804839

class Account:
    def __init__(self, login: str, password: str) -> None:
        self._login: str = login
        self._password: int = Account.hash_function(password = password)
    
    @staticmethod
    def hash_function(password: str):
        hash_value = sum(
            ord(char) * index
            for char, index in zip(password, range(len(password)))
        )
        return hash_value % 10**9
    
    @property
    def login(self) -> str:
        return self._login
    
    @login.setter
    def login(self, login: str) -> None:
        raise AttributeError("Изменение логина невозможно")
    
    @property
    def password(self) -> str:
        return self._password
    
    @password.setter
    def password(self, password: str) -> None:
        hash_password = Account.hash_function(password = password)
        self._password = hash_password
        
```

```python
# Date: 01.02.2024
# Step: 4.7.15
# Url: https://stepik.org/lesson/794698/step/15?unit=797450

class StrExtension:
    
    def __init__(self) -> None:
        pass
    
    @classmethod
    def remove_vowels(cls, string: str) -> str:
        for char in ("a", "e", "i", "o", "u", "y"):
            string = (
                string
                .replace(char, "")
                .replace(char.upper(), "")
            )
        
        return string
    
    @classmethod
    def leave_alpha(cls, string: str) -> str:
        for char in string:
            if not char.isalpha():
                string = string.replace(char, "")
        
        return string
    
    @classmethod
    def replace_all(cls, string: str, chars: str, char: str) -> str:
        for i in chars:
            string = string.replace(i, char)
        
        return string
```

# 4.8 Декоратор @singledispatchmethod

```python
# Date: 01.02.2024
# Step: 4.8.10
# Url: https://stepik.org/lesson/802381/step/10?unit=805389

# IMPORTS
from functools import singledispatchmethod

class Processor:
    
    @singledispatchmethod
    def process(data):
        raise TypeError("Аргумент переданного типа не поддерживается")
    
    @process.register(int)
    @process.register(float)
    def from_int_float(data):
        return data * 2
    
    @process.register(str)
    def from_str(data):
        return data.upper()
    
    @process.register(list)
    def from_list(data) -> list:
        return sorted(data)
    
    @process.register(tuple)
    def from_tuple(data) -> tuple:
        return tuple(sorted(data))    
```

```python
# Date: 01.02.2024
# Step: 4.8.11
# Url: https://stepik.org/lesson/802381/step/11?unit=805389

# IMPORTS
from functools import singledispatchmethod

class Formatter:
    @singledispatchmethod
    def format(self) -> None:
        raise TypeError("Аргумент переданного типа не поддерживается")
    
    @format.register(int)
    def from_int(self) -> None:
        print(f"Целое число: {self}")
    
    @format.register(float)
    def from_float(self) -> None:
        print(f"Вещественное число: {self}")
    
    @format.register(tuple)
    def from_tuple(self) -> None:
        print(f"Элементы кортежа: {', '.join(str(i) for i in self)}")
    
    @format.register(list)
    def from_list(self) -> None:
        print(f"Элементы списка: {', '.join(str(i) for i in self)}")
    
    @format.register(dict)
    def from_dict(self) -> None:
        print(f"Пары словаря: {', '.join([str(i) for i in a.items()])}")
```
