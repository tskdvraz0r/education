```python
%%ipytest
# Date: 18.01.2024
# Step: 2.2.15
# Url: https://stepik.org/lesson/265077/step/15?unit=246025

# Functions
def repeat_after_me_1(
    strings: list[str],
    is_stdin: bool = False
) -> list[str]:
    """
    Note:
        Функция принимает на вход список со строками и возвращает их в той же последовательности.

    Args:
        strings (list[str]): Список со строками.
        is_stdin (bool, optional): Принимаются ли данные из "потока". По умолчанию False.
    
    Raises:
        TypeError(f"Аргумент ожидается тип данных {value_type}.")
        TypeError("Элемент ожидает тип данных 'str'.")

    Returns:
        list[str]: Список со строками в той же последовательности.
    """

    # DATA VALIDATION
    for value, value_type in zip((strings, is_stdin), (list, bool)):
        if not isinstance(value, value_type):
            raise TypeError(f"Аргумент ожидается тип данных {value_type}.")

    if not all(isinstance(i,str) for i in strings):
        raise TypeError("Элемент ожидает тип данных 'str'.")

    if is_stdin:
        strings = [i.replace("\n", "") for i in strings]
    
    # MAIN ALGORITHM
    return strings

# Tests
@pytest.mark.parametrize(
    ["test_strings", "test_is_stdin", "expected_result"],
    [
        (["I was", "born", "this way"], False, ["I was", "born", "this way"]),
        (["I love", "Python", "so much"], False, ["I love", "Python", "so much"]),
        (["Меня зовут", "Тимур", "Я люблю математику"], False, ["Меня зовут", "Тимур", "Я люблю математику"]),

        (["I was\n", "born\n", "this way"], True, ["I was", "born", "this way"]),
        (["I love\n", "Python\n", "so much"], True, ["I love", "Python", "so much"]),
        (["Меня зовут\n", "Тимур\n", "Я люблю математику"], True, ["Меня зовут", "Тимур", "Я люблю математику"])
    ]
)
def test_good_cases_repeat_after_me_1(test_strings, test_is_stdin, expected_result) -> None:
    assert repeat_after_me_1(
        strings = test_strings,
        is_stdin = test_is_stdin
    ) == expected_result

# Main Program
if __name__ == "__main__":
    print(
        *repeat_after_me_1(
            strings = ["Я", "изучаю", "python"],
            is_stdin = False
        ),
        sep = "\n"
    )
```

```python
%%ipytest
# Date: 18.01.2024
# Step: 2.3.7
# Url: https://stepik.org/lesson/275252/step/7?unit=256355

# Functions
def custom_separator(
    sep_and_strings: list[str],
    is_stdin: bool = False
) -> str:
    """
    Note:
        Функция принимает на вход список с разделителем и строками, и возвращает данные строки через указанный разделитель.

    Args:
        sep_and_strings (list[str]): Список с разделителем и строками.
        is_stdin (bool, optional): Используется ли "поток" данных. По умолчанию False.
    
    Raises:
        TypeError(f"Аргумент ожидается тип данных {value_type}.")
        TypeError("Элемент ожидает тип данных 'str'.")
        
    Returns:
        str: Последовательность строк через указанный разделитель.
    """

    # DATA VALIDATION
    for value, value_type in zip((sep_and_strings, is_stdin), (list, bool)):
        if not isinstance(value, value_type):
            raise TypeError(f"Аргумент ожидается тип данных {value_type}.")

    if is_stdin:
        sep_and_strings = [i.replace("\n", "") for i in sep_and_strings]

    if not all(isinstance(x, str) for x in sep_and_strings):
        raise TypeError("Элемент ожидает тип данных 'str'.")

    # Разделить данные по переменным
    separator = sep_and_strings.pop(0)
    list_with_strings = sep_and_strings

    return separator.join(list_with_strings)


# Tests
@pytest.mark.parametrize(
    ["test_sep_and_strings", "test_is_stdin", "expected_result"],
    [
        (["**", "I", "like", "Python"], False, "I**like**Python"),
        (["-", "I", "like", "Python"], False, "I-like-Python"),
        (["->", "I", "like", "Python"], False, "I->like->Python"),

        (["**\n", "I\n", "like\n", "Python"], True, "I**like**Python"),
        (["-\n", "I\n", "like\n", "Python"], True, "I-like-Python"),
        (["->\n", "I\n", "like\n", "Python"], True, "I->like->Python")
    ]
)
def test_good_cases_custom_separator(test_sep_and_strings, test_is_stdin, expected_result) -> None:
    assert custom_separator(
        sep_and_strings = test_sep_and_strings,
        is_stdin = test_is_stdin
    ) == expected_result


# Main Program
if __name__ == "__main__":
    print(
        custom_separator(
            sep_and_strings = ["_", "first", "second", "third"],
            is_stdin = False
        )
    )
```

```python
%%ipytest
# Date: 18.01.2024
# Step: 2.4.6
# Url: https://stepik.org/lesson/265079/step/6?unit=246027

# Functions
def three_consecutive_numbers(number: int) -> tuple[int, int, int]:
    """
    Notes:
        Функция принимает на вход целочисленной число и возвращает последовательность из 3 чисел, начинающуюся с указаного числа.

    Args:
        number (int): Целочисленное число.

    Raises:
        TypeError: "Аргумент ожидает тип 'int'."

    Returns:
        tuple[int, int, int]: Три последовательных числа.
    """

    # DATA VALIDATION
    if not isinstance(number, int):
        raise TypeError("Аргумент ожидает тип 'int'.")
    
    # MAIN ALGORITHM
    return number, number + 1, number + 2

# Tests
@pytest.mark.parametrize(
    ["test_number", "expected_result"],
    [
        (1, (1, 2, 3)),
        (0, (0, 1, 2)),
        (-1, (-1, 0, 1))
    ]
)
def test_good_cases_three_consecutive_numbers(test_number, expected_result) -> None:
    assert three_consecutive_numbers(
        number = test_number
    ) == expected_result

# Main program
if __name__ == "__main__":
    print(
        *three_consecutive_numbers(
            number = 3
        ),
        sep = "\n"
    )
```

```python
%%ipytest
# Date: 18.01.2024
# Step: 2.4.7
# Url: https://stepik.org/lesson/265079/step/7?unit=246027

# Functions
def sum_of_three_numbers(
    numbers: list[int, int, int] | list[str, str, str],
    is_stdin: bool = False
) -> int:
    """
    Notes:
        Функция принимает на вход список с тремя целыми числами и возвращает сумму.

    Args:
        numbers (list[int, int, int] | list[str, str, str]): Список с целыми числами.
        is_stdin (bool, optional): Используется ли поток данных. По умолчанию False.

    Raises:
        TypeError(f"Аргумент ожидает тип данных {value_type}.")
        IndexError("Аргумент ожидает 3 элемента.")
        TypeError("Элемент ожидает тип данных 'int'.")

    Returns:
        int: Сумма трех чисел.
    """

    # DATA VALIDATION
    for value, value_type in zip(
        (numbers, is_stdin),
        (list, bool)
    ):
        if not isinstance(value, value_type):
            raise TypeError(f"Аргумент ожидает тип данных {value_type}.")
    
    if len(numbers) != 3:
        raise IndexError("Аргумент ожидает 3 элемента.")

    if is_stdin:
        numbers: list[int, int, int] = [int(i) for i in numbers]

    if not all(isinstance(i, int) for i in numbers):
        raise TypeError("Элемент ожидает тип данных 'int'.")

    # MAIN ALGORITHM
    return sum(numbers)

# Tests
@pytest.mark.parametrize(
    ["test_numbers", "test_is_stdin", "expected_result"],
    [
        ([9, 11, 2], False, 22),
        ([-1, 10, 1], False, 10),
        ([-7, -10, -3], False, -20),
        
        (["9\n", "11\n", "2"], True, 22),
        (["-1\n", "10\n", "1"], True, 10),
        (["-7\n", "-10\n", "-3"], True, -20)
    ]
)
def test_good_cases_sum_of_three_numbers(test_numbers, test_is_stdin, expected_result) -> None:
    assert sum_of_three_numbers(
        numbers = test_numbers,
        is_stdin = test_is_stdin
    ) == expected_result

# Main program
if __name__ == "__main__":
    print(
        sum_of_three_numbers(
            numbers = [1, 2, 3],
            is_stdin = False
        )
    )
```

```python
%%ipytest
# Date: 18.01.2024
# Step: 2.4.13
# Url: https://stepik.org/lesson/265079/step/13?unit=246027

# Functions
def arithmetic_progression(
    numbers: list[int, int, int] | list[str, str, str],
    is_stdin: bool = False
) -> int:
    """
    Note:
        Функция принимает на вход три целых числа и возвращает арифметическую прогрессиию.
        
        Формула арифметической прогрессии: a_(n) = a_(n - 1) + d

    Args:
        numbers (list[int]): Список с целыми числами.
        is_stdin (bool, optional): Используется ли "поток" данных. По умолчанию False.
    
    Raises:
        TypeError(f"Аргумент ожидает тип данных {value_type}.")
        IndexError("Аргумент ожидает 3 элемента.")
        TypeError("Элемент ожидает тип данных 'int'.")

    Returns:
        int: Арифметическая прогрессия.
    """

    # DATA VALIDATION
    for value, value_type in zip(
        (numbers, is_stdin),
        (list, bool)
    ):
        if not isinstance(value, value_type):
            raise TypeError(f"Аргумент ожидает тип данных {value_type}.")
    
    if len(numbers) != 3:
        raise IndexError("Аргумент ожидает 3 элемента.")
    
    if is_stdin:
        numbers = [int(i) for i in numbers]
        
    if not all(isinstance(i, int) for i in numbers):
        raise TypeError("Элемент ожидает тип данных 'int'.")
    
    # MAIN ALGORITHM
    a, d, n = numbers

    return a + d * (n - 1)


# Tests
@pytest.mark.parametrize(
    ["test_numbers", "test_is_stdin", "expected_result"],
    [
        ([1, 1, 10], False, 10),
        ([-1, 1, 2], False, 0),
        ([100, 50, 1], False, 100),

        (["1\n", "1\n", "10"], True, 10),
        (["-1\n", "1\n", "2"], True, 0),
        (["100\n", "50\n", "1"], True, 100)
    ]
)
def test_good_cases_arithmetic_progression(test_numbers, test_is_stdin, expected_result) -> None:
    assert arithmetic_progression(
        numbers = test_numbers,
        is_stdin = test_is_stdin
    ) == expected_result

# Main Program
if __name__ == "__main__":
    print(
        arithmetic_progression(
            numbers = [1, 2, 10],
            is_stdin = True
        )
    )
```
