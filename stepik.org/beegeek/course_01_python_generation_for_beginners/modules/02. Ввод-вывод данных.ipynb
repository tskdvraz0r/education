```python
%%ipytest
# Date: 18.01.2024
# Step: 2.2.13
# Url: https://stepik.org/lesson/265077/step/13?unit=246025

# Functions
def greeting(name: str) -> str:
    """
    Notes:
        Функция принимает на вход имя пользователя и приветствует его по шаблону: f"Привет, {name}".

    Args:
        name (str): Имя пользователя.
    
    Raises:
        TypeError(f"Аргумент ожидается тип данных 'str'.")

    Returns:
        str: Строка приветствия пользователя.
    """
    
    # DATA VALIDATION
    if not isinstance(name, str):
        raise TypeError(f"Аргумент ожидается тип данных 'str'.")
    
    # MAIN ALGORITHM
    return f"Привет, {name}"

# Tests
@pytest.mark.parametrize(
    ["test_name", "expected_result"],
    [
        ("Тимур", "Привет, Тимур"),
        ("Гвидо", "Привет, Гвидо"),
        ("Питоняшка", "Привет, Питоняшка")
    ]
)
def test_good_cases_greeting(test_name, expected_result) -> None:
    assert greeting(name = test_name) == expected_result

# Main program
if __name__ == "__main__":
    print(
        greeting(name = "Путник")
    )
```

```python
%%ipytest
# Date: 18.01.2024
# Step: 2.2.15
# Url: https://stepik.org/lesson/265077/step/15?unit=246025

# Functions
def repeat_after_me_1(
    strings: list[str],
    is_stdin: bool = False
) -> list[str]:
    """
    Note:
        Функция принимает на вход список со строками и возвращает их в той же последовательности.

    Args:
        strings (list[str]): Список со строками.
        is_stdin (bool, optional): Принимаются ли данные из "потока". По умолчанию False.
    
    Raises:
        TypeError(f"Аргумент ожидается тип данных {value_type}.")
        TypeError("Элемент ожидает тип данных 'str'.")

    Returns:
        list[str]: Список со строками в той же последовательности.
    """

    # DATA VALIDATION
    for value, value_type in zip((strings, is_stdin), (list, bool)):
        if not isinstance(value, value_type):
            raise TypeError(f"Аргумент ожидается тип данных {value_type}.")

    for element in strings:
        if not isinstance(element, str):
            raise TypeError("Элемент ожидает тип данных 'str'.")

    if is_stdin:
        strings = [i.replace("\n", "") for i in strings]
    
    # MAIN ALGORITHM
    return strings

# Tests
@pytest.mark.parametrize(
    ["test_strings", "test_is_stdin", "expected_result"],
    [
        (["I was", "born", "this way"], False, ["I was", "born", "this way"]),
        (["I love", "Python", "so much"], False, ["I love", "Python", "so much"]),
        (["Меня зовут", "Тимур", "Я люблю математику"], False, ["Меня зовут", "Тимур", "Я люблю математику"]),

        (["I was\n", "born\n", "this way"], True, ["I was", "born", "this way"]),
        (["I love\n", "Python\n", "so much"], True, ["I love", "Python", "so much"]),
        (["Меня зовут\n", "Тимур\n", "Я люблю математику"], True, ["Меня зовут", "Тимур", "Я люблю математику"])
    ]
)
def test_good_cases_repeat_after_me_1(test_strings, test_is_stdin, expected_result) -> None:
    assert repeat_after_me_1(
        strings = test_strings,
        is_stdin = test_is_stdin
    ) == expected_result

# Main Program
if __name__ == "__main__":
    print(
        *repeat_after_me_1(
            strings = ["Я", "изучаю", "python"],
            is_stdin = False
        ),
        sep = "\n"
    )
```

```python
%%ipytest
# Date: 18.01.2024
# Step: 2.2.16
# Url: https://stepik.org/lesson/265077/step/16?unit=246025

# Functions
def repeat_after_me_2(
    strings: list[str],
    is_stdin: bool = False
) -> list[str]:
    """
    Note:
        Функция принимает на вход список со строками и возвращает их в обратной последовательности.

    Args:
        strings (list[str]): Список со строками.
        is_stdin (bool, optional): Принимаются ли данные из "потока". По умолчанию False.
    
    Raises:
        TypeError(f"Аргумент ожидается тип данных {value_type}.")
        TypeError("Элемент ожидает тип данных 'str'.")
    
    Returns:
        list[str]: Список со строками в обратной последовательности.
    """
    
    # DATA VALIDATION
    for value, value_type in zip((strings, is_stdin), (list, bool)):
        if not isinstance(value, value_type):
            raise TypeError(f"Аргумент ожидается тип данных {value_type}.")

    for element in strings:
        if not isinstance(element, str):
            raise TypeError("Элемент ожидает тип данных 'str'.")

    if is_stdin:
        strings = [i.replace("\n", "") for i in strings]
    
    # MAIN ALGORITHM
    return strings[::-1]

# Tests
@pytest.mark.parametrize(
    ["test_strings", "test_is_stdin", "expected_result"],
    [
        (["I was", "born", "this way"], False, ["this way", "born", "I was"]),
        (["I love", "Python", "so much"], False, ["so much", "Python", "I love"]),
        (["Меня зовут", "Тимур", "Я люблю математику"], False, ["Я люблю математику", "Тимур", "Меня зовут"]),

        (["I was\n", "born\n", "this way"], True, ["this way", "born", "I was"]),
        (["I love\n", "Python\n", "so much"], True, ["so much", "Python", "I love"]),
        (["Меня зовут\n", "Тимур\n", "Я люблю математику"], True, ["Я люблю математику", "Тимур", "Меня зовут"])
    ]
)
def test_good_cases_repeat_after_me_2(test_strings, test_is_stdin, expected_result) -> None:
    assert repeat_after_me_2(
        strings = test_strings,
        is_stdin = test_is_stdin
    ) == expected_result


# Main Program
if __name__ == "__main__":
    print(
        *repeat_after_me_2(
            strings = ["Я", "изучаю", "python"],
            is_stdin = False
        ),
        sep = "\n"
    )
```
