```python
import sys

import pytest
import ipytest

ipytest.autoconfig()
```

# Step 12

```python
%%ipytest

def i_like_python() -> str:
    """
    Notes:
        Функция возвращает список строк ["I", "like", "Python"].

    Returns:
        str: Список строк ["I", "like", "Python"]
    
    Examples:
        >>> i_like_python()
        >>> ["I", "like", "Python"]
    """
    
    return ["I", "like", "Python"]

@pytest.mark.parametrize(
    ["expected_result"],
    [
        (["I", "like", "Python"], )
    ]
)
def test_good_case_i_like_python(expected_result) -> None:
    i_like_python() == expected_result

if __name__ == "__main__":
    print(
        *i_like_python(),
        sep="***"
    )
```

```python
%%ipytest

def greatings(name: str) -> str:
    """
    Notes:
        Функция принимает на вход имя и возвращает строку приветствия.
    Args:
        name (str): Имя.

    Returns:
        str: Строка приветствия вида: f"Привет, {name}".
    
    Examples:
        >>> greatings(name="ПитоняшкаЭ)
        >>> "Привет, Питоняшка"
    """
    
    return f"Привет, {name}"

@pytest.mark.parametrize(
    ["test_name", "expected_result"],
    [
        ("Тирион", "Привет, Тирион"),
        ("Тимуп", "Привет, Тимур"),
        ("Джон", "Привет, Джон"),
    ]
)
def test_good_case_greatings(
        test_name,
        expected_result
) -> None:
    greatings(name=test_name) == expected_result

if __name__ == "__main__":
    print(
        greatings(name="Питоняшка"),
        end="!"
    )

    print("\n")
```

```python
%%ipytest

def custom_sep(
        sep_and_strings: list[str],
        is_stdin: bool = False
) -> str:
    """
    Notes:
        Функиця считывает данные из потока или принимает на вход список со строками   

    Args:
        sep_and_strings (list[str]): Список с разделителем и строками.
        is_stdin (bool, optional): Используется ли поток данных. По умолчанию False.

    Returns:
        str: Строки, разделённые "разделителем".
    
    Examples:
        >>> custom_sep(sep_and_strings=["*", "Раз", "Два", "Три"], is_stdin=False)
        >>> "Раз*Два*Три"
        
        >>> custom_sep(sep_and_strings=["\n", "Money often\n", "costs\n", "too much"], is_stdin=True)
        >>> "Money often costs too much"
    """
    
    if is_stdin:
        sep_and_strings = [i.replace("\n", "") for i in sep_and_strings]

    separator: str = sep_and_strings.pop(0)

    return separator.join(sep_and_strings)

@pytest.mark.parametrize(
    ["test_sep_and_strings", "test_is_stdin", "expected_result"],
    [
        (["*", "Раз", "Два", "Три"], False, "Раз*Два*Три"),
        (["\n", "Money often\n", "costs\n", "too much"], True, "Money often costs too much")
    ]
)
def test_good_cases_custom_sep(
        test_sep_and_strings,
        test_is_stdin,
        expected_result
) -> None:
    custom_sep(
        sep_and_strings=test_sep_and_strings,
        is_stdin=test_is_stdin
    ) == expected_result

if __name__ == "__main__":
    print(
        custom_sep(
            sep_and_strings=["-*-", "Python", "C", "C++"],
            is_stdin=True
        )
    )
```
