```python
%%ipytest

def star_triangle() -> list[str]:
    """
    Notes:
        Функция возвращает список со строками, который при распаковке образует треугольник.

    Returns:
        list[str]: Список со строками ["*", "**", "***", ...].
    
    Examples:
        >>> star_triangle()
        >>> ["*", "**", "***", "****", "*****", "******", "*******"]
    """
    
    return ["*" * i for i in range(1, 8)]

@pytest.mark.parametrize(
    ["expected_result"],
    [(["*", "**", "***", "****", "*****", "******", "*******"],)]
)
def test_good_case_star_triangle(expected_result) -> None:
    assert star_triangle() == expected_result

if __name__ == "__main__":
    print(*star_triangle(), sep="\n")
```

```python
%%ipytest

def greatings(name: str) -> str:
    """
    Notes:
        Функция принимает на вход имя и возвращает строку приветствия.

    Args:
        name (str): Имя.

    Returns:
        str: Строка приветствия: f"Привет, {name}".
    
    Examples:
        >>> greatings(name="Питоняшка")
        >>> "Привет, Питоняшка"
    """
    
    return f"Привет, {name}"

@pytest.mark.parametrize(
    ["test_name", "expected_result"],
    [
        ("Тимур", "Привет, Тимур"),
        ("Гвидо", "Привет, Гвидо"),
        ("Python", "Привет, Python")
    ]
)
def test_good_case_greatings(
        test_name,
        expected_result
) -> None:
    assert greatings(name=test_name) == expected_result

if __name__ == "__main__":
    print(
        greatings(name="Питоняшка")
    )
```

# Step 23

```python
%%ipytest

def repeat_after_me_1(
        strings: list[str],
        is_stdin: bool = False
) -> list[str]:
    """
    Notes:
        Функция принимает на вход список со строками или считывает данные из потока данных. 
        В первом случае - возвращает список неизменным, во втором, очищает от символов переноса 
        строки.

    Args:
        strings (list[str]): Список со строками.
        is_stdin (bool, optional): Используется ли поток данных. По умолчанию False.

    Returns:
        list[str]: Список со строками, очищенные от символов переноса строки.
    
    Examples:
        >>> repeat_after_me_1(strings=["I was", "born", "this way"], is_stdin=False)
        >>> ["I was", "born", "this way"]
        
        >>> repeat_after_me_1(strings=["I was\n", "born\n", "this way"], is_stdin=True)
        >>> ["I was", "born", "this way"]
    """
    
    return (
        [i.replace("\n", "") for i in strings]
        if is_stdin
        else strings
    )

@pytest.mark.parametrize(
    ["test_strings", "test_is_stdin", "expected_result"],
    [
        (["I was", "born", "this way"], False, ["I was", "born", "this way"]),
        (["I was\n", "born\n", "this way"], True, ["I was", "born", "this way"])
    ]
)
def test_good_case_repeat_after_me_1(
    test_strings,
    test_is_stdin,
    expected_result
) -> None:
    assert repeat_after_me_1(
        strings=test_strings,
        is_stdin=test_is_stdin
    ) == expected_result

if __name__ == "__main__":
    print(
        *repeat_after_me_1(
            strings=["I was", "born", "this way"],
            is_stdin=True
        ),
        sep="\n"
    )
```

```python
%%ipytest

def repeat_after_me_2(
        strings: list[str],
        is_stdin: bool = False
) -> list[str]:
    """
    Notes:
        Функция принимает на вход список со строками или считывает данные из потока данных. 
        В первом случае - возвращает список в обратном порядке, во втором, очищает от символов переноса 
        строки и возвращает в обратном порядке.

    Args:
        strings (list[str]): Список со строками.
        is_stdin (bool, optional): Используется ли поток данных. По умолчанию False.

    Returns:
        list[str]: Список со строками, очищенные от символов переноса строки в обратном порядке.
    
    Examples:
        >>> repeat_after_me_1(strings=["I was", "born", "this way"], is_stdin=False)
        >>> ["this way", "born", "I was"]
        
        >>> repeat_after_me_1(strings=["I was\n", "born\n", "this way"], is_stdin=True)
        >>> ["this way", "born", "I was"]
    """
    
    
    return (
        [i.replace("\n", "") for i in strings][::-1]
        if is_stdin
        else strings[::-1]
    )

@pytest.mark.parametrize(
    ["test_strings", "test_is_stdin", "expected_result"],
    [
        (["I was", "born", "this way"], False, ["this way", "born", "I was"]),
        (["I was\n", "born\n", "this way"], True, ["this way", "born", "I was"])
    ]
)
def test_good_cases_repeat_after_me_2(
        test_strings,
        test_is_stdin,
        expected_result
) -> None:
    assert repeat_after_me_2(
        strings=test_strings,
        is_stdin=test_is_stdin
    ) == expected_result

if __name__ == "__main__":
    print(
        *repeat_after_me_2(
            strings=["I was", "born", "this way"],
            is_stdin=True
        ),
        sep="\n"
    )
```
